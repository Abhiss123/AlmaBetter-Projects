### **What are Graph Neural Networks (GNNs)?**

Graph Neural Networks (GNNs) are a type of machine learning model that works well with data structured in the form of *graphs*. In a graph, things like pages, users, or keywords can be represented as *nodes*, and relationships between them (like links between pages or user clicks) are represented as *edges*. GNNs help in analyzing and learning from this kind of data.

### **How do GNNs apply to SEO?**

In **SEO (Search Engine Optimization),** GNNs can be used to understand and model complex relationships between different elements like:
1. **Content**: How pieces of content on a website are related.
2. **User behavior**: How users interact with the website, including which pages they visit, how long they stay, and where they click.
3. **Link structures**: The internal and external links between pages.

A GNN can analyze these relationships to optimize the website, making it easier for search engines like Google to understand the site’s content and its relevance to user queries. This can boost the website’s ranking on search results.

````python
# URLs of the website pages you want to analyze
urls = [
    'https://thatware.co/',  # Homepage
    'https://thatware.co/services/',  # Services page
    'https://thatware.co/advanced-seo-services/',  # Advanced SEO services page
    # Add more URLs as needed
]

# Step 1: Function to extract links from each URL
# This function fetches the content of each page, parses the HTML, and retrieves all internal and external links.
def extract_links(url):
    try:
        # Step 1.1: Send a GET request to the URL to retrieve the page content
        # The 'requests.get()' function makes an HTTP request to the given URL and gets the HTML content.
        response = requests.get(url)

        # Step 1.2: Parse the HTML content using BeautifulSoup
        # BeautifulSoup parses the page content (response.text) and makes it easy to extract data from HTML.
        soup = BeautifulSoup(response.text, 'html.parser')

        # Step 1.3: Initialize an empty set to store links
        # A set is used here to store links because it automatically removes duplicates.
        links = set()

        # Step 1.4: Find all <a> tags (anchor tags) in the HTML that contain the href attribute
        # Each <a> tag represents a link, and the href attribute contains the URL.
        for link in soup.find_all('a', href=True):
            href = link['href']  # Extract the URL from the href attribute

            # Step 1.5: Filter links that start with 'http' or '/', which means they are valid web links
            # 'http' identifies external links, and '/' identifies internal links.
            if href.startswith('http') or href.startswith('/'):
                # Step 1.6: Add the link to the set
                links.add(href)

        # Step 1.7: Return the set of extracted links for this URL
        return links

    except Exception as e:
        # If any error occurs during the fetching or parsing of the page, print the error and return an empty set
        print(f"Error fetching or parsing the URL {url}: {e}")
        return set()

# Example: To run the function on one of the URLs, let's print the links found on the homepage
for url in urls:
    print(f"Links found on {url}:")
    print(extract_links(url))
    print("\n")
````

````python
# Import necessary libraries
import networkx as nx  # For building graph structures (nodes/edges)
import requests        # To fetch webpage content
from bs4 import BeautifulSoup  # To extract text and links from webpages
import matplotlib.pyplot as plt  # For visualizing the graph

# Step 2: Create a graph to represent the website structure
def build_graph(urls):
    G = nx.Graph()  # This initializes an empty graph where we will add nodes (webpages) and edges (links)

    # Step 2.1: Loop through each URL in the list of URLs
    for url in urls:
        G.add_node(url)  # This adds the URL as a "node" in the graph

        # Step 2.2: Extract all links from the current URL
        links = extract_links(url)

        # Step 2.3: For each link found on the page, check if it's an internal link
        for link in links:
            # Only add internal links (those starting with 'https://thatware.co' or '/')
            if link.startswith('https://thatware.co') or link.startswith('/'):
                full_link = link if link.startswith('http') else f'https://thatware.co{link}'

                # Add an "edge" between the current URL and the linked page
                G.add_edge(url, full_link)

    # Step 2.4: Return the complete graph and print the connections
    print("\nGraph Nodes (Pages):")
    print(G.nodes)
    print("\nGraph Edges (Links):")
    print(G.edges)

    return G

# Step 3: Build and visualize the graph
graph = build_graph(urls)

# Optional: Visualize the graph
nx.draw(graph, with_labels=True, node_size=1000, node_color='lightblue', font_size=10, font_weight='bold')
plt.show()
````

````python
# Step 4: Define the GNN Model
class GNNModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(GNNModel, self).__init__()
        self.conv1 = nn.Linear(input_dim, hidden_dim)  # First graph layer
        self.conv2 = nn.Linear(hidden_dim, output_dim)  # Second graph layer

    def forward(self, x, adj):
        x = torch.relu(self.conv1(torch.mm(adj, x)))  # First layer with ReLU
        x = torch.relu(self.conv2(torch.mm(adj, x)))  # Second layer with ReLU
        return x

# Step 5: Prepare dummy data for testing the GNN
# Here we assume each node has 3 features, and we create a simple adjacency matrix to represent the graph connections.
# This is just a placeholder for demonstration purposes.

# Number of nodes (pages)
num_nodes = len(G.nodes)

# Create dummy feature matrix (num_nodes x 3) where each node has 3 features
features = torch.randn(num_nodes, 3)

# Create a simple adjacency matrix (num_nodes x num_nodes) where the graph's structure is encoded
adj_matrix = torch.eye(num_nodes)  # Here we use an identity matrix for simplicity

# Initialize the GNN model with 3 input features, 4 hidden features, and 2 output features
model = GNNModel(input_dim=3, hidden_dim=4, output_dim=2)

# Step 6: Run the model and display the output
output = model(features, adj_matrix)
print("\nGNN Model Output (for each page):")
print(output)
````

````python
# Import necessary libraries
import networkx as nx  # For building graph structures (nodes/edges)
import requests        # To fetch webpage content
from bs4 import BeautifulSoup  # To extract text and links from webpages
import torch           # PyTorch for machine learning and GNNs
import torch.nn as nn  # Neural network layers for GNN
import numpy as np     # For working with matrices and arrays

# URLs of the website pages you provided
urls = [
    'https://thatware.co/',
    'https://thatware.co/services/',
    'https://thatware.co/advanced-seo-services/',
    # Add more URLs as needed
]

# Step 1: Function to extract links from each URL
# This function fetches the content of each page, parses the HTML, and retrieves all internal and external links.
def extract_links(url):
    try:
        # Send a GET request to the URL to retrieve the page content
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract all <a> tags, which contain links
        links = set()  # Using a set to avoid duplicate links
        for link in soup.find_all('a', href=True):
            href = link['href']

            # Ignore links that don't point to valid pages
            if href.startswith('http') or href.startswith('/'):
                links.add(href)

        # Return the set of extracted links
        return links

    except Exception as e:
        print(f"Error fetching or parsing the URL {url}: {e}")
        return set()

# Step 2: Create a graph to represent the website structure
# Here, we create a graph where each page (URL) is a node, and the links between pages are edges.
def build_graph(urls):
    G = nx.Graph()  # Create an empty graph

    # Add nodes (webpages) to the graph
    for url in urls:
        G.add_node(url)  # Add the URL as a node

        # Extract links from the current page
        links = extract_links(url)

        # Add edges (connections between pages) based on the extracted links
        for link in links:
            # Only add internal links to the graph (links within the same website)
            if link.startswith('https://thatware.co') or link.startswith('/'):
                full_link = link if link.startswith('http') else f'https://thatware.co{link}'
                G.add_edge(url, full_link)  # Add an edge between the two URLs

    return G

# Step 3: Visualize the graph (optional step, useful for understanding the structure)
# This step will print the nodes and edges of the graph.
def visualize_graph(G):
    print("Nodes (Pages):", G.nodes())
    print("Edges (Links):", G.edges())

# Step 4: Define the GNN Model
# Here we define the Graph Neural Network (GNN) model. GNN will analyze the graph to learn about the relationship between pages (nodes) and links (edges).
class GNNModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(GNNModel, self).__init__()
        # First graph convolution layer
        self.conv1 = nn.Linear(input_dim, hidden_dim)
        # Second graph convolution layer
        self.conv2 = nn.Linear(hidden_dim, output_dim)

    def forward(self, x, adj):
        # Apply the first layer (with ReLU activation)
        x = torch.relu(self.conv1(torch.mm(adj, x)))
        # Apply the second layer (with ReLU activation)
        x = torch.relu(self.conv2(torch.mm(adj, x)))
        return x

# Step 5: Prepare the input data for the GNN
# GNNs require a node feature matrix and an adjacency matrix (describes how nodes are connected).
def prepare_gnn_data(G):
    num_nodes = len(G.nodes)
    node_features = torch.eye(num_nodes)  # One-hot encoded feature matrix (each page is represented by a unique vector)
    adj_matrix = nx.adjacency_matrix(G).todense()  # Adjacency matrix that represents the connections between nodes
    adj_matrix = torch.tensor(adj_matrix, dtype=torch.float32)

    return node_features, adj_matrix

# Step 6: Train the GNN
# In this step, we define how the GNN will be trained to learn the link structure and predict the best link optimizations.
def train_gnn(G):
    # Prepare data
    node_features, adj_matrix = prepare_gnn_data(G)

    # Fixing the output dimensions: We change the output_dim to match the number of nodes in the graph
    model = GNNModel(input_dim=node_features.shape[1], hidden_dim=16, output_dim=node_features.shape[0])

    # Define a loss function and an optimizer
    criterion = nn.MSELoss()  # Mean Squared Error loss function
    optimizer = torch.optim.Adam(model.parameters(), lr=0.01)  # Adam optimizer

    # Training loop
    for epoch in range(100):  # 100 epochs
        optimizer.zero_grad()  # Zero the gradients
        output = model(node_features, adj_matrix)  # Forward pass
        loss = criterion(output, adj_matrix)  # Calculate loss
        loss.backward()  # Backpropagate the gradients
        optimizer.step()  # Update the model's weights

        if epoch % 10 == 0:  # Print the loss every 10 epochs
            print(f"Epoch {epoch}, Loss: {loss.item()}")

    return output, adj_matrix

# Step 7: Generate Recommendations to Improve Internal Links
# After training, we analyze the output and recommend new links or improvements in the internal linking structure.
def generate_recommendations(G, output, adj_matrix):
    num_nodes = len(G.nodes)
    adjacency_current = adj_matrix.detach().numpy()  # Convert the adjacency matrix to a numpy array for easier analysis
    adjacency_predicted = output.detach().numpy()  # Predicted links from the GNN model

    recommendations = []  # List to store link recommendations

    # Compare the current and predicted adjacency matrices
    for i in range(num_nodes):
        for j in range(num_nodes):
            # If the model predicts a link but the current structure doesn't have one, recommend adding it
            if adjacency_predicted[i][j] > 0.5 and adjacency_current[i][j] == 0:
                node_i = list(G.nodes())[i]  # Get the URL of node i
                node_j = list(G.nodes())[j]  # Get the URL of node j
                recommendations.append(f"Consider linking {node_i} to {node_j}")

    # Print recommendations
    if recommendations:
        print("Recommended Internal Links to Improve SEO:")
        for rec in recommendations:
            print(rec)
    else:
        print("No new internal linking recommendations found.")

# Step 8: Putting everything together
# Build the graph, visualize it, train the GNN, and generate recommendations for internal links.
G = build_graph(urls)  # Build the graph from the URLs
visualize_graph(G)     # Visualize the graph (optional)
output, adj_matrix = train_gnn(G)  # Train the GNN on the website structure
generate_recommendations(G, output, adj_matrix)  # Generate internal linking recommendations
````
