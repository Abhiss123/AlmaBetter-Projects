# What is K-Means Clustering?

At its core, **K-Means Clustering** is a machine learning technique used to group or classify data into different clusters (groups). Imagine you have a set of data points, and you want to divide them into separate groups that are similar to each other. This is what K-Means does: it finds patterns in the data and groups things that are alike into clusters.

**For example**, if you have data about your website users (**age, interests, or behavior**), K-Means can help divide them into clusters like **"young adults," "seniors," or "frequent buyers,"** so you can better understand and target them.


# How does K-Means Clustering work?

1. **Input Data:** You provide the algorithm with some data, such as user characteristics or webpage performance statistics.

2. **Choose Number of Clusters (K):** You decide how many groups (or clusters) you want to divide your data into. This is where the "K" in K-Means comes from. If you want to split users into 3 groups, K would be 3.

3.  **Assign to Clusters:** The algorithm looks at the data and tries to divide it into clusters by finding data points that are close to each other in some way (similar).

4.  **Adjust Clusters:** It keeps adjusting these clusters to make sure that each group is as tight and distinct from others as possible.

5.  **Final Clusters:** Once it has grouped the data points, it outputs the final clusters.

#  How to Decide on the Number of Clusters?

In K-Means Clustering, **clusters** represent groups of similar data points (in this case, similar web pages based on behavior). Deciding on the number of clusters depends on **what insights you want to gain and the structure of the data.**

Here’s how to approach it:

1. **Start Small:** It’s common to start with a small number of clusters (like 3 or 5) and see how well the data fits into these groups. You can adjust based on the results.

* **Example:** If you have 100 pages, starting with 3 clusters might group pages into:
*  **High traffic, high engagement** (popular pages).

*  **Low traffic, low engagement** (underperforming pages).

*  **Low traffic, high engagement** (niche but valuable pages).

2.  **Use the Elbow Method:** This is a common technique used to determine the right number of clusters. It involves running K-Means with **different numbers of clusters and plotting the sum of squared distances from each point to its assigned cluster center.** The **“elbow”** point on the graph helps you decide the optimal number of clusters.



# Step 1: Import the Necessary Libraries

      from sklearn.cluster import KMeans
      from sklearn.preprocessing import StandardScaler

**What this does:**

* **KMeans:** This is the algorithm we use to group (cluster) similar data points together. In our case, the data points are web pages, and we want to group them based on how users behave on those pages.

* **StandardScaler:** This is a tool that helps to "scale" the data. Scaling makes sure that all the numbers are treated fairly by the algorithm, especially when the numbers are very different in size **(like views vs. engagement time).**

#  Step 2: Select the Numeric Columns for Clustering

       features = user_behavior[['Views', 'Average engagement time per active user']]

**What this does:**

*  We are selecting the columns **'Views' and 'Average engagement time per active user'** from our data (which is about web pages).

*  These two columns represent the **numeric information** we want to use for grouping web pages. **'Views'** tells us how many people visit the page, and **'Average engagement time'** tells us how long they stay on the page.

**Example:** Imagine you have a list of 10 web pages, and each page has two numbers:

*  One number tells you how many people visited the page (Views).

*  The other number tells you how long each visitor spent on that page (Engagement Time).

# Step 3: Scale the Features

     scaler = StandardScaler()
     scaled_features = scaler.fit_transform(features)

**What this does:**

*  **Scaling** means adjusting the numbers so they are easier for the algorithm to work with.

*  **Why is this important?** If one column (like Views) has very large numbers (e.g., 40,000 views) and another column (like Engagement Time) has smaller numbers (e.g., 20 seconds), the algorithm might give too much importance to the larger numbers (Views). Scaling makes sure both columns are treated equally.

**Example:**

*  Imagine you’re comparing apples and oranges. If one box of apples has 100 fruits and the other box of oranges has only 5 fruits, it’s hard to compare them. Scaling makes them more comparable by adjusting the units so that the difference isn’t overwhelming.

*  After scaling, both columns will have a **mean of 0 and a standard deviation of 1,** meaning that both are on the same level for comparison.

# Step 4: Apply the K-Means Clustering Algorithm

     kmeans = KMeans(n_clusters=3, random_state=42)
     kmeans.fit(scaled_features)

**What this does:**

*  We are applying the **K-Means Clustering algorithm** to the scaled data.

*  **K-Means** tries to group the web pages into **3 clusters** (because we set **n_clusters=3**), based on their Views and Engagement Time.

**Example:**

*  Imagine you’re sorting books in a library. You want to create 3 groups based on how often each book is borrowed and how long people spend reading each book. The K-Means algorithm will figure out which books are similar to each other based on those two criteria (borrow frequency and reading time).

*   Similarly, here the computer is grouping similar web pages based on how many people visit them and how long people stay on them.

**Why random_state=42?**

*  This just ensures that the algorithm gives the same result every time you run it. If you leave this out, you might get slightly different results each time.

# Step 5: Add the Cluster Labels to the Original Data

       user_behavior['Cluster'] = kmeans.labels_

**What this does:**

*  After K-Means groups the **web pages into 3 clusters,** it assigns a label to each page, telling us which cluster it belongs to **(Cluster 0, Cluster 1, or Cluster 2).**

*  Here, we are adding a new column called **'Cluster'** to the original **user_behavior data,** which shows which cluster each page belongs to.

# Step 6: Display the Data with Cluster Labels

      print("\nUser Behavior Data with Clusters:")
      print(user_behavior.head())

**What this does:**

*  This simply shows the first few rows of the data, including the new Cluster labels. Now you can see which pages belong to which cluster.

**Example:**

* After the computer sorts the pages into 3 clusters, this will show something like:

  *   Page 1: Cluster 0 (low traffic, low engagement)
  *   Page 2: Cluster 1 (high traffic, high engagement)
  *   Page 3: Cluster 2 (low traffic, high engagement)

* **Why this is useful:** Once you know which pages belong to each cluster, you can take action based on the performance of those pages. For example, if Cluster 0 contains underperforming pages, you can focus on improving those pages.




```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Selecting the numeric columns for clustering
features = user_behavior[['Views', 'Average engagement time per active user']]

# Scaling the features to have mean = 0 and standard deviation = 1
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Applying K-Means clustering algorithm with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(scaled_features)

# Adding the cluster labels back to the original data
user_behavior['Cluster'] = kmeans.labels_

# Display the data with cluster labels
print("\nUser Behavior Data with Clusters:")
print(user_behavior.head())
```

```python
# Step 1: Import necessary libraries
import pandas as pd  # This library helps us work with CSV files and tables of data
from sklearn.cluster import KMeans  # This is the machine learning algorithm for clustering
from sklearn.preprocessing import StandardScaler  # This helps us scale our data so it's ready for clustering

# Step 2: Connect Google Drive to Google Colab
# This is necessary so we can access the CSV files stored in Google Drive
from google.colab import drive
drive.mount('/content/drive')  # This line connects your Google Drive to the Colab environment

# Step 3: Load the CSV files from Google Drive
# We will now load the necessary data from the file paths provided
user_behavior_data = pd.read_csv('/content/drive/MyDrive/User Behaviour Data/User Behavior Data.csv')  # This loads the user behavior data into a table
event_data = pd.read_csv('/content/drive/MyDrive/User Behaviour Data/Event data-export (1).csv')  # This loads event-related data into a table
demographics_data = pd.read_csv('/content/drive/MyDrive/User Behaviour Data/Demographics.csv')  # This loads demographic data into a table

# Step 4: Display the column names to ensure we have the correct data loaded
# This is just to check that we have loaded the data correctly and to see what columns we can use
print("User Behavior Data Columns:")
print(user_behavior_data.columns)  # This will show us the columns (features) available in the user behavior data

print("\nEvent Data Columns:")
print(event_data.columns)  # This will show us the columns in the event data

print("\nDemographics Data Columns:")
print(demographics_data.columns)  # This will show us the columns in the demographic data

# Step 5: Select the relevant columns from the User Behavior Data
# We are selecting the columns that we need for clustering
# 'Page path and screen class' is the URL of the webpage, 'Views' tells us how many times the page was viewed,
# and 'Average engagement time per active user' tells us how long users spend on the page
user_behavior = user_behavior_data[['Page path and screen class', 'Views', 'Average engagement time per active user']]

# Step 6: Display data types to make sure we have the right types of values
# We need to check if the data is in the correct format (numeric values) before applying clustering
print("\nData types before conversion:")
print(user_behavior.dtypes)

# Step 7: Convert any non-numeric values in 'Views' and 'Average engagement time per active user' to numeric
# If there are any non-numeric values (like strings), we convert them to numeric and handle any errors
user_behavior['Views'] = pd.to_numeric(user_behavior['Views'], errors='coerce')  # Convert 'Views' column to numeric
user_behavior['Average engagement time per active user'] = pd.to_numeric(user_behavior['Average engagement time per active user'], errors='coerce')  # Convert 'Average engagement time per active user' to numeric

# Step 8: Remove any rows with missing (NaN) values after the conversion
# This ensures we only have valid numeric data for clustering
user_behavior = user_behavior.dropna(subset=['Views', 'Average engagement time per active user'])

# Step 9: Display the data types again to confirm the conversion was successful
print("\nData types after conversion:")
print(user_behavior.dtypes)

# Step 10: Select the features for clustering
# Here, we choose the 'Views' and 'Average engagement time per active user' columns to group (cluster) similar pages together
features = user_behavior[['Views', 'Average engagement time per active user']]

# Step 11: Scale the features before applying K-Means
# Scaling is necessary because K-Means is sensitive to the scale of the data
# We want both 'Views' and 'Average engagement time per active user' to be treated equally
scaler = StandardScaler()  # Initialize the scaler
scaled_features = scaler.fit_transform(features)  # Scale the data

# Step 12: Apply K-Means clustering
# We are now applying K-Means to group the web pages into 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)  # We choose 3 clusters, you can experiment with more or fewer clusters
kmeans.fit(scaled_features)  # This fits the K-Means algorithm to the data and assigns clusters

# Step 13: Add the cluster labels back to the original data
# The K-Means algorithm will assign a 'Cluster' label (0, 1, or 2) to each webpage
user_behavior['Cluster'] = kmeans.labels_  # This creates a new column 'Cluster' with the cluster assignment

# Step 14: Display the first few rows of the data with cluster labels
# Now we can see which web pages have been grouped together
print("\nUser Behavior Data with Clusters:")
print(user_behavior.head())  # This shows the first 5 rows of the data along with the assigned clusters

# Step 15: Analyze the clusters by calculating the average for each cluster
# We group the data by the 'Cluster' and calculate the mean (average) for the numeric columns
cluster_summary = user_behavior.groupby('Cluster')[['Views', 'Average engagement time per active user']].mean()

# Step 16: Display the cluster summary
# This will show the average 'Views' and 'Average engagement time per active user' for each cluster
print("\nCluster Summary:")
print(cluster_summary)

# Step 17: Save the clustered data to a CSV file
# This saves the data with the cluster labels so you can analyze it or share it later
user_behavior.to_csv('/content/drive/MyDrive/User Behaviour Data/clustered_user_behavior_data.csv', index=False)  # Save the data with clusters to Google Drive

print("Clustered data saved successfully!")  # Confirmation that the data has been saved
```

# What is K-Means Clustering in Simple Terms?

Think of **K-Means Clustering** like sorting socks into different piles. Imagine you have a bunch of socks of different colors and patterns, and you want to organize them into groups. **K-Means does the same thing with data:** it groups similar **data** points (in your case, web pages) together.

In this case, we applied K-Means clustering to group pages from a website based on:

*  **Views:** How many times a page has been viewed.
*  **Average engagement time per active user:** How long users are spending on that page.
