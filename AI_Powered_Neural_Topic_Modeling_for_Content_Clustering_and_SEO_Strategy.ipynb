Neural Topic Modeling enhances this by using deep learning (neural networks) to identify complex, nuanced topics in the content, improving the accuracy of topic discovery.

````python
# Step 1: Fetch content from URLs
# This function will retrieve the text content from each webpage.
# Input: List of URLs (webpages)
# Output: List of raw text content from each URL
def fetch_webpage_content(urls):
    content_list = []  # List to store the content of each webpage
    for url in urls:
        try:
            # Send an HTTP request to get the webpage content
            response = requests.get(url)
            # Use BeautifulSoup to extract the text content from the HTML of the webpage
            soup = BeautifulSoup(response.content, 'html.parser')
            # Get all the visible text from the webpage
            text = soup.get_text(separator=' ')  # Separate each text block with a space
            content_list.append(text)  # Add the text content to the list
        except Exception as e:
            print(f"Failed to fetch {url}: {str(e)}")  # Print an error message if the webpage cannot be fetched
    return content_list
````

````python
# Step 2: Preprocess the text
# This function cleans the text by removing digits, stopwords, and unnecessary symbols.
# Input: List of raw text content
# Output: Cleaned and preprocessed text (list of text)
def preprocess_text(content_list):
    stop_words = set(stopwords.words('english'))  # Get default English stopwords from nltk

    # Additional custom stopwords and meaningless symbols to remove (e.g., punctuation)
    custom_stopwords = ['the', 'and', 'when', 'where', 'is', 'also', 'can', 'then', 'here', 'there', ':', ';',
                        '(', ')', '&', '*', '[', ']', '{', '}', '.', ',', '"', "'", '-', '_', '!', '?', '/']
    stop_words = stop_words.union(custom_stopwords)  # Combine default stopwords with custom stopwords

    preprocessed_content = []  # List to store the cleaned text content

    for content in content_list:
        content = content.lower()  # Convert all text to lowercase
        content = re.sub(r'\d+', '', content)  # Remove digits (numbers) from the text
        content = re.sub(r'[^\w\s]', '', content)  # Remove punctuation and symbols
        words = content.split()  # Split the text into individual words
        # Remove stopwords from the list of words
        filtered_words = [word for word in words if word not in stop_words]
        preprocessed_content.append(' '.join(filtered_words))  # Join the filtered words back into a string

    return preprocessed_content
````

The **`preprocess_text`** function is designed to clean text data. When we collect raw text from webpages, it contains unnecessary information like punctuation, numbers, and common words like "the," "and," "is."

````python
# Step 3: Vectorize the text (Bag of Words model)
# This step converts the cleaned text into a numerical format that the model can understand.
# Input: Preprocessed text
# Output: Word count matrix (Bag of Words model)
def vectorize_text(preprocessed_content):
    # Initialize a CountVectorizer object to convert text into a matrix of word counts
    vectorizer = CountVectorizer(max_df=0.9, min_df=2, stop_words='english')
    # Fit and transform the preprocessed content into a word count matrix
    word_matrix = vectorizer.fit_transform(preprocessed_content)
    return vectorizer, word_matrix
````

### What is the Purpose of This Function?

The **`vectorize_text`** function converts text into a **numerical format** that the machine learning model can understand. This step is necessary because machine learning algorithms cannot work with plain text—they need numbers. So, this function turns the cleaned text (from the previous step) into a **Bag of Words (BoW) model**, which is essentially a table that counts how many times each word appears in the text.

````python
# Step 7: Convert Similarity Matrix to Human-Readable Recommendations
# This function generates human-readable recommendations based on the similarity of topics between pages.
# Input: List of URLs, topic distributions, and number of topics
# Output: Human-readable recommendations (most similar pages for each URL)
def recommend_similar_pages(urls, page_topics, num_topics):
    # Initialize an empty matrix for storing topic distributions of each page
    topic_vectors = np.zeros((len(page_topics), num_topics))

    # Fill in the topic distribution for each page
    for i, topics in enumerate(page_topics):
        for topic_num, topic_value in topics:
            topic_vectors[i, topic_num] = topic_value

    # Compute the cosine similarity between the topic distributions of all pages
    similarity_matrix = cosine_similarity(topic_vectors)

    # Create a dictionary to store recommendations for each page
    recommendations = {}

    for i in range(len(urls)):
        similar_pages = []  # List to store similar pages and their similarity scores
        for j in range(len(urls)):
            if i != j:  # Avoid recommending the page to itself
                similarity_score = similarity_matrix[i][j]  # Get the similarity score between pages
                similar_pages.append((urls[j], similarity_score))

        # Sort the similar pages by similarity score (highest to lowest)
        similar_pages = sorted(similar_pages, key=lambda x: x[1], reverse=True)

        # Keep the top 3 most similar pages for each page
        recommendations[urls[i]] = similar_pages[:3]

    return recommendations

print(recommendations)
````

  - **`sorted(similar_pages, key=lambda x: x[1], reverse=True)`**: This sorts the similar pages by their similarity score (from highest to lowest). The **`x[1]`** refers to the similarity score in each tuple (URL, similarity score).
  - **`similar_pages[:3]`**: After sorting, this keeps only the top 3 most similar pages for each URL.
  - These top 3 similar pages are then stored in the **`recommendations`** dictionary for the current page (`urls[i]`).

- **Why this is important**:
  - We don’t want to overwhelm the user with too many recommendations, so we limit the suggestions to the top 3 most similar pages. This ensures the recommendations are useful and focused.

````python
# Install necessary libraries by running the following command in your terminal:
# pip install requests beautifulsoup4 nltk sklearn gensim

# Import required libraries for the project
import requests  # To fetch the content of web pages from the internet
from bs4 import BeautifulSoup  # To extract and parse HTML content from web pages
import re  # To use regular expressions for cleaning text
import nltk  # Natural Language Toolkit for working with text data
from nltk.corpus import stopwords  # To remove common stopwords from the text
from sklearn.feature_extraction.text import CountVectorizer  # To convert text into a numerical format
from sklearn.metrics.pairwise import cosine_similarity  # To calculate the similarity between different web pages
from gensim import corpora, models  # To create the neural topic model
import numpy as np  # To handle arrays and matrices

# Download English stopwords using the nltk library (common words like 'the', 'is', etc.)
nltk.download('stopwords')

# Step 1: Fetch content from URLs
# This function will retrieve the text content from each webpage.
# Input: List of URLs (webpages)
# Output: List of raw text content from each URL
def fetch_webpage_content(urls):
    content_list = []  # List to store the content of each webpage
    for url in urls:
        try:
            # Send an HTTP request to get the webpage content
            response = requests.get(url)
            # Use BeautifulSoup to extract the text content from the HTML of the webpage
            soup = BeautifulSoup(response.content, 'html.parser')
            # Get all the visible text from the webpage
            text = soup.get_text(separator=' ')  # Separate each text block with a space
            content_list.append(text)  # Add the text content to the list
        except Exception as e:
            print(f"Failed to fetch {url}: {str(e)}")  # Print an error message if the webpage cannot be fetched
    return content_list

# Step 2: Preprocess the text
# This function cleans the text by removing digits, stopwords, and unnecessary symbols.
# Input: List of raw text content
# Output: Cleaned and preprocessed text (list of text)
def preprocess_text(content_list):
    stop_words = set(stopwords.words('english'))  # Get default English stopwords from nltk

    # Additional custom stopwords and meaningless symbols to remove (e.g., punctuation)
    custom_stopwords = ['the', 'and', 'when', 'where', 'is', 'also', 'can', 'then', 'here', 'there', ':', ';',
                        '(', ')', '&', '*', '[', ']', '{', '}', '.', ',', '"', "'", '-', '_', '!', '?', '/']
    stop_words = stop_words.union(custom_stopwords)  # Combine default stopwords with custom stopwords

    preprocessed_content = []  # List to store the cleaned text content

    for content in content_list:
        content = content.lower()  # Convert all text to lowercase
        content = re.sub(r'\d+', '', content)  # Remove digits (numbers) from the text
        content = re.sub(r'[^\w\s]', '', content)  # Remove punctuation and symbols
        words = content.split()  # Split the text into individual words
        # Remove stopwords from the list of words
        filtered_words = [word for word in words if word not in stop_words]
        preprocessed_content.append(' '.join(filtered_words))  # Join the filtered words back into a string

    return preprocessed_content

# Step 3: Vectorize the text (Bag of Words model)
# This step converts the cleaned text into a numerical format that the model can understand.
# Input: Preprocessed text
# Output: Word count matrix (Bag of Words model)
def vectorize_text(preprocessed_content):
    # Initialize a CountVectorizer object to convert text into a matrix of word counts
    vectorizer = CountVectorizer(max_df=0.9, min_df=2, stop_words='english')
    # Fit and transform the preprocessed content into a word count matrix
    word_matrix = vectorizer.fit_transform(preprocessed_content)
    return vectorizer, word_matrix

# Step 4: Create the Topic Model
# This function creates a Latent Dirichlet Allocation (LDA) topic model.
# Input: Preprocessed text and number of topics
# Output: LDA model and corpus (bag of words representation of the documents)
def create_topic_model(preprocessed_content, num_topics=5):
    # Tokenize the preprocessed content into individual words
    tokenized_data = [text.split() for text in preprocessed_content]
    # Create a dictionary from the tokenized data
    dictionary = corpora.Dictionary(tokenized_data)
    # Convert the tokenized data into a Bag of Words corpus
    corpus = [dictionary.doc2bow(text) for text in tokenized_data]
    # Create the LDA topic model with the specified number of topics
    lda_model = models.LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=10)
    return lda_model, corpus

# Step 5: Assign topics to pages
# This function assigns a topic distribution to each page.
# Input: LDA model and corpus
# Output: List of topic distributions for each page
def assign_topics_to_pages(lda_model, corpus):
    page_topics = []  # List to store the topic distributions for each page
    for doc in corpus:
        # Get the topic distribution for each document
        page_topics.append(lda_model.get_document_topics(doc, minimum_probability=0))
    return page_topics

# Step 6: Display the top keywords for each topic
# This function shows the most important words for each topic, which can be used for SEO optimization.
# Input: LDA model
# Output: Prints the top keywords for each topic
def display_top_keywords(lda_model, num_keywords=10):
    # Print the top keywords for each topic
    for idx, topic in lda_model.print_topics(-1, num_keywords):
        print(f"Topic {idx} - Top {num_keywords} Keywords: {topic}")

# Step 7: Convert Similarity Matrix to Human-Readable Recommendations
# This function generates human-readable recommendations based on the similarity of topics between pages.
# Input: List of URLs, topic distributions, and number of topics
# Output: Human-readable recommendations (most similar pages for each URL)
def recommend_similar_pages(urls, page_topics, num_topics):
    # Initialize an empty matrix for storing topic distributions of each page
    topic_vectors = np.zeros((len(page_topics), num_topics))

    # Fill in the topic distribution for each page
    for i, topics in enumerate(page_topics):
        for topic_num, topic_value in topics:
            topic_vectors[i, topic_num] = topic_value

    # Compute the cosine similarity between the topic distributions of all pages
    similarity_matrix = cosine_similarity(topic_vectors)

    # Create a dictionary to store recommendations for each page
    recommendations = {}

    for i in range(len(urls)):
        similar_pages = []  # List to store similar pages and their similarity scores
        for j in range(len(urls)):
            if i != j:  # Avoid recommending the page to itself
                similarity_score = similarity_matrix[i][j]  # Get the similarity score between pages
                similar_pages.append((urls[j], similarity_score))

        # Sort the similar pages by similarity score (highest to lowest)
        similar_pages = sorted(similar_pages, key=lambda x: x[1], reverse=True)

        # Keep the top 3 most similar pages for each page
        recommendations[urls[i]] = similar_pages[:3]

    return recommendations

# Main Program Execution
# List of URLs to be analyzed
urls = [
    'https://thatware.co/',
    'https://thatware.co/services/',
    'https://thatware.co/advanced-seo-services/',
    'https://thatware.co/digital-marketing-services/',
    'https://thatware.co/business-intelligence-services/',
    'https://thatware.co/link-building-services/',
    'https://thatware.co/branding-press-release-services/',
    'https://thatware.co/conversion-rate-optimization/',
    'https://thatware.co/social-media-marketing/',
    'https://thatware.co/content-proofreading-services/',
    'https://thatware.co/website-design-services/',
    'https://thatware.co/web-development-services/',
    'https://thatware.co/app-development-services/',
    'https://thatware.co/website-maintenance-services/',
    'https://thatware.co/bug-testing-services/',
    'https://thatware.co/software-development-services/',
    'https://thatware.co/competitor-keyword-analysis/'
]

# Step 1: Fetch content from the URLs
webpage_content = fetch_webpage_content(urls)

# Step 2: Preprocess the text content
preprocessed_content = preprocess_text(webpage_content)

# Step 3: Create the topic model
lda_model, corpus = create_topic_model(preprocessed_content)

# Step 4: Assign topics to pages (Content Organization)
page_topics = assign_topics_to_pages(lda_model, corpus)

# Step 5: Display top keywords for each topic (SEO Optimization)
display_top_keywords(lda_model)

# Step 6: Generate human-readable recommendations for each page
recommendations = recommend_similar_pages(urls, page_topics, num_topics=5)

# Step 7: Display the recommendations in a human-readable format
print("\nHuman-Readable Recommendations:")
for page, similar_pages in recommendations.items():
    print(f"\nFor the page '{page}':")
    for similar_page, score in similar_pages:
        print(f" - Similar page: {similar_page} (Similarity score: {score:.2f})")

````
