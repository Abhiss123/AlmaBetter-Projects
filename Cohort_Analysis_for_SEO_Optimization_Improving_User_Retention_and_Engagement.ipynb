# **Part 1: Metadata Extraction Code**
### **What this code does**:
- It **extracts metadata** from a list of webpage URLs using Python libraries like `BeautifulSoup` and `requests`. Metadata includes:
   - **Page title** (name of the webpage).
   - **Meta description** (short description used by search engines).
   - **Meta keywords** (words that describe the page for SEO purposes).
   - **H1 and H2 tags** (headings on the page).
   - **Word count** (total words on the page).

### **Why is this important**:
- It collects **important SEO information** about each webpage, which can help you analyze how well the content is optimized for search engines.

````python
# Import necessary libraries
import pandas as pd  # Used for data storage and processing in tabular format
import requests  # Used to fetch webpage content from the given URLs
from bs4 import BeautifulSoup  # Used to parse and extract information from HTML content

# Step 1: Define the list of URLs to extract content and metadata
# Purpose: These URLs represent the web pages whose metadata (like title, description, keywords, headings) will be extracted
urls = [
    "https://webtool.co/advanced-seo-service/",
    "https://webtool.co/content-optimization-using-ai/",
    "https://webtool.co/cora/",
    "https://webtool.co/cosine-similarity/",
    "https://webtool.co/bagofwords/",
    "https://webtool.co/tf-idf-checker/",
    "https://webtool.co/lda/",
    "https://webtool.co/cooccurence/",
    "https://webtool.co/keydensity/",
    "https://webtool.co/proximity/",
    "https://webtool.co/semantic/",
    "https://webtool.co/n-gram/",
    "https://webtool.co/sentiment/",
    "https://webtool.co/contact-us/",
    "https://webtool.co/fitness-based-seo-service/",
    "https://webtool.co/attorney-based-seo-service/",
    "https://webtool.co/medical-seo-service/",
    "https://webtool.co/photography-seo-service/",
    "https://webtool.co/banking-seo-service/",
    "https://webtool.co/fashion-based-seo-service/",
    "https://webtool.co/real-estate-seo-service/",
    "https://webtool.co/adult-seo-service/",
    "https://webtool.co/cbd-seo-service/",
    "https://webtool.co/crypto-seo-service/",
    "https://webtool.co/ecommerce-seo-service/",
    "https://webtool.co/education-based-seo/",
    "https://webtool.co/gaming-seo/",
    "https://webtool.co/igaming-seo-service/",
    "https://webtool.co/cosmetics-seo-service/",
    "https://webtool.co/glass-wall-seo/"
]

# Step 2: Initialize an empty list to store the extracted metadata and content
# Purpose: This list will store dictionaries with extracted details for each webpage (like title, description, keywords, etc.)
content_data = []

# Step 3: Loop through each URL to fetch and extract metadata
# Purpose: For each webpage, we fetch the content, parse the HTML, and extract relevant metadata for SEO analysis.
for url in urls:
    try:
        # Fetch the webpage content using HTTP GET request
        # Purpose: Get the HTML content of the webpage
        response = requests.get(url, timeout=10)  # A timeout of 10 seconds ensures we don't wait indefinitely
        response.raise_for_status()  # Raise an error if the request fails (e.g., 404 or 500 error)

        # Parse the fetched HTML content using BeautifulSoup
        # Purpose: Make the webpage content structured and readable for extraction
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract the title of the webpage
        # Purpose: The title is essential for SEO as it summarizes the page content
        page_title = soup.title.string.strip() if soup.title else "No Title"

        # Extract the meta description tag content
        # Purpose: Meta descriptions provide a short summary of the page for search engines and users
        meta_description = soup.find('meta', attrs={'name': 'description'})
        meta_description = meta_description['content'].strip() if meta_description else "No Meta Description"

        # Extract the meta keywords tag content
        # Purpose: Although less relevant today, meta keywords indicate relevant keywords for SEO
        meta_keywords = soup.find('meta', attrs={'name': 'keywords'})
        meta_keywords = meta_keywords['content'].strip() if meta_keywords else "No Keywords"

        # Extract all H1 (main heading) tags from the page
        # Purpose: H1 tags define the main topic of the page and are critical for SEO
        h1_tags = "; ".join([h1.get_text(strip=True) for h1 in soup.find_all('h1')])

        # Extract all H2 (sub-heading) tags from the page
        # Purpose: H2 tags are used for subtopics and help organize the content for better readability
        h2_tags = "; ".join([h2.get_text(strip=True) for h2 in soup.find_all('h2')])

        # Calculate the total word count of the page
        # Purpose: Word count is important for SEO; search engines prefer comprehensive, detailed content
        main_content = soup.get_text(separator=" ", strip=True)  # Extract all visible text from the webpage
        word_count = len(main_content.split())  # Split the text into words and count them

        # Append all extracted data as a dictionary to the content_data list
        content_data.append({
            "URL": url,
            "Page title": page_title.lower().strip(),  # Convert title to lowercase and strip spaces for uniformity
            "Meta Description": meta_description,
            "Meta Keywords": meta_keywords,
            "H1 Tags": h1_tags,
            "H2 Tags": h2_tags,
            "Word Count": word_count
        })

    except Exception as e:
        # Handle any errors (e.g., network errors, invalid URLs) gracefully
        # Purpose: Ensure the script continues running even if one URL fails
        content_data.append({
            "URL": url,
            "Page title": "Error",
            "Meta Description": f"Error fetching content: {e}",
            "Meta Keywords": "Error",
            "H1 Tags": "Error",
            "H2 Tags": "Error",
            "Word Count": 0
        })

# Step 4: Convert the extracted metadata and content into a DataFrame
# Purpose: A DataFrame provides a structured and tabular format, making the data easy to analyze and export
content_df = pd.DataFrame(content_data)

# Step 5: Save the extracted data to a CSV file
# Purpose: Saving the data ensures it can be used later for merging with other datasets or performing further analysis
output_path = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Extracted_Content_Metadata.csv'
content_df.to_csv(output_path, index=False)  # Save the DataFrame as a CSV file without the index column

# Step 6: Print a sample of the extracted data for verification
# Purpose: Allow the user to verify that the data extraction worked as expected
print("Extracted Metadata and Content (Sample Rows):")
print(content_df.head())  # Print the first few rows of the extracted data
print(f"Extracted content saved to: {output_path}")  # Confirm the file save location
````

# **Part 2: Fuzzy Matching Code**
### **What this code does**:
- It matches webpage **titles** from two different datasets:
   1. Behavioral data (contains metrics like user activity, views, and engagement).
   2. Extracted metadata (contains SEO information such as page titles and descriptions).
- If the titles are slightly different, it uses **fuzzy matching** to find the closest match.

### **Why is this important**:
- Sometimes, the **titles of webpages** in different datasets might not match exactly due to spelling variations, special characters, or formatting differences.
- This step ensures that **relevant SEO metadata** (like titles, descriptions, and word count) is linked to the **behavioral data**.

---



````python
# Install necessary library for fuzzy matching
# 'fuzzywuzzy' is used for fuzzy string matching between similar texts
!pip install fuzzywuzzy

# Import required libraries
import pandas as pd  # Pandas for working with tabular data (dataframes)
import difflib  # difflib for fuzzy string matching

# -------------------------------------------------------------
# Step 1: Load the two datasets
# Purpose: We are combining behavioral data with metadata using page titles.
# -------------------------------------------------------------
# Define the file paths for the datasets
behavioral_data_path = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Merged_Group_1_and_Group_2_Dataset.csv'
metadata_path = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Extracted_Content_Metadata.csv'

# Load behavioral dataset: Contains user interactions and activity
behavioral_df = pd.read_csv(behavioral_data_path)

# Load metadata dataset: Contains extracted SEO metadata like titles, descriptions, etc.
metadata_df = pd.read_csv(metadata_path)

# -------------------------------------------------------------
# Step 2: Normalize 'Page title' column for consistency
# Purpose: Ensure both datasets use a consistent format for the 'page_title' column
# -------------------------------------------------------------
# Rename the 'Page title' column to 'page_title' for both datasets to avoid confusion
behavioral_df.rename(columns={'Page title': 'page_title'}, inplace=True)
metadata_df.rename(columns={'Page title': 'page_title'}, inplace=True)

# Standardize the page titles by converting them to lowercase and stripping extra spaces
# This ensures that minor differences in formatting (like capitalization) don't affect the matching process
behavioral_df['page_title'] = behavioral_df['page_title'].str.lower().str.strip()
metadata_df['page_title'] = metadata_df['page_title'].str.lower().str.strip()

# -------------------------------------------------------------
# Step 3: Define a fuzzy matching function using difflib
# Purpose: Match page titles between behavioral data and metadata even if they are slightly different
# -------------------------------------------------------------
def fuzzy_match_titles(behavioral_titles, metadata_titles, threshold=0.8):
    """
    This function matches titles between two lists using fuzzy string matching.
    It finds the closest matching metadata title for each behavioral title.

    Parameters:
    - behavioral_titles: List of unique page titles from behavioral data
    - metadata_titles: List of unique page titles from metadata
    - threshold: Similarity cutoff percentage (default is 80%)

    Returns:
    - A dictionary where keys are behavioral titles and values are matched metadata titles
    """
    matched_titles = {}  # Create an empty dictionary to store the matches

    for title in behavioral_titles:
        # Find the closest match for each title in metadata_titles
        closest_match = difflib.get_close_matches(title, metadata_titles, n=1, cutoff=threshold)
        matched_titles[title] = closest_match[0] if closest_match else None  # Store the match or None

    return matched_titles

# -------------------------------------------------------------
# Step 4: Perform fuzzy matching between the datasets
# Purpose: Identify similar titles between behavioral and metadata datasets
# -------------------------------------------------------------
print("Matching Page Titles using Fuzzy Matching...")
matched_titles_dict = fuzzy_match_titles(
    behavioral_df['page_title'].unique(),  # Unique titles from behavioral data
    metadata_df['page_title'].unique(),    # Unique titles from metadata
    threshold=0.8                          # Similarity threshold set to 80%
)

# -------------------------------------------------------------
# Step 5: Create a DataFrame from matched titles
# Purpose: Store the matched results for further processing
# -------------------------------------------------------------
matched_titles_df = pd.DataFrame(
    list(matched_titles_dict.items()),  # Convert dictionary to list of tuples
    columns=['behavioral_title', 'matched_metadata_title']  # Define column names
)

# -------------------------------------------------------------
# Step 6: Merge matched titles with metadata
# Purpose: Enrich metadata with matched titles to combine relevant information
# -------------------------------------------------------------
metadata_merged = pd.merge(
    matched_titles_df,  # Matched titles DataFrame
    metadata_df,        # Original metadata DataFrame
    how='left',         # Left join to retain all matched titles
    left_on='matched_metadata_title',  # Key: matched metadata title
    right_on='page_title'              # Key: original page title in metadata
)

# -------------------------------------------------------------
# Step 7: Merge enriched metadata with behavioral data
# Purpose: Combine user activity data with SEO metadata for analysis
# -------------------------------------------------------------
final_combined_df = pd.merge(
    behavioral_df,      # Behavioral data DataFrame
    metadata_merged,    # Enriched metadata DataFrame
    how='left',         # Left join to retain all behavioral data
    left_on='page_title',  # Key: page title in behavioral data
    right_on='behavioral_title'  # Key: matched title from fuzzy matching
)

# -------------------------------------------------------------
# Step 8: Handle missing metadata
# Purpose: Replace missing values with default placeholders to maintain data integrity
# -------------------------------------------------------------
final_combined_df['URL'] = final_combined_df['URL'].fillna('No URL')
final_combined_df['Meta Description'] = final_combined_df['Meta Description'].fillna('No Meta Description')
final_combined_df['Meta Keywords'] = final_combined_df['Meta Keywords'].fillna('No Keywords')
final_combined_df['H1 Tags'] = final_combined_df['H1 Tags'].fillna('No H1 Tags')
final_combined_df['H2 Tags'] = final_combined_df['H2 Tags'].fillna('No H2 Tags')
final_combined_df['Word Count'] = final_combined_df['Word Count'].fillna(0)

# -------------------------------------------------------------
# Step 9: Simplify column names and filter unmatched rows
# Purpose: Clean up and debug data for clarity
# -------------------------------------------------------------
# Combine the two page title columns into a single column for clarity
final_combined_df['final_page_title'] = final_combined_df['page_title_x'].fillna(final_combined_df['page_title_y'])

# Identify rows where metadata was not matched (still missing)
unmatched_rows = final_combined_df[final_combined_df['URL'] == 'No URL']

# -------------------------------------------------------------
# Step 10: Save the final combined dataset
# Purpose: Export the combined dataset for further cohort analysis
# -------------------------------------------------------------
output_path = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Final_Combined_Dataset_FuzzyMatch.csv'
final_combined_df.to_csv(output_path, index=False)

# Debugging Outputs: Verify the results
print("Final Combined Dataset Columns:", final_combined_df.columns)
print("Rows with Missing Metadata after Fuzzy Matching:")
print(unmatched_rows[['final_page_title']])

print("Sample Combined Dataset:")
print(final_combined_df.head())

print(f"Final combined dataset saved to: {output_path}")
print(f"Total rows with missing metadata: {len(unmatched_rows)}")
````

# **Part 3: Cohort Analysis Code**
### **What this code does**:
- It performs **Cohort Analysis**, which divides users into groups (called cohorts) based on when they first interacted with the website (like per week or month).
- It calculates:
   1. **User Retention**: How many users returned after their first visit.
   2. **Drop-off Analysis**: The week-over-week decrease in user activity.
   3. **Engagement Metrics**: Metrics like views, bounce rate, and average engagement time.

- This part also generates visualizations such as:
   - **Cohort Retention Heatmap**: Shows retention percentages for different weeks.
   - **Weekly Trends**: Tracks active users and engagement metrics.

### **Why is this important**:
- Cohort Analysis helps you identify **user behavior trends** over time.
   - Which user groups (cohorts) are staying engaged?
   - Where are users dropping off, and why?
- The heatmaps and charts provide a **clear visual representation** of user retention and engagement.
---


````python
# Import necessary libraries for data analysis and visualization
import pandas as pd  # Pandas is used for manipulating and analyzing tabular data
import matplotlib.pyplot as plt  # Matplotlib helps visualize data with graphs and charts
import seaborn as sns  # Seaborn is used to create visually appealing and informative visualizations

# Step 1: Load the final combined dataset from the second part
# Purpose: Load the dataset that was cleaned and enriched in the previous steps.
# This dataset contains user activity data along with enriched metadata like page titles, URLs, and engagement metrics.
final_combined_path = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Final_Combined_Dataset_FuzzyMatch.csv'
final_combined_df = pd.read_csv(final_combined_path)

# Step 2: Data Preprocessing for Cohort Analysis
# Purpose: Prepare the data for analyzing trends over time by creating 'Week' and 'Month' columns.

# Convert the 'Date' column into datetime format
# Why: Converting to datetime ensures we can use date-based functions for analysis.
final_combined_df['Date'] = pd.to_datetime(final_combined_df['Date'], format='%Y%m%d')

# Extract the week and month from the 'Date' column for grouping and segmentation
# 'Week' column groups user activity by week
final_combined_df['Week'] = final_combined_df['Date'].dt.to_period('W')  # Weekly period (Sunday to Saturday)
# 'Month' column groups user activity by month
final_combined_df['Month'] = final_combined_df['Date'].dt.to_period('M')  # Monthly period

# Convert 'Week' into a string format for compatibility in plotting and grouping
final_combined_df['Week_str'] = final_combined_df['Week'].astype(str)

# Step 3: Calculate Retention and Engagement Trends
# Purpose: Analyze weekly trends of active users and key engagement metrics such as sessions and bounce rates.

# Group data by 'Week_str' to analyze trends over time
# Aggregations:
# - 'Active users': Total active users in each week
# - 'Engaged sessions': Total number of sessions where users actively engaged
# - 'Bounce rate': Average bounce rate across all pages for each week
# - 'Average engagement time per session': Mean engagement time users spent per session
weekly_engagement = final_combined_df.groupby('Week_str').agg({
    'Active users': 'sum',
    'Engaged sessions': 'sum',
    'Bounce rate': 'mean',
    'Average engagement time per session': 'mean'
}).reset_index()

# Step 4: Visualize Retention Trends with Line Charts
# Purpose: Visualize how active users and engagement metrics change over time to identify patterns and trends.

# Plot 1: Weekly Active Users Over Time
plt.figure(figsize=(10, 6))  # Set the size of the plot
sns.lineplot(data=weekly_engagement, x='Week_str', y='Active users', marker='o')
plt.title('Weekly Active Users Over Time')  # Title of the plot
plt.xlabel('Week')  # X-axis label
plt.ylabel('Number of Active Users')  # Y-axis label
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout for better spacing
plt.show()  # Display the plot

# Plot 2: Weekly Engagement Metrics Over Time
# Why: Visualizing multiple engagement metrics (engaged sessions, bounce rate, avg. engagement time) on one graph
# helps us identify which weeks performed well or poorly.
plt.figure(figsize=(10, 6))
sns.lineplot(data=weekly_engagement, x='Week_str', y='Engaged sessions', marker='o', label='Engaged Sessions')
sns.lineplot(data=weekly_engagement, x='Week_str', y='Bounce rate', marker='o', label='Bounce Rate')
sns.lineplot(data=weekly_engagement, x='Week_str', y='Average engagement time per session', marker='o', label='Avg. Engagement Time')
plt.title('Engagement Metrics Over Time')  # Title of the plot
plt.xlabel('Week')  # X-axis label
plt.ylabel('Metric Values')  # Y-axis label
plt.legend()  # Show legend to identify metrics
plt.xticks(rotation=45)  # Rotate x-axis labels for better spacing
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()

# Step 5: Derive Actionable Insights Based on Engagement
# Purpose: Identify the top 5 pages with the highest average engagement time.
# Why: Pages with high engagement time indicate that users find the content valuable.

top_engaged_pages = final_combined_df.groupby('final_page_title').agg({
    'Average engagement time per session': 'mean',  # Average time users spent on each page
    'Views': 'sum',  # Total number of views for each page
    'Active users': 'sum'  # Total active users for each page
}).sort_values(by='Average engagement time per session', ascending=False).head(5)

# Print the top 5 pages for actionable insights
print("Top 5 Pages with Highest Engagement Time per Session:")
print(top_engaged_pages)

# Step 6: Analyze Traffic Insights for Top URLs
# Purpose: Analyze which URLs perform the best based on views, active users, and bounce rate.

# Filter rows where metadata (URLs) exist (excluding missing or default placeholders)
metadata_insights = final_combined_df[final_combined_df['URL'] != 'No URL'].groupby('URL').agg({
    'Views': 'sum',  # Total views for each URL
    'Active users': 'sum',  # Total active users for each URL
    'Bounce rate': 'mean'  # Average bounce rate for each URL
}).sort_values(by='Views', ascending=False).head(10)

# Print the top 10 performing URLs based on total views and active users
print("Top Performing URLs Based on Views and Active Users:")
print(metadata_insights)

# Step 7: Save Weekly Engagement Data for Further Use
# Purpose: Save the weekly aggregated engagement data to a CSV file for future analysis or reporting.

weekly_engagement_output = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Weekly_Engagement_Analysis.csv'
weekly_engagement.to_csv(weekly_engagement_output, index=False)  # Save to CSV without the index column
print(f"Weekly engagement data saved to: {weekly_engagement_output}")
````

# **Part 4: Conversion Rate Analysis Code**
### **What this code does**:
- It calculates the **Conversion Rate** for different cohorts (groups of users based on months).
- It identifies:
   1. The **high-performing cohort** (group with the highest conversion rate).
   2. The **low-performing cohort** (group with the lowest conversion rate).

- It generates a bar chart called **"Conversion Rate by Cohort"** to visually compare the performance of different cohorts.

### **Why is this important**:
- The conversion rate measures how many users **engage with the content** after visiting the website.
- By comparing cohorts, you can:
   - Find out what strategies worked for the **best-performing group**.
   - Identify issues that caused low performance and fix them.

- This helps you make **data-driven decisions** to improve user engagement and overall website performance.

````python
# Import necessary libraries for analysis and visualization
import pandas as pd  # Used to handle and analyze tabular data
import numpy as np  # Provides numerical computations and data operations
import matplotlib.pyplot as plt  # Used for creating visualizations like graphs and charts
import seaborn as sns  # Simplifies and beautifies data visualizations

# Step 1: Load the Final Combined Dataset
# Purpose: Load the dataset that contains user activity, session metrics, and engagement data.
# This dataset has already been cleaned and enriched in previous steps.
final_combined_path = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Final_Combined_Dataset_FuzzyMatch.csv'
final_combined_df = pd.read_csv(final_combined_path)

# Step 2: Preprocess Data for Cohort Analysis
# Purpose: Prepare the data to analyze user retention by grouping it into weeks and months.

# Convert the 'Date' column to a proper datetime format
# Why: Converting it to datetime enables us to perform date-related calculations.
final_combined_df['Date'] = pd.to_datetime(final_combined_df['Date'], format='%Y%m%d')

# Extract 'Acquisition Week' (week of user interaction) and 'Cohort Month' (month of user interaction)
# Why: These fields allow us to group users by their first interaction period.
final_combined_df['Acquisition Week'] = final_combined_df['Date'].dt.to_period('W').astype(str)  # Weekly cohorts
final_combined_df['Cohort Month'] = final_combined_df['Date'].dt.to_period('M')  # Monthly cohorts

# Step 3: Build the Retention Matrix
# Purpose: Create a matrix that shows user retention across weeks.

def build_retention_matrix(df):
    """
    Creates a user retention matrix that calculates the retention percentage over weeks.
    """
    # Add a 'Week of Activity' column (current week of user activity)
    # Why: This helps us track in which week the users were active after acquisition.
    df['Week of Activity'] = df['Date'].dt.to_period('W').astype(str)

    # Group by 'Acquisition Week' and 'Week of Activity' to count active users
    # Why: We want to calculate how many users remain active in subsequent weeks after acquisition.
    cohort_data = df.groupby(['Acquisition Week', 'Week of Activity']).agg({'Active users': 'sum'})

    # Pivot the table to create a matrix where rows = acquisition week, columns = week of activity
    retention_matrix = cohort_data.unstack(level='Week of Activity').fillna(0)

    # Normalize each cohort by dividing by the first week's size to get retention as a percentage
    # Why: This helps us compare user retention relative to the first week's activity.
    cohort_sizes = retention_matrix.iloc[:, 0]  # The first column represents cohort sizes
    retention_rate = retention_matrix.divide(cohort_sizes, axis=0) * 100  # Convert to percentage
    return retention_rate

# Generate the retention matrix
retention_rate = build_retention_matrix(final_combined_df)

# Step 4: Generate and Visualize Cohort Retention Heatmap
# Purpose: Visualize user retention over weeks using a heatmap.

plt.figure(figsize=(12, 8))  # Set the size of the plot
sns.heatmap(retention_rate, annot=True, fmt=".1f", cmap="Blues", cbar_kws={'label': 'Retention (%)'})
plt.title("Cohort Retention Analysis Heatmap")  # Title of the heatmap
plt.xlabel("Week of Activity")  # X-axis: weeks after acquisition
plt.ylabel("Acquisition Week")  # Y-axis: user acquisition weeks
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to avoid overlap
plt.savefig('/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Cohort_Retention_Heatmap.png')  # Save the plot
plt.show()  # Display the heatmap

# Step 5: Summarize Cohort Data
# Purpose: Summarize the total cohort size and Week 1 retention rate for each cohort.

cohort_summary = pd.DataFrame({
    'Total Users': retention_rate.iloc[:, 0],  # Total users in each cohort's first week
    'Week 1 Retention (%)': retention_rate.iloc[:, 1] if retention_rate.shape[1] > 1 else 0  # Week 1 retention
})

# Display the summary table for decision-makers
print("Cohort Summary Table:")
print(cohort_summary)

# Step 6: Drop-off Analysis (Week-over-Week Decline)
# Purpose: Calculate the week-over-week drop-off rates to identify when users are leaving.

drop_off_rates = retention_rate.diff(axis=1).fillna(0).iloc[:, 1:] * -1  # Negative changes as drop-off rates
print("Drop-off Analysis (Week-over-Week):")
print(drop_off_rates)

# Step 7: Analyze Conversion Metrics by Cohort Month
# Purpose: Group data by cohort month and analyze conversion metrics like bounce rates and engagement rates.

conversion_insights = final_combined_df.groupby('Cohort Month').agg({
    'Sessions': 'sum',  # Total user sessions in the cohort month
    'Engaged sessions': 'sum',  # Total engaged sessions
    'Bounce rate': 'mean',  # Average bounce rate
    'Engagement rate': 'mean',  # Average engagement rate
    'Views': 'sum'  # Total page views
}).reset_index()

# Calculate Conversion Rate as the percentage of engaged sessions
conversion_insights['Conversion Rate'] = (conversion_insights['Engaged sessions'] / conversion_insights['Sessions']) * 100

# Visualize Conversion Rate by Cohort Month
plt.figure(figsize=(12, 6))
sns.barplot(data=conversion_insights, x='Cohort Month', y='Conversion Rate', palette='pastel')
plt.title('Conversion Rate by Cohort')  # Chart title
plt.xlabel('Cohort Month')  # X-axis label
plt.ylabel('Conversion Rate (%)')  # Y-axis label
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()

# Step 8: Identify High-Performing and Low-Performing Cohorts
# Purpose: Highlight which cohorts are performing the best and worst based on conversion rates.

high_performance = conversion_insights.loc[conversion_insights['Conversion Rate'].idxmax()]  # Highest conversion rate
low_performance = conversion_insights.loc[conversion_insights['Conversion Rate'].idxmin()]  # Lowest conversion rate

# Print insights for actionable recommendations
print("\nHigh-Performance Cohort Insights:")
print(high_performance)

print("\nLow-Performance Cohort Insights:")
print(low_performance)

# Recommendations for improving website performance
print("\nActionable Recommendations:")
print(f"Focus on replicating strategies from the high-performing cohort ({high_performance['Cohort Month']}).")
print(f"Analyze and address issues in the low-performing cohort ({low_performance['Cohort Month']}).")

# Step 9: Save Outputs for Future Analysis
# Purpose: Save all generated insights to CSV files for reporting or further use.

retention_output = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Cohort_Retention_Data.csv'
summary_output = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Cohort_Summary_Data.csv'
dropoff_output = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/DropOff_Analysis.csv'
conversion_output = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Conversion_Insights.csv'

# Save all outputs to files
retention_rate.to_csv(retention_output)
cohort_summary.to_csv(summary_output)
drop_off_rates.to_csv(dropoff_output)
conversion_insights.to_csv(conversion_output)

print("\nOutputs Saved Successfully:")
print(f"- Retention data: {retention_output}")
print(f"- Cohort summary: {summary_output}")
print(f"- Drop-off analysis: {dropoff_output}")
print(f"- Conversion insights: {conversion_output}")
````

````python
# Import necessary libraries
import pandas as pd  # Used for data storage and processing in tabular format
import requests  # Used to fetch webpage content from the given URLs
from bs4 import BeautifulSoup  # Used to parse and extract information from HTML content

# Step 1: Define the list of URLs to extract content and metadata
# Purpose: These URLs represent the web pages whose metadata (like title, description, keywords, headings) will be extracted
urls = [
    "https://webtool.co/advanced-seo-service/",
    "https://webtool.co/content-optimization-using-ai/",
    "https://webtool.co/cora/",
    "https://webtool.co/cosine-similarity/",
    "https://webtool.co/bagofwords/",
    "https://webtool.co/tf-idf-checker/",
    "https://webtool.co/lda/",
    "https://webtool.co/cooccurence/",
    "https://webtool.co/keydensity/",
    "https://webtool.co/proximity/",
    "https://webtool.co/semantic/",
    "https://webtool.co/n-gram/",
    "https://webtool.co/sentiment/",
    "https://webtool.co/contact-us/",
    "https://webtool.co/fitness-based-seo-service/",
    "https://webtool.co/attorney-based-seo-service/",
    "https://webtool.co/medical-seo-service/",
    "https://webtool.co/photography-seo-service/",
    "https://webtool.co/banking-seo-service/",
    "https://webtool.co/fashion-based-seo-service/",
    "https://webtool.co/real-estate-seo-service/",
    "https://webtool.co/adult-seo-service/",
    "https://webtool.co/cbd-seo-service/",
    "https://webtool.co/crypto-seo-service/",
    "https://webtool.co/ecommerce-seo-service/",
    "https://webtool.co/education-based-seo/",
    "https://webtool.co/gaming-seo/",
    "https://webtool.co/igaming-seo-service/",
    "https://webtool.co/cosmetics-seo-service/",
    "https://webtool.co/glass-wall-seo/"
]

# Step 2: Initialize an empty list to store the extracted metadata and content
# Purpose: This list will store dictionaries with extracted details for each webpage (like title, description, keywords, etc.)
content_data = []

# Step 3: Loop through each URL to fetch and extract metadata
# Purpose: For each webpage, we fetch the content, parse the HTML, and extract relevant metadata for SEO analysis.
for url in urls:
    try:
        # Fetch the webpage content using HTTP GET request
        # Purpose: Get the HTML content of the webpage
        response = requests.get(url, timeout=10)  # A timeout of 10 seconds ensures we don't wait indefinitely
        response.raise_for_status()  # Raise an error if the request fails (e.g., 404 or 500 error)

        # Parse the fetched HTML content using BeautifulSoup
        # Purpose: Make the webpage content structured and readable for extraction
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract the title of the webpage
        # Purpose: The title is essential for SEO as it summarizes the page content
        page_title = soup.title.string.strip() if soup.title else "No Title"

        # Extract the meta description tag content
        # Purpose: Meta descriptions provide a short summary of the page for search engines and users
        meta_description = soup.find('meta', attrs={'name': 'description'})
        meta_description = meta_description['content'].strip() if meta_description else "No Meta Description"

        # Extract the meta keywords tag content
        # Purpose: Although less relevant today, meta keywords indicate relevant keywords for SEO
        meta_keywords = soup.find('meta', attrs={'name': 'keywords'})
        meta_keywords = meta_keywords['content'].strip() if meta_keywords else "No Keywords"

        # Extract all H1 (main heading) tags from the page
        # Purpose: H1 tags define the main topic of the page and are critical for SEO
        h1_tags = "; ".join([h1.get_text(strip=True) for h1 in soup.find_all('h1')])

        # Extract all H2 (sub-heading) tags from the page
        # Purpose: H2 tags are used for subtopics and help organize the content for better readability
        h2_tags = "; ".join([h2.get_text(strip=True) for h2 in soup.find_all('h2')])

        # Calculate the total word count of the page
        # Purpose: Word count is important for SEO; search engines prefer comprehensive, detailed content
        main_content = soup.get_text(separator=" ", strip=True)  # Extract all visible text from the webpage
        word_count = len(main_content.split())  # Split the text into words and count them

        # Append all extracted data as a dictionary to the content_data list
        content_data.append({
            "URL": url,
            "Page title": page_title.lower().strip(),  # Convert title to lowercase and strip spaces for uniformity
            "Meta Description": meta_description,
            "Meta Keywords": meta_keywords,
            "H1 Tags": h1_tags,
            "H2 Tags": h2_tags,
            "Word Count": word_count
        })

    except Exception as e:
        # Handle any errors (e.g., network errors, invalid URLs) gracefully
        # Purpose: Ensure the script continues running even if one URL fails
        content_data.append({
            "URL": url,
            "Page title": "Error",
            "Meta Description": f"Error fetching content: {e}",
            "Meta Keywords": "Error",
            "H1 Tags": "Error",
            "H2 Tags": "Error",
            "Word Count": 0
        })

# Step 4: Convert the extracted metadata and content into a DataFrame
# Purpose: A DataFrame provides a structured and tabular format, making the data easy to analyze and export
content_df = pd.DataFrame(content_data)

# Step 5: Save the extracted data to a CSV file
# Purpose: Saving the data ensures it can be used later for merging with other datasets or performing further analysis
output_path = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Extracted_Content_Metadata.csv'
content_df.to_csv(output_path, index=False)  # Save the DataFrame as a CSV file without the index column

# Step 6: Print a sample of the extracted data for verification
# Purpose: Allow the user to verify that the data extraction worked as expected
print("Extracted Metadata and Content (Sample Rows):")
print(content_df.head())  # Print the first few rows of the extracted data
print(f"Extracted content saved to: {output_path}")  # Confirm the file save location


# Install necessary library for fuzzy matching
# 'fuzzywuzzy' is used for fuzzy string matching between similar texts
!pip install fuzzywuzzy

# Import required libraries
import pandas as pd  # Pandas for working with tabular data (dataframes)
import difflib  # difflib for fuzzy string matching

# -------------------------------------------------------------
# Step 1: Load the two datasets
# Purpose: We are combining behavioral data with metadata using page titles.
# -------------------------------------------------------------
# Define the file paths for the datasets
behavioral_data_path = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Merged_Group_1_and_Group_2_Dataset.csv'
metadata_path = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Extracted_Content_Metadata.csv'

# Load behavioral dataset: Contains user interactions and activity
behavioral_df = pd.read_csv(behavioral_data_path)

# Load metadata dataset: Contains extracted SEO metadata like titles, descriptions, etc.
metadata_df = pd.read_csv(metadata_path)

# -------------------------------------------------------------
# Step 2: Normalize 'Page title' column for consistency
# Purpose: Ensure both datasets use a consistent format for the 'page_title' column
# -------------------------------------------------------------
# Rename the 'Page title' column to 'page_title' for both datasets to avoid confusion
behavioral_df.rename(columns={'Page title': 'page_title'}, inplace=True)
metadata_df.rename(columns={'Page title': 'page_title'}, inplace=True)

# Standardize the page titles by converting them to lowercase and stripping extra spaces
# This ensures that minor differences in formatting (like capitalization) don't affect the matching process
behavioral_df['page_title'] = behavioral_df['page_title'].str.lower().str.strip()
metadata_df['page_title'] = metadata_df['page_title'].str.lower().str.strip()

# -------------------------------------------------------------
# Step 3: Define a fuzzy matching function using difflib
# Purpose: Match page titles between behavioral data and metadata even if they are slightly different
# -------------------------------------------------------------
def fuzzy_match_titles(behavioral_titles, metadata_titles, threshold=0.8):
    """
    This function matches titles between two lists using fuzzy string matching.
    It finds the closest matching metadata title for each behavioral title.

    Parameters:
    - behavioral_titles: List of unique page titles from behavioral data
    - metadata_titles: List of unique page titles from metadata
    - threshold: Similarity cutoff percentage (default is 80%)

    Returns:
    - A dictionary where keys are behavioral titles and values are matched metadata titles
    """
    matched_titles = {}  # Create an empty dictionary to store the matches

    for title in behavioral_titles:
        # Find the closest match for each title in metadata_titles
        closest_match = difflib.get_close_matches(title, metadata_titles, n=1, cutoff=threshold)
        matched_titles[title] = closest_match[0] if closest_match else None  # Store the match or None

    return matched_titles

# -------------------------------------------------------------
# Step 4: Perform fuzzy matching between the datasets
# Purpose: Identify similar titles between behavioral and metadata datasets
# -------------------------------------------------------------
print("Matching Page Titles using Fuzzy Matching...")
matched_titles_dict = fuzzy_match_titles(
    behavioral_df['page_title'].unique(),  # Unique titles from behavioral data
    metadata_df['page_title'].unique(),    # Unique titles from metadata
    threshold=0.8                          # Similarity threshold set to 80%
)

# -------------------------------------------------------------
# Step 5: Create a DataFrame from matched titles
# Purpose: Store the matched results for further processing
# -------------------------------------------------------------
matched_titles_df = pd.DataFrame(
    list(matched_titles_dict.items()),  # Convert dictionary to list of tuples
    columns=['behavioral_title', 'matched_metadata_title']  # Define column names
)

# -------------------------------------------------------------
# Step 6: Merge matched titles with metadata
# Purpose: Enrich metadata with matched titles to combine relevant information
# -------------------------------------------------------------
metadata_merged = pd.merge(
    matched_titles_df,  # Matched titles DataFrame
    metadata_df,        # Original metadata DataFrame
    how='left',         # Left join to retain all matched titles
    left_on='matched_metadata_title',  # Key: matched metadata title
    right_on='page_title'              # Key: original page title in metadata
)

# -------------------------------------------------------------
# Step 7: Merge enriched metadata with behavioral data
# Purpose: Combine user activity data with SEO metadata for analysis
# -------------------------------------------------------------
final_combined_df = pd.merge(
    behavioral_df,      # Behavioral data DataFrame
    metadata_merged,    # Enriched metadata DataFrame
    how='left',         # Left join to retain all behavioral data
    left_on='page_title',  # Key: page title in behavioral data
    right_on='behavioral_title'  # Key: matched title from fuzzy matching
)

# -------------------------------------------------------------
# Step 8: Handle missing metadata
# Purpose: Replace missing values with default placeholders to maintain data integrity
# -------------------------------------------------------------
final_combined_df['URL'] = final_combined_df['URL'].fillna('No URL')
final_combined_df['Meta Description'] = final_combined_df['Meta Description'].fillna('No Meta Description')
final_combined_df['Meta Keywords'] = final_combined_df['Meta Keywords'].fillna('No Keywords')
final_combined_df['H1 Tags'] = final_combined_df['H1 Tags'].fillna('No H1 Tags')
final_combined_df['H2 Tags'] = final_combined_df['H2 Tags'].fillna('No H2 Tags')
final_combined_df['Word Count'] = final_combined_df['Word Count'].fillna(0)

# -------------------------------------------------------------
# Step 9: Simplify column names and filter unmatched rows
# Purpose: Clean up and debug data for clarity
# -------------------------------------------------------------
# Combine the two page title columns into a single column for clarity
final_combined_df['final_page_title'] = final_combined_df['page_title_x'].fillna(final_combined_df['page_title_y'])

# Identify rows where metadata was not matched (still missing)
unmatched_rows = final_combined_df[final_combined_df['URL'] == 'No URL']

# -------------------------------------------------------------
# Step 10: Save the final combined dataset
# Purpose: Export the combined dataset for further cohort analysis
# -------------------------------------------------------------
output_path = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Final_Combined_Dataset_FuzzyMatch.csv'
final_combined_df.to_csv(output_path, index=False)

# Debugging Outputs: Verify the results
print("Final Combined Dataset Columns:", final_combined_df.columns)
print("Rows with Missing Metadata after Fuzzy Matching:")
print(unmatched_rows[['final_page_title']])

print("Sample Combined Dataset:")
print(final_combined_df.head())

print(f"Final combined dataset saved to: {output_path}")
print(f"Total rows with missing metadata: {len(unmatched_rows)}")

# Import necessary libraries for data analysis and visualization
import pandas as pd  # Pandas is used for manipulating and analyzing tabular data
import matplotlib.pyplot as plt  # Matplotlib helps visualize data with graphs and charts
import seaborn as sns  # Seaborn is used to create visually appealing and informative visualizations

# Step 1: Load the final combined dataset from the second part
# Purpose: Load the dataset that was cleaned and enriched in the previous steps.
# This dataset contains user activity data along with enriched metadata like page titles, URLs, and engagement metrics.
final_combined_path = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Final_Combined_Dataset_FuzzyMatch.csv'
final_combined_df = pd.read_csv(final_combined_path)

# Step 2: Data Preprocessing for Cohort Analysis
# Purpose: Prepare the data for analyzing trends over time by creating 'Week' and 'Month' columns.

# Convert the 'Date' column into datetime format
# Why: Converting to datetime ensures we can use date-based functions for analysis.
final_combined_df['Date'] = pd.to_datetime(final_combined_df['Date'], format='%Y%m%d')

# Extract the week and month from the 'Date' column for grouping and segmentation
# 'Week' column groups user activity by week
final_combined_df['Week'] = final_combined_df['Date'].dt.to_period('W')  # Weekly period (Sunday to Saturday)
# 'Month' column groups user activity by month
final_combined_df['Month'] = final_combined_df['Date'].dt.to_period('M')  # Monthly period

# Convert 'Week' into a string format for compatibility in plotting and grouping
final_combined_df['Week_str'] = final_combined_df['Week'].astype(str)

# Step 3: Calculate Retention and Engagement Trends
# Purpose: Analyze weekly trends of active users and key engagement metrics such as sessions and bounce rates.

# Group data by 'Week_str' to analyze trends over time
# Aggregations:
# - 'Active users': Total active users in each week
# - 'Engaged sessions': Total number of sessions where users actively engaged
# - 'Bounce rate': Average bounce rate across all pages for each week
# - 'Average engagement time per session': Mean engagement time users spent per session
weekly_engagement = final_combined_df.groupby('Week_str').agg({
    'Active users': 'sum',
    'Engaged sessions': 'sum',
    'Bounce rate': 'mean',
    'Average engagement time per session': 'mean'
}).reset_index()

# Step 4: Visualize Retention Trends with Line Charts
# Purpose: Visualize how active users and engagement metrics change over time to identify patterns and trends.

# Plot 1: Weekly Active Users Over Time
plt.figure(figsize=(10, 6))  # Set the size of the plot
sns.lineplot(data=weekly_engagement, x='Week_str', y='Active users', marker='o')
plt.title('Weekly Active Users Over Time')  # Title of the plot
plt.xlabel('Week')  # X-axis label
plt.ylabel('Number of Active Users')  # Y-axis label
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout for better spacing
plt.show()  # Display the plot

# Plot 2: Weekly Engagement Metrics Over Time
# Why: Visualizing multiple engagement metrics (engaged sessions, bounce rate, avg. engagement time) on one graph
# helps us identify which weeks performed well or poorly.
plt.figure(figsize=(10, 6))
sns.lineplot(data=weekly_engagement, x='Week_str', y='Engaged sessions', marker='o', label='Engaged Sessions')
sns.lineplot(data=weekly_engagement, x='Week_str', y='Bounce rate', marker='o', label='Bounce Rate')
sns.lineplot(data=weekly_engagement, x='Week_str', y='Average engagement time per session', marker='o', label='Avg. Engagement Time')
plt.title('Engagement Metrics Over Time')  # Title of the plot
plt.xlabel('Week')  # X-axis label
plt.ylabel('Metric Values')  # Y-axis label
plt.legend()  # Show legend to identify metrics
plt.xticks(rotation=45)  # Rotate x-axis labels for better spacing
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()

# Step 5: Derive Actionable Insights Based on Engagement
# Purpose: Identify the top 5 pages with the highest average engagement time.
# Why: Pages with high engagement time indicate that users find the content valuable.

top_engaged_pages = final_combined_df.groupby('final_page_title').agg({
    'Average engagement time per session': 'mean',  # Average time users spent on each page
    'Views': 'sum',  # Total number of views for each page
    'Active users': 'sum'  # Total active users for each page
}).sort_values(by='Average engagement time per session', ascending=False).head(5)

# Print the top 5 pages for actionable insights
print("Top 5 Pages with Highest Engagement Time per Session:")
print(top_engaged_pages)

# Step 6: Analyze Traffic Insights for Top URLs
# Purpose: Analyze which URLs perform the best based on views, active users, and bounce rate.

# Filter rows where metadata (URLs) exist (excluding missing or default placeholders)
metadata_insights = final_combined_df[final_combined_df['URL'] != 'No URL'].groupby('URL').agg({
    'Views': 'sum',  # Total views for each URL
    'Active users': 'sum',  # Total active users for each URL
    'Bounce rate': 'mean'  # Average bounce rate for each URL
}).sort_values(by='Views', ascending=False).head(10)

# Print the top 10 performing URLs based on total views and active users
print("Top Performing URLs Based on Views and Active Users:")
print(metadata_insights)

# Step 7: Save Weekly Engagement Data for Further Use
# Purpose: Save the weekly aggregated engagement data to a CSV file for future analysis or reporting.

weekly_engagement_output = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Weekly_Engagement_Analysis.csv'
weekly_engagement.to_csv(weekly_engagement_output, index=False)  # Save to CSV without the index column
print(f"Weekly engagement data saved to: {weekly_engagement_output}")



# Import necessary libraries for analysis and visualization
import pandas as pd  # Used to handle and analyze tabular data
import numpy as np  # Provides numerical computations and data operations
import matplotlib.pyplot as plt  # Used for creating visualizations like graphs and charts
import seaborn as sns  # Simplifies and beautifies data visualizations

# Step 1: Load the Final Combined Dataset
# Purpose: Load the dataset that contains user activity, session metrics, and engagement data.
# This dataset has already been cleaned and enriched in previous steps.
final_combined_path = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Final_Combined_Dataset_FuzzyMatch.csv'
final_combined_df = pd.read_csv(final_combined_path)

# Step 2: Preprocess Data for Cohort Analysis
# Purpose: Prepare the data to analyze user retention by grouping it into weeks and months.

# Convert the 'Date' column to a proper datetime format
# Why: Converting it to datetime enables us to perform date-related calculations.
final_combined_df['Date'] = pd.to_datetime(final_combined_df['Date'], format='%Y%m%d')

# Extract 'Acquisition Week' (week of user interaction) and 'Cohort Month' (month of user interaction)
# Why: These fields allow us to group users by their first interaction period.
final_combined_df['Acquisition Week'] = final_combined_df['Date'].dt.to_period('W').astype(str)  # Weekly cohorts
final_combined_df['Cohort Month'] = final_combined_df['Date'].dt.to_period('M')  # Monthly cohorts

# Step 3: Build the Retention Matrix
# Purpose: Create a matrix that shows user retention across weeks.

def build_retention_matrix(df):
    """
    Creates a user retention matrix that calculates the retention percentage over weeks.
    """
    # Add a 'Week of Activity' column (current week of user activity)
    # Why: This helps us track in which week the users were active after acquisition.
    df['Week of Activity'] = df['Date'].dt.to_period('W').astype(str)

    # Group by 'Acquisition Week' and 'Week of Activity' to count active users
    # Why: We want to calculate how many users remain active in subsequent weeks after acquisition.
    cohort_data = df.groupby(['Acquisition Week', 'Week of Activity']).agg({'Active users': 'sum'})

    # Pivot the table to create a matrix where rows = acquisition week, columns = week of activity
    retention_matrix = cohort_data.unstack(level='Week of Activity').fillna(0)

    # Normalize each cohort by dividing by the first week's size to get retention as a percentage
    # Why: This helps us compare user retention relative to the first week's activity.
    cohort_sizes = retention_matrix.iloc[:, 0]  # The first column represents cohort sizes
    retention_rate = retention_matrix.divide(cohort_sizes, axis=0) * 100  # Convert to percentage
    return retention_rate

# Generate the retention matrix
retention_rate = build_retention_matrix(final_combined_df)

# Step 4: Generate and Visualize Cohort Retention Heatmap
# Purpose: Visualize user retention over weeks using a heatmap.

plt.figure(figsize=(12, 8))  # Set the size of the plot
sns.heatmap(retention_rate, annot=True, fmt=".1f", cmap="Blues", cbar_kws={'label': 'Retention (%)'})
plt.title("Cohort Retention Analysis Heatmap")  # Title of the heatmap
plt.xlabel("Week of Activity")  # X-axis: weeks after acquisition
plt.ylabel("Acquisition Week")  # Y-axis: user acquisition weeks
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to avoid overlap
plt.savefig('/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Cohort_Retention_Heatmap.png')  # Save the plot
plt.show()  # Display the heatmap

# Step 5: Summarize Cohort Data
# Purpose: Summarize the total cohort size and Week 1 retention rate for each cohort.

cohort_summary = pd.DataFrame({
    'Total Users': retention_rate.iloc[:, 0],  # Total users in each cohort's first week
    'Week 1 Retention (%)': retention_rate.iloc[:, 1] if retention_rate.shape[1] > 1 else 0  # Week 1 retention
})

# Display the summary table for decision-makers
print("Cohort Summary Table:")
print(cohort_summary)

# Step 6: Drop-off Analysis (Week-over-Week Decline)
# Purpose: Calculate the week-over-week drop-off rates to identify when users are leaving.

drop_off_rates = retention_rate.diff(axis=1).fillna(0).iloc[:, 1:] * -1  # Negative changes as drop-off rates
print("Drop-off Analysis (Week-over-Week):")
print(drop_off_rates)

# Step 7: Analyze Conversion Metrics by Cohort Month
# Purpose: Group data by cohort month and analyze conversion metrics like bounce rates and engagement rates.

conversion_insights = final_combined_df.groupby('Cohort Month').agg({
    'Sessions': 'sum',  # Total user sessions in the cohort month
    'Engaged sessions': 'sum',  # Total engaged sessions
    'Bounce rate': 'mean',  # Average bounce rate
    'Engagement rate': 'mean',  # Average engagement rate
    'Views': 'sum'  # Total page views
}).reset_index()

# Calculate Conversion Rate as the percentage of engaged sessions
conversion_insights['Conversion Rate'] = (conversion_insights['Engaged sessions'] / conversion_insights['Sessions']) * 100

# Visualize Conversion Rate by Cohort Month
plt.figure(figsize=(12, 6))
sns.barplot(data=conversion_insights, x='Cohort Month', y='Conversion Rate', palette='pastel')
plt.title('Conversion Rate by Cohort')  # Chart title
plt.xlabel('Cohort Month')  # X-axis label
plt.ylabel('Conversion Rate (%)')  # Y-axis label
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.show()

# Step 8: Identify High-Performing and Low-Performing Cohorts
# Purpose: Highlight which cohorts are performing the best and worst based on conversion rates.

high_performance = conversion_insights.loc[conversion_insights['Conversion Rate'].idxmax()]  # Highest conversion rate
low_performance = conversion_insights.loc[conversion_insights['Conversion Rate'].idxmin()]  # Lowest conversion rate

# Print insights for actionable recommendations
print("\nHigh-Performance Cohort Insights:")
print(high_performance)

print("\nLow-Performance Cohort Insights:")
print(low_performance)

# Recommendations for improving website performance
print("\nActionable Recommendations:")
print(f"Focus on replicating strategies from the high-performing cohort ({high_performance['Cohort Month']}).")
print(f"Analyze and address issues in the low-performing cohort ({low_performance['Cohort Month']}).")

# Step 9: Save Outputs for Future Analysis
# Purpose: Save all generated insights to CSV files for reporting or further use.

retention_output = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Cohort_Retention_Data.csv'
summary_output = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Cohort_Summary_Data.csv'
dropoff_output = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/DropOff_Analysis.csv'
conversion_output = '/content/drive/MyDrive/ Dataset For Cohort Analysis for SEO Model/Conversion_Insights.csv'

# Save all outputs to files
retention_rate.to_csv(retention_output)
cohort_summary.to_csv(summary_output)
drop_off_rates.to_csv(dropoff_output)
conversion_insights.to_csv(conversion_output)

print("\nOutputs Saved Successfully:")
print(f"- Retention data: {retention_output}")
print(f"- Cohort summary: {summary_output}")
print(f"- Drop-off analysis: {dropoff_output}")
print(f"- Conversion insights: {conversion_output}")

````

# **1. What is a Cohort?**
A **cohort** refers to a group of users who started their activity during the same period (like a specific month or week). Cohort Analysis helps analyze their behavior over time and compares user groups.
