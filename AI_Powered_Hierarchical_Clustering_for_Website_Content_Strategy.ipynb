### **Hierarchical Clustering for Content Strategy**

**1. What is Hierarchical Clustering for Content Strategy?**  
Hierarchical clustering is a technique used to group similar items into a tree-like structure (called a hierarchy). For content strategy, this means organizing website content into meaningful groups based on their similarity. The goal is to improve the website's structure, make content easier to navigate, and enhance user experience and SEO (search engine optimization).

**2. What are its Use Cases and Real-Life Implementations?**  
Hierarchical clustering is used in various ways, such as:  
- **Grouping similar pages:** For example, blog posts about similar topics can be grouped together.  
- **Improving internal linking:** By linking related pages, users can navigate better and stay longer on the site.  
- **Optimizing content strategy:** It helps identify gaps in content and suggests areas for improvement.  
- **E-commerce:** Categorizing products into meaningful clusters (e.g., shoes, clothing, accessories) for better navigation.  
- **News websites:** Organizing articles by topics like sports, politics, and entertainment.  

**3. How Does it Work in a Website Context?**  
For a website, hierarchical clustering works by analyzing the text content of your pages to find similarities. It could involve analyzing keywords, topics, or themes on each page. The model then groups pages that cover similar topics into clusters. For example, if your site has 500 pages, hierarchical clustering can identify clusters like “Technology,” “Health,” and “Finance,” with subcategories within each.

**4. What Kind of Data Does it Need?**  
To perform hierarchical clustering, you need:  
- **Page content data:** The text content from each page of your website (e.g., blog articles, product descriptions).  
- **Metadata:** Titles, keywords, and categories of the pages.  
- **Optional: URLs:** If extracting text automatically, the URLs of the pages can be used.  

The data can either be in raw text form or organized in a **CSV file**, where each row represents a webpage and includes its URL, title, and text content.

**5. Does it Require Webpage URLs or CSV Data?**  
- **If you have URLs:** You can use Python code to fetch (or "scrape") the content from those URLs automatically.  
- **If you have CSV data:** It’s easier because the content is already provided. No need to fetch anything manually.  

**6. What Output Does it Provide?**  
The hierarchical clustering model provides:  
- **Clustered groups:** For example, it might group 500 webpages into 10 clusters based on their topics.  
- **Visual tree structure (Dendrogram):** This is a diagram that shows how pages are grouped at different levels of similarity.  
- **Actionable insights:** Suggestions for improving internal linking, optimizing categories, or identifying content gaps.  

**7. How Does it Enhance Content Strategy and Internal Linking?**  
- **Improved structure:** Helps organize your site by clustering similar content together, making navigation logical and user-friendly.  
- **Better internal linking:** Pages within the same cluster can link to each other, boosting SEO and user engagement.

# **1st Part : Data Collection and Scraping Content**

````python
# Import necessary libraries
import requests  # To fetch the content from URLs
from bs4 import BeautifulSoup  # To parse HTML and extract metadata

# Step 1: Define a list of URLs to scrape content from
"""
This list contains the URLs of the webpages we want to analyze.
Each URL represents a specific webpage on the website.
The goal is to gather meaningful data (text content, title, description) from each page.
"""
urls = [
    'https://thatware.co/',
    'https://thatware.co/services/',
    'https://thatware.co/advanced-seo-services/',
    'https://thatware.co/digital-marketing-services/',
    'https://thatware.co/business-intelligence-services/',
    'https://thatware.co/link-building-services/',
    'https://thatware.co/branding-press-release-services/',
    'https://thatware.co/conversion-rate-optimization/',
    'https://thatware.co/social-media-marketing/',
    'https://thatware.co/content-proofreading-services/',
    'https://thatware.co/website-design-services/',
    'https://thatware.co/web-development-services/',
    'https://thatware.co/app-development-services/',
    'https://thatware.co/website-maintenance-services/',
    'https://thatware.co/bug-testing-services/',
    'https://thatware.co/software-development-services/',
    'https://thatware.co/competitor-keyword-analysis/'
]

# Step 2: Function to scrape content and metadata from each URL
"""
This function performs the following tasks:
1. Sends a request to the webpage URL to fetch the HTML content.
2. Parses the HTML using BeautifulSoup to extract:
   - Title: The title of the webpage.
   - Meta Description: A short description found in the webpage's metadata.
   - Main Content: The main text content extracted from all <p> tags (paragraphs).
Purpose: To gather structured data from each webpage for analysis.
"""
def scrape_content(url):
    try:
        # Fetch the webpage HTML
        response = requests.get(url)  # Sends an HTTP GET request to the URL
        response.raise_for_status()  # Ensures the request was successful
        soup = BeautifulSoup(response.text, 'html.parser')  # Parse HTML content

        # Extract the title of the page
        title = soup.title.string if soup.title else 'No Title Found'

        # Extract the meta description of the page
        meta_description = soup.find('meta', attrs={'name': 'description'})
        meta_description = meta_description['content'] if meta_description else 'No Meta Description Found'

        # Extract the main text content by gathering all paragraph tags
        paragraphs = soup.find_all('p')  # Finds all <p> (paragraph) tags in the HTML
        text_content = ' '.join([para.get_text() for para in paragraphs])  # Combine all paragraph text into a single string

        return title, meta_description, text_content  # Return the extracted data as a tuple
    except Exception as e:
        # Handle errors gracefully and return an error message
        return "Error", "Error", f"Error: {str(e)}"

# Step 3: Scrape content for all URLs in the list
"""
Purpose: Loop through each URL and scrape the content using the scrape_content function.
The scraped data (title, description, content) is stored in a structured list called `data`.
"""
data = []  # Create an empty list to store the scraped data
for url in urls:
    # Call the scrape_content function for each URL
    title, description, content = scrape_content(url)

    # Append the scraped data as a dictionary to the `data` list
    data.append({
        'url': url,
        'title': title,
        'description': description,
        'content': content
    })

# Step 4: Output the scraped data
"""
Print the scraped content to verify the output.
This step ensures that the content from each webpage is correctly collected.
"""
print("\nScraped Data:")
for page in data:
    print(f"URL: {page['url']}")
    print(f"Title: {page['title']}")
    print(f"Meta Description: {page['description']}")
    print(f"Content (First 300 Characters): {page['content'][:300]}...\n")  # Display only the first 300 characters of the content
````

# **2nd Part:** **Numerical Transformation and Clustering**

````python
# Import the necessary libraries for clustering and visualization
from sklearn.feature_extraction.text import TfidfVectorizer  # To convert text data into numerical vectors
from scipy.cluster.hierarchy import dendrogram, linkage  # For hierarchical clustering and visualization
import matplotlib.pyplot as plt  # To create visualizations

# Step 5: Convert text content into numerical format using TF-IDF
"""
TF-IDF (Term Frequency-Inverse Document Frequency):
1. Converts textual data into numerical vectors.
2. Measures the importance of each word in a document relative to all other documents in the dataset.
3. Creates a matrix where rows represent documents and columns represent unique words, and the values indicate the significance of each word in the document.

Purpose: To prepare the textual data for mathematical analysis and clustering.
"""
vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)  # Only use the top 1000 most important words
X = vectorizer.fit_transform(documents)  # Transform the documents into a TF-IDF matrix

# Output the shape of the TF-IDF matrix
"""
The shape of the matrix represents the number of documents (rows) and the number of features (columns).
This confirms that the text data has been successfully converted into numerical format.
"""
print("\nTF-IDF Matrix Shape:")
print(f"Documents: {X.shape[0]}, Features: {X.shape[1]}")

# Step 6: Perform hierarchical clustering
"""
Hierarchical Clustering:
1. Groups the documents (web pages) based on their similarity.
2. Uses the Ward linkage method to minimize the variance within clusters.

Purpose: To identify groups (clusters) of similar pages based on their textual content.
"""
linkage_matrix = linkage(X.toarray(), method='ward')  # Perform clustering using Ward's method

# Step 7: Visualize the clustering using a dendrogram
"""
A dendrogram visually represents the clustering process:
1. URLs are shown along the x-axis.
2. The height of the lines on the y-axis indicates the dissimilarity between pages.

Purpose: To visualize how the pages are grouped into clusters and understand their relationships.
"""
plt.figure(figsize=(15, 8))  # Set the size of the plot
dendrogram(
    linkage_matrix,  # The clustering results
    labels=[item['url'] for item in data],  # Use the URLs as labels
    orientation='top',  # The tree grows downward
    leaf_rotation=90,  # Rotate the labels for readability
    leaf_font_size=10  # Font size for labels
)
plt.title('Hierarchical Clustering Dendrogram')  # Title of the plot
plt.xlabel('Web Pages')  # Label for the x-axis
plt.ylabel('Distance (Dissimilarity)')  # Label for the y-axis
plt.show()  # Display the dendrogram

# Output Explanation
"""
The dendrogram provides a visual representation of the clusters:
1. URLs (webpages) are shown as individual labels at the bottom of the plot.
2. Shorter branches (low height) indicate that the pages are very similar and belong to the same cluster.
3. Longer branches (higher height) indicate less similarity between pages or clusters.

Use Case:
The dendrogram helps identify clusters of similar pages and decide how to group them for better content strategy.
"""
````

- **Why This is Important:**

- **Example:**
   Suppose the word "SEO" appears frequently on one webpage but not on others. TF-IDF assigns a higher weight to "SEO" for that page. On the other hand, words like "the" or "and," which are common across all documents, get very low weights.

````python
# Import necessary libraries
import requests  # To fetch the content from URLs
from bs4 import BeautifulSoup  # To parse HTML and extract metadata
from sklearn.feature_extraction.text import TfidfVectorizer  # To convert textual data into numerical format
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster  # For hierarchical clustering and visualizations
import matplotlib.pyplot as plt  # To create visual plots (dendrogram)

# Step 1: Define a list of URLs to scrape content from
"""
This list contains the URLs of webpages we want to analyze for clustering.
Each URL represents a page on the website whose content will be grouped based on similarity.
"""
urls = [
    'https://thatware.co/',
    'https://thatware.co/services/',
    'https://thatware.co/advanced-seo-services/',
    'https://thatware.co/digital-marketing-services/',
    'https://thatware.co/business-intelligence-services/',
    'https://thatware.co/link-building-services/',
    'https://thatware.co/branding-press-release-services/',
    'https://thatware.co/conversion-rate-optimization/',
    'https://thatware.co/social-media-marketing/',
    'https://thatware.co/content-proofreading-services/',
    'https://thatware.co/website-design-services/',
    'https://thatware.co/web-development-services/',
    'https://thatware.co/app-development-services/',
    'https://thatware.co/website-maintenance-services/',
    'https://thatware.co/bug-testing-services/',
    'https://thatware.co/software-development-services/',
    'https://thatware.co/competitor-keyword-analysis/'
]

# Step 2: Function to scrape content and metadata from each URL
"""
This function:
1. Fetches the webpage HTML using the URL.
2. Extracts the title, meta description, and main text content from the page.
Purpose: To gather meaningful information (text) from each page for clustering analysis.
"""
def scrape_content(url):
    try:
        # Fetch the webpage HTML
        response = requests.get(url)  # Sends a GET request to the URL to fetch the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')  # Parses the HTML content using BeautifulSoup

        # Extract the title of the page
        title = soup.title.string if soup.title else 'No Title Found'

        # Extract the meta description of the page
        meta_description = soup.find('meta', attrs={'name': 'description'})
        meta_description = meta_description['content'] if meta_description else 'No Meta Description Found'

        # Extract the main text content by gathering all paragraphs
        paragraphs = soup.find_all('p')  # Finds all <p> (paragraph) tags in the HTML
        text_content = ' '.join([para.get_text() for para in paragraphs])  # Combines all paragraph text into a single string

        return title, meta_description, text_content  # Return the extracted data
    except Exception as e:
        return "Error", "Error", "Error: " + str(e)  # Return error information if fetching fails

# Step 3: Scrape content for all URLs
"""
Purpose: Loop through the list of URLs and scrape the content for each webpage.
Store the title, meta description, and text content for every page in a list called `data`.
"""
data = []  # Create an empty list to store the scraped data
for url in urls:
    title, description, content = scrape_content(url)  # Call the scrape_content function
    data.append({  # Append the extracted content as a dictionary
        'url': url,
        'title': title,
        'description': description,
        'content': content
    })

# Step 4: Prepare content for clustering
"""
Extract only the main text content from the scraped data.
Purpose: The content (text) will be used to calculate similarities between pages for clustering.
"""
documents = [item['content'] for item in data]  # Extract the 'content' field from each dictionary in `data`

# Step 5: Convert text content into numerical format using TF-IDF
"""
TF-IDF (Term Frequency-Inverse Document Frequency):
1. Converts textual data into numerical vectors.
2. Measures the importance of each word in a document relative to all other documents.
Purpose: To create a numerical representation of the content so it can be analyzed mathematically.
"""
vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)  # Use the top 1000 most important words
X = vectorizer.fit_transform(documents)  # Transform the documents into a TF-IDF matrix

# Step 6: Perform hierarchical clustering
"""
Hierarchical clustering groups pages based on their similarity.
1. Linkage method ('ward') ensures that variance within clusters is minimized.
2. The clustering output will help us visualize how pages are grouped.
Purpose: To identify groups (clusters) of related pages.
"""
linkage_matrix = linkage(X.toarray(), method='ward')  # Perform clustering using Ward's method

# Step 7: Visualize the clustering using a dendrogram
"""
A dendrogram visually shows the clustering process.
1. URLs are shown on the x-axis.
2. Similarity between pages is represented by the height of lines on the y-axis.
Purpose: To provide a visual representation of how pages are grouped.
"""
plt.figure(figsize=(15, 8))  # Set the size of the plot
dendrogram(linkage_matrix, labels=[item['url'] for item in data], orientation='top', leaf_rotation=90)
plt.title('Hierarchical Clustering Dendrogram')  # Add a title to the plot
plt.xlabel('Web Pages')  # Label for the x-axis
plt.ylabel('Distance')  # Label for the y-axis
plt.show()  # Display the dendrogram

# Step 8: Extract clusters and actionable insights
"""
1. Cut the dendrogram at a threshold distance to create clusters.
2. Assign each page to a cluster based on its similarity to others.
Purpose: To group similar pages into clusters and generate actionable insights.
"""
threshold = 1.0  # Define the threshold distance for cutting the dendrogram
clusters = fcluster(linkage_matrix, threshold, criterion='distance')  # Assign cluster labels to pages

# Group pages by their cluster ID
clustered_pages = {}
for i, cluster_id in enumerate(clusters):
    if cluster_id not in clustered_pages:
        clustered_pages[cluster_id] = []
    clustered_pages[cluster_id].append(data[i]['url'])  # Append each URL to its respective cluster

# Display the clustered pages
print("\nClustered Pages:")
for cluster_id, urls in clustered_pages.items():
    print(f"\nCluster {cluster_id}:")
    for url in urls:
        print(f" - {url}")

# Generate actionable insights for each cluster
"""
For each cluster:
1. Suggest a common theme based on the URLs.
2. Recommend internal linking between pages in the cluster.
3. Highlight content gaps for clusters with only one page.
Purpose: To provide practical recommendations for improving website organization and content strategy.
"""
print("\nActionable Insights:")
for cluster_id, urls in clustered_pages.items():
    print(f"\nFor Cluster {cluster_id}:")
    if len(urls) > 1:
        print(f" - Group these pages together under a common theme.")
        print(f" - Suggested Theme: {', '.join(urls[:2])}")  # Suggest a theme based on the first two URLs
        print(f" - Improve internal linking by linking these pages together.")
        print(f" - Ensure the pages in this cluster have consistent categories or menus.")
    else:
        print(f" - This cluster has only one page. Consider creating more content related to this topic.")
        print(f" - Suggested Related Content: Add pages related to this topic to expand the cluster.")

# Step 9: Summary
"""
Summarize the findings:
1. Use clusters to create logical categories for the website.
2. Improve navigation and SEO by linking pages within the same cluster.
3. Create new content for clusters with only one or very few pages.
Purpose: To summarize the project's recommendations for enhancing the website's structure.
"""
print("\nSummary:")
print("1. Use the clustered pages to organize your website into logical categories or sections.")
print("2. Improve user navigation and SEO by linking pages within the same cluster.")
print("3. Address content gaps by creating new pages for clusters with only one or very few pages.")
````
