# **1. Part 1: Scraping Web Content**
**Purpose:**  
This part fetches the content of webpages from a list of URLs and saves it into a CSV file.

**Steps in this Part:**
- **Mount Google Drive:**  
  Connects to Google Drive to access the file containing URLs and save the output CSV.

- **Define `scrape_content` Function:**  
  Fetches the HTML content of a webpage and extracts readable text (e.g., paragraphs).

- **Read URLs from File:**  
  Loads the list of webpage URLs from a file into memory.

- **Loop Through URLs:**  
  Iterates over each URL, scrapes the content using `scrape_content`, and saves the result in a list.

- **Save Scraped Data to CSV:**  
  After scraping all URLs, saves the content to a CSV file for further processing.

---


````python
# File Path for the dataset (URLs) and output CSV
FILE_PATH = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/seo_services_urls.txt"
OUTPUT_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/scraped_content.csv"

# Import required libraries
import pandas as pd  # For handling datasets
import requests  # For sending HTTP requests to fetch web pages
from bs4 import BeautifulSoup  # For parsing and extracting text from web pages
from google.colab import drive  # To mount Google Drive and access files

# Step 1: Mount Google Drive
# Explanation:
# - Google Colab is a cloud environment and cannot directly access your local files.
# - We mount Google Drive to access files stored in your Google Drive.
drive.mount('/content/drive')  # This command links Google Drive to your Colab environment.

# Step 2: Function to Scrape Content from a Single URL
# Explanation:
# - This function fetches the content of a webpage using its URL.
# - It extracts only the visible text (e.g., paragraphs) from the webpage.
def scrape_content(url):
    """
    Scrapes visible text content from a webpage given its URL.
    - Fetches the webpage using the requests library.
    - Parses the HTML content using BeautifulSoup and extracts text from <p> tags.
    """
    try:
        # Send an HTTP GET request to fetch the webpage
        response = requests.get(url, timeout=10)  # Timeout ensures we don't wait indefinitely.
        response.raise_for_status()  # Raises an HTTPError for bad responses (e.g., 404, 500).

        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract visible text from all <p> tags (paragraphs)
        text = ' '.join([p.get_text() for p in soup.find_all('p')])

        # Return the cleaned text content
        return text.strip()  # `.strip()` removes any leading/trailing spaces or newlines.

    except Exception as e:
        # If an error occurs (e.g., network error, invalid URL), print the error message.
        print(f"Error scraping {url}: {e}")
        return None  # Return None to indicate that scraping failed.

# Step 3: Main Function to Scrape All URLs and Save to a CSV File
# Explanation:
# - This function reads the list of URLs from a file.
# - It loops through each URL, scrapes the content, and saves the results to a CSV file.
def main():
    # Step 3.1: Load URLs from the file
    # Explanation:
    # - The URLs are stored in a text file, with one URL per line.
    # - We use pandas to read the file into a DataFrame for easier handling.
    try:
        urls = pd.read_csv(FILE_PATH, header=None, names=["URLs"])  # Load URLs into a DataFrame
    except FileNotFoundError:
        # If the file is not found, print an error message and stop the script.
        print("File not found. Please check the FILE_PATH.")
        return  # Stop the function execution.

    # Step 3.2: Initialize a list to store the scraped data
    # Explanation:
    # - This list will hold dictionaries with two keys: "URL" and "Content".
    # - Each dictionary represents one scraped webpage.
    scraped_data = []

    # Step 3.3: Loop through each URL and scrape content
    # Explanation:
    # - For each URL in the DataFrame, scrape the content using the `scrape_content` function.
    for idx, row in urls.iterrows():
        # Print progress (e.g., "Scraping URL 1/100...")
        print(f"Scraping URL {idx + 1}/{len(urls)}: {row['URLs']}")

        # Call the `scrape_content` function to get the webpage content
        content = scrape_content(row["URLs"])

        # If scraping is successful (i.e., content is not None), save the data
        if content:
            scraped_data.append({"URL": row["URLs"], "Content": content})  # Append to the list

        # Step 3.4: Preview the scraped content for the first 20 URLs
        # Explanation:
        # - To verify that the scraping is working, we display a short preview for the first 20 URLs.
        if idx < 20:  # Only preview the first 20 URLs
            print(f"\nPreview of URL {idx + 1}:\n")
            print(content[:500])  # Display the first 500 characters of the content
            print("\n" + "-" * 80 + "\n")  # Separator for better readability

    # Step 3.5: Save the scraped data to a CSV file
    # Explanation:
    # - The scraped data is saved in a CSV file for further analysis.
    # - Each row in the CSV contains the URL and the corresponding scraped content.
    scraped_df = pd.DataFrame(scraped_data)  # Convert the list of dictionaries into a DataFrame
    scraped_df.to_csv(OUTPUT_CSV, index=False)  # Save the DataFrame to a CSV file (no index column)
    print(f"Scraped data saved to {OUTPUT_CSV}")  # Confirm that the file has been saved.

# Step 4: Run the Main Function
# Explanation:
# - The `if __name__ == "__main__"` block ensures that the script runs only when executed directly.
# - This prevents accidental execution if the script is imported as a module.
if __name__ == "__main__":
    main()
````

# **2. Part 2: Preprocessing Text Data**
**Purpose:**  
This part cleans the raw text data scraped from webpages to prepare it for analysis.

**Steps in this Part:**
- **Mount Google Drive:**  
  Ensures the script can access input and output files stored in your Google Drive.

- **Download NLTK Data Files:**  
  Downloads stopwords (common words like "the", "is") and lemmatization rules to clean the text.

- **Define `preprocess_text` Function:**  
  - Removes unwanted characters (e.g., punctuation, numbers).
  - Converts text to lowercase.
  - Removes stopwords.
  - Reduces words to their base form (e.g., "running" → "run").

- **Load Scraped Data:**  
  Reads the scraped text data from the CSV file created in Part 1.

- **Clean the Text:**  
  Applies the `preprocess_text` function to clean each webpage's content.

- **Save Preprocessed Data to CSV:**  
  Saves the cleaned text data into a new CSV file.

---


````python
# File Paths for Input and Output
SCRAPED_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/scraped_content.csv"
PROCESSED_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/preprocessed_content.csv"

# Import Required Libraries
import pandas as pd  # For handling tabular data
import re  # For text cleaning (e.g., removing special characters)
import nltk  # For natural language processing tasks
from nltk.corpus import stopwords  # To remove common, meaningless words (e.g., "the", "is")
from nltk.stem import WordNetLemmatizer  # To reduce words to their base form (e.g., "running" -> "run")
from google.colab import drive  # To access Google Drive in Colab

# Step 1: Mount Google Drive
# Why this step?
# - Google Colab runs in the cloud and doesn't directly access your files.
# - We mount Google Drive to read and save files stored in your Google Drive.
drive.mount('/content/drive')

# Step 2: Download NLTK Data Files
# Why this step?
# - NLTK is a library for processing text data, but it requires some files (like stopwords and lemmatization rules) to work.
# - We ensure these files are downloaded before running the script.
nltk.download('stopwords')  # Download stopwords for English
nltk.download('wordnet')  # Download WordNet for lemmatization

# Step 3: Initialize NLP Tools
# Explanation:
# - We set up a list of stopwords and initialize the lemmatizer for later use in text cleaning.
stop_words = set(stopwords.words('english'))  # Set of common words to exclude
lemmatizer = WordNetLemmatizer()  # To reduce words to their base form

# Step 4: Function to Clean and Preprocess Text
# Explanation:
# - This function cleans and preprocesses the text to make it suitable for analysis.
# - It removes unnecessary characters, stopwords, and performs lemmatization.

def preprocess_text(text):
    """
    Cleans and preprocesses the given text by:
    1. Removing special characters, punctuation, and numbers.
    2. Converting text to lowercase.
    3. Removing common stopwords.
    4. Performing lemmatization (reducing words to their base form).

    Why this step?
    - Text data from web scraping is often messy (e.g., contains HTML tags, special characters, etc.).
    - Preprocessing standardizes and simplifies the text for easier analysis.
    """
    try:
        # Remove special characters, punctuation, and numbers
        text = re.sub(r'[^a-zA-Z\s]', '', text)  # Keep only letters and spaces

        # Convert text to lowercase for consistency
        text = text.lower()

        # Tokenize the text into words
        words = text.split()

        # Remove stopwords (e.g., "the", "is", "in")
        words = [word for word in words if word not in stop_words]

        # Perform lemmatization to reduce words to their base form
        words = [lemmatizer.lemmatize(word) for word in words]

        # Rejoin the cleaned words into a single string
        return ' '.join(words)

    except Exception as e:
        # If an error occurs during preprocessing, print the error and return None
        print(f"Error during preprocessing: {e}")
        return None

# Step 5: Main Function to Preprocess the Dataset
# Explanation:
# - This function reads the scraped content, preprocesses it, and saves the results to a new file.
def main():
    try:
        # Step 5.1: Load the scraped content
        # Why this step?
        # - The input file contains web content scraped from various URLs.
        # - We need to load this data to process it.
        scraped_data = pd.read_csv(SCRAPED_CSV)  # Load the scraped data into a DataFrame

        # Step 5.2: Preprocess the 'Content' column
        # Why this step?
        # - The 'Content' column contains raw text scraped from webpages, which needs cleaning.
        print("Preprocessing text content...")
        scraped_data['Processed_Content'] = scraped_data['Content'].apply(preprocess_text)  # Apply preprocessing

        # Step 5.3: Save the preprocessed data to a new CSV file
        # Why this step?
        # - The cleaned text is saved for further analysis (e.g., topic modeling, similarity analysis).
        scraped_data.to_csv(PROCESSED_CSV, index=False)  # Save preprocessed data to a CSV file
        print(f"Preprocessed data saved to {PROCESSED_CSV}")  # Confirm that the data has been saved

        # Step 5.4: Display a preview of the preprocessed data
        # Why this step?
        # - To verify that the preprocessing worked as expected.
        print("\nPreview of Preprocessed Data:")
        print(scraped_data[['URL', 'Processed_Content']].head(40))  # Display the first 40 rows

    except FileNotFoundError:
        # If the input file is missing, print an error message
        print("Scraped content file not found. Please check the file path.")
    except Exception as e:
        # Catch any other errors and print them
        print(f"An error occurred: {e}")

# Step 6: Execute the Main Function
# Explanation:
# - This ensures that the script runs when executed directly.
if __name__ == "__main__":
    main()
````

# **3. Part 3: Extracting Topics with LSA**
**Purpose:**  
This part performs Latent Semantic Analysis (LSA) to extract topics from the cleaned text data.

**Steps in this Part:**
- **Load Preprocessed Data:**  
  Reads the cleaned text data from the CSV file created in Part 2.

- **Apply TF-IDF Vectorization:**  
  Converts text into numerical values using **TF-IDF** (Term Frequency-Inverse Document Frequency), which gives more weight to important words.

- **Apply LSA (Latent Semantic Analysis):**  
  Reduces the high-dimensional TF-IDF matrix into a smaller set of **topics**.

- **Extract Keywords for Topics:**  
  For each topic, identifies the top keywords that define the topic.

- **Assign Topics to Documents:**  
  Assigns the most relevant topic to each webpage based on the content.

- **Save LSA Results to CSV:**  
  Saves the topics and their associated documents to a CSV file.

- **Generate Webpage Mapping:**  
  Maps document IDs to webpage titles for easier reference.

- **Generate Heatmap:**  
  Creates a visual heatmap of document similarities to identify closely related webpages.

---


````python
# Import Required Libraries
import numpy as np  # For numerical computations, such as matrix operations
import pandas as pd  # For handling tabular data
from sklearn.feature_extraction.text import TfidfVectorizer  # For converting text to numerical representation
from sklearn.decomposition import TruncatedSVD  # For Latent Semantic Analysis (LSA)
import seaborn as sns  # For creating heatmaps
import matplotlib.pyplot as plt  # For plotting graphs
from google.colab import drive  # For accessing Google Drive in Colab

# Mount Google Drive
# Why this step?
# - This allows the script to access files stored in your Google Drive, which is essential for reading input files and saving output files.
drive.mount('/content/drive')

# File Paths
# Define paths for input and output files.
PROCESSED_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/preprocessed_content.csv"  # Input file
LSA_RESULTS_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/lsa_results.csv"  # Output for LSA results
ENHANCED_HEATMAP_PATH = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/enhanced_heatmap.png"  # Heatmap output
SIMILARITY_MATRIX_PATH = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/similarity_matrix.npy"  # Similarity matrix
WEBPAGE_MAPPING_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/webpage_mapping.csv"  # Document ID to webpage title mapping

def balanced_lsa_model():
    """
    This function performs Latent Semantic Analysis (LSA) on preprocessed text data.
    It uses a series of steps to extract meaningful topics from the text and visualize the similarity between documents.
    """
    try:
        # Step 1: Load Preprocessed Data
        # Why this step?
        # - This step reads the cleaned and preprocessed text data for further analysis.
        # - The data is assumed to be saved as a CSV file after preprocessing.
        print("Loading preprocessed content...")
        data = pd.read_csv(PROCESSED_CSV)
        print(f"Loaded {data.shape[0]} documents.\n")  # Show the number of documents loaded

        # Step 2: TF-IDF Vectorization
        # Why this step?
        # - Converts text data into a numerical format (TF-IDF matrix) suitable for machine learning.
        # - TF-IDF emphasizes important words in a document relative to a collection of documents.
        print("Vectorizing text data with TF-IDF...")
        vectorizer = TfidfVectorizer(max_features=3000, stop_words='english')  # Limit to 3000 features for efficiency
        tfidf_matrix = vectorizer.fit_transform(data['Processed_Content'])  # Convert text to numerical representation
        feature_names = vectorizer.get_feature_names_out()  # Get the names of the words in the TF-IDF matrix
        print(f"TF-IDF Vectorization Complete. Shape: {tfidf_matrix.shape}\n")  # Print the dimensions of the matrix

        # Step 3: Apply Latent Semantic Analysis (LSA)
        # Why this step?
        # - LSA reduces the high-dimensional TF-IDF matrix into a smaller set of topics (dimensionality reduction).
        # - Each topic captures key themes in the data.
        print("Applying Latent Semantic Analysis (LSA)...")
        num_topics = 20  # Specify the number of topics to extract
        lsa_model = TruncatedSVD(n_components=num_topics, random_state=42)  # Initialize the LSA model
        lsa_matrix = lsa_model.fit_transform(tfidf_matrix)  # Transform the TF-IDF matrix into a topic representation
        print(f"LSA Transformation Complete. Topics Extracted: {num_topics}\n")

        # Step 4: Extract Topics and Save Results
        # Why this step?
        # - Extract and display the top keywords for each topic to understand what each topic represents.
        print("Extracting topics and keywords...")
        topics = []
        for idx, component in enumerate(lsa_model.components_):
            # For each topic, select the top 10 words
            terms = [feature_names[i] for i in component.argsort()[-10:]]
            topics.append(" + ".join(terms))  # Combine the top words into a single string
            print(f"Topic {idx + 1}: {topics[-1]}")  # Print the topic keywords

        # Assign each document its most dominant topic
        data['Dominant_Topic'] = lsa_matrix.argmax(axis=1)  # Find the topic with the highest score for each document
        data['Topic_Keywords'] = data['Dominant_Topic'].apply(lambda x: topics[x])  # Add the keywords for each topic

        # Save the LSA results to a CSV file
        data.to_csv(LSA_RESULTS_CSV, index=False)
        print(f"\nLSA Results saved to: {LSA_RESULTS_CSV}")

        # Step 5: Generate Document-to-Webpage Mapping
        # Why this step?
        # - Maps each document ID to a webpage title for easier interpretation.
        print("\nHandling webpage mapping...")
        webpage_titles = [
            "Homepage", "Blog Page", "Contact Us", "Services Page", "About Us",
            # Add more titles corresponding to your documents...
        ]
        webpage_mapping = pd.DataFrame({
            'Document ID': range(len(webpage_titles)),
            'Webpage Title': webpage_titles[:len(data)]  # Ensure mapping matches the number of documents
        })
        webpage_mapping.to_csv(WEBPAGE_MAPPING_CSV, index=False)
        print(f"Webpage mapping saved to: {WEBPAGE_MAPPING_CSV}")

        # Step 6: Generate Topic Distribution Summary
        # Why this step?
        # - Summarizes the number of documents associated with each topic.
        print("\nTopic Distribution Summary:")
        topic_counts = data['Dominant_Topic'].value_counts()
        for topic, count in topic_counts.items():
            print(f"Topic {topic}: {count} documents")

        # Step 7: Compute and Save Similarity Matrix
        # Why this step?
        # - Computes a similarity matrix to find how similar documents are to one another.
        print("\nGenerating similarity matrix...")
        similarity_matrix = np.dot(lsa_matrix, lsa_matrix.T)  # Compute the similarity between documents
        np.save(SIMILARITY_MATRIX_PATH, similarity_matrix)  # Save the similarity matrix
        print(f"Similarity matrix saved to: {SIMILARITY_MATRIX_PATH}")

        # Step 8: Generate an Enhanced Heatmap
        # Why this step?
        # - Visualizes the similarity matrix as a heatmap for easier analysis of document relationships.
        print("\nGenerating enhanced heatmap...")
        plt.figure(figsize=(12, 10))  # Set the size of the heatmap
        heatmap = sns.heatmap(
            similarity_matrix,
            cmap="RdYlGn",  # Red-to-green color map
            annot=False,  # Disable cell annotations for better readability
            cbar_kws={'label': 'Similarity Score'}  # Add a color bar with a label
        )

        # Add titles and labels to the heatmap
        heatmap.set_title("Document Semantic Similarity Heatmap", fontsize=16)
        heatmap.set_xlabel("Documents", fontsize=12)
        heatmap.set_ylabel("Documents", fontsize=12)

        # Add a footnote explaining the heatmap
        plt.figtext(
            0.5, -0.1,
            "This heatmap shows the similarity between documents based on topics.\n"
            "Red = High Similarity (e.g., Documents on similar topics).\n"
            "Green = Low Similarity (e.g., Documents on unrelated topics).",
            wrap=True, horizontalalignment='center', fontsize=12
        )

        # Save and display the heatmap
        plt.tight_layout()
        plt.savefig(ENHANCED_HEATMAP_PATH)
        plt.show()
        print(f"Enhanced heatmap saved to: {ENHANCED_HEATMAP_PATH}")

    except Exception as e:
        # Print an error message if something goes wrong
        print(f"An error occurred: {e}")

# Execute the function
balanced_lsa_model()
````

# **4. Part 4: Content Recommendations and Gap Analysis**
**Purpose:**  
This part generates actionable insights, such as related webpage recommendations, high-value keywords, and content gaps.

**Steps in this Part:**
- **Load LSA Results and Webpage Mapping:**  
  Reads the LSA topics and webpage mapping from the files created in Part 3.

- **Generate Content Recommendations:**  
  Finds similar webpages for each document using a similarity matrix and creates a list of recommendations.

- **Extract High-Value Keywords:**  
  Identifies the most important keywords for each topic, helping optimize content.

- **Highlight Content Gaps:**  
  Analyzes the number of documents per topic to find underrepresented topics. Suggests creating more content for these topics.

- **Save Results to Files:**  
  Saves the recommendations, keywords, and content gaps to separate files.

---


````python
# Import necessary libraries
import pandas as pd  # For working with tabular data
import numpy as np  # For numerical computations
from google.colab import drive  # To access files in Google Drive

# Step 1: Mount Google Drive
# Why this step?
# - To access input and output files stored in Google Drive.
# - This is necessary because the files used in this code are located in a specific folder in the user's Drive.
drive.mount('/content/drive')

# Step 2: File Paths
# Why this step?
# - Define file paths for input and output to ensure the script reads and writes files at the correct locations.
LSA_RESULTS_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/lsa_results.csv"  # Input LSA results
CONTENT_RECOMMENDATIONS_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/content_recommendations.csv"  # Output for recommendations
TOPIC_KEYWORDS_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/topic_keywords.csv"  # Output for topic keywords
CONTENT_GAPS_TXT = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/content_gaps.txt"  # Output for content gap analysis
SIMILARITY_MATRIX_PATH = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/similarity_matrix.npy"  # Input similarity matrix
WEBPAGE_MAPPING_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/webpage_mapping.csv"  # Input webpage mapping

def lsa_content_analysis():
    """
    Perform content analysis using Latent Semantic Analysis (LSA) results.
    The function:
    1. Generates content recommendations for webpages.
    2. Extracts high-value keywords for topics.
    3. Highlights content gaps for improvement.
    """
    try:
        # Step 3: Load LSA Results
        # Why this step?
        # - The LSA results contain processed documents and their dominant topics, essential for generating recommendations and analyzing gaps.
        print("Loading LSA results...")
        data = pd.read_csv(LSA_RESULTS_CSV)
        print(f"Loaded {data.shape[0]} documents.\n")  # Show the number of documents loaded

        # Step 4: Load Webpage Mapping
        # Why this step?
        # - The mapping file links each document (identified by its ID) to its corresponding webpage title for clarity.
        print("Loading webpage mapping...")
        webpage_mapping = pd.read_csv(WEBPAGE_MAPPING_CSV)

        # Step 5: Validate Webpage Mapping
        # Why this step?
        # - Ensure every document in the LSA results has a corresponding title.
        # - If the mapping is incomplete, add placeholders for missing entries.
        if len(webpage_mapping) < data.shape[0]:
            print("Warning: Webpage mapping is incomplete. Adding placeholders for missing entries.")
            missing_entries = [
                {'Document ID': missing_id, 'Webpage Title': f"Document {missing_id}"}
                for missing_id in range(len(webpage_mapping), data.shape[0])
            ]
            webpage_mapping = pd.concat([webpage_mapping, pd.DataFrame(missing_entries)], ignore_index=True)
        print("Webpage mapping loaded successfully.\n")

        # Step 6: Generate Content Recommendations
        # Why this step?
        # - Use the similarity matrix to find the most similar documents for each webpage.
        print("Generating content recommendations...")
        similarity_matrix = np.load(SIMILARITY_MATRIX_PATH)  # Load precomputed similarity matrix

        recommendations = []  # List to store recommendations
        for i, row in enumerate(similarity_matrix):
            # Find the top 5 most similar documents (excluding the document itself)
            similar_docs = row.argsort()[::-1][1:6]  # Sort in descending order and exclude the first element
            similar_titles = []  # List to store titles of similar documents
            for doc in similar_docs:
                # If the document ID is mapped, retrieve its title
                if doc in webpage_mapping['Document ID'].values:
                    title = webpage_mapping.loc[webpage_mapping['Document ID'] == doc, 'Webpage Title'].values[0]
                else:
                    title = f"Document {doc} (Title Missing)"
                similar_titles.append(title)  # Add the title to the list of similar titles
            # Store the recommendations for the current document
            recommendations.append({
                'Webpage_Title': webpage_mapping.loc[webpage_mapping['Document ID'] == i, 'Webpage Title'].values[0],
                'Similar_Webpages': similar_titles
            })

        # Save the recommendations to a CSV file
        recommendations_df = pd.DataFrame(recommendations)
        recommendations_df.to_csv(CONTENT_RECOMMENDATIONS_CSV, index=False)
        print(f"Content recommendations saved to {CONTENT_RECOMMENDATIONS_CSV}\n")

        # Preview the first 5 recommendations in the console
        print("Preview of Content Recommendations:")
        for _, rec in recommendations_df.head(5).iterrows():
            print(f"Webpage: {rec['Webpage_Title']}")
            print(f" - Similar Pages: {', '.join(rec['Similar_Webpages'])}")

        # Step 7: Extract High-Value Keywords
        # Why this step?
        # - Identify the most important keywords for each topic to provide actionable insights for content improvement.
        print("\nExtracting high-value keywords...")
        topic_keywords = data[['Dominant_Topic', 'Topic_Keywords']].drop_duplicates()
        topic_keywords = topic_keywords.sort_values('Dominant_Topic').reset_index(drop=True)

        # Save the keywords to a CSV file
        topic_keywords.to_csv(TOPIC_KEYWORDS_CSV, index=False)
        print(f"Topic keywords saved to {TOPIC_KEYWORDS_CSV}\n")

        # Preview the first 5 topic keywords in the console
        print("Preview of Topic Keywords:")
        print(topic_keywords.head(5).to_string(index=False))

        # Step 8: Highlight Content Gaps
        # Why this step?
        # - Analyze topic distribution to identify topics with insufficient content and suggest improvements.
        print("\nHighlighting content gaps...")
        topic_distribution = data['Dominant_Topic'].value_counts()  # Count documents per topic
        mean_documents_per_topic = topic_distribution.mean()  # Calculate the average number of documents per topic
        content_gaps = topic_distribution[topic_distribution < mean_documents_per_topic]  # Identify underrepresented topics

        # Save the content gaps to a text file
        with open(CONTENT_GAPS_TXT, 'w') as f:
            f.write("Content Gaps Report\n")
            f.write("=" * 50 + "\n")
            f.write(f"Average Documents per Topic: {mean_documents_per_topic:.2f}\n")
            f.write("Topics with fewer than average documents:\n\n")
            for topic, count in content_gaps.items():
                f.write(f"Topic {topic}: {count} documents\n")
        print(f"Content gaps report saved to {CONTENT_GAPS_TXT}\n")

        # Preview content gaps in the console
        print("Preview of Content Gaps:")
        print(f"Average Documents per Topic: {mean_documents_per_topic:.2f}")
        for topic, count in content_gaps.items():
            print(f"Topic {topic}: {count} documents")
            print(f"  Suggestion: Add more content for topic {topic}")

    except Exception as e:
        # Handle errors gracefully
        print(f"An error occurred: {e}")

# Execute the function
lsa_content_analysis()
````

# **Latent Semantic Analysis (LSA) Model Code**

````python
# File Path for the dataset (URLs) and output CSV
FILE_PATH = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/seo_services_urls.txt"
OUTPUT_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/scraped_content.csv"

# Import required libraries
import pandas as pd  # For handling datasets
import requests  # For sending HTTP requests to fetch web pages
from bs4 import BeautifulSoup  # For parsing and extracting text from web pages
from google.colab import drive  # To mount Google Drive and access files

# Step 1: Mount Google Drive
# Explanation:
# - Google Colab is a cloud environment and cannot directly access your local files.
# - We mount Google Drive to access files stored in your Google Drive.
drive.mount('/content/drive')  # This command links Google Drive to your Colab environment.

# Step 2: Function to Scrape Content from a Single URL
# Explanation:
# - This function fetches the content of a webpage using its URL.
# - It extracts only the visible text (e.g., paragraphs) from the webpage.
def scrape_content(url):
    """
    Scrapes visible text content from a webpage given its URL.
    - Fetches the webpage using the requests library.
    - Parses the HTML content using BeautifulSoup and extracts text from <p> tags.
    """
    try:
        # Send an HTTP GET request to fetch the webpage
        response = requests.get(url, timeout=10)  # Timeout ensures we don't wait indefinitely.
        response.raise_for_status()  # Raises an HTTPError for bad responses (e.g., 404, 500).

        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract visible text from all <p> tags (paragraphs)
        text = ' '.join([p.get_text() for p in soup.find_all('p')])

        # Return the cleaned text content
        return text.strip()  # `.strip()` removes any leading/trailing spaces or newlines.

    except Exception as e:
        # If an error occurs (e.g., network error, invalid URL), print the error message.
        print(f"Error scraping {url}: {e}")
        return None  # Return None to indicate that scraping failed.

# Step 3: Main Function to Scrape All URLs and Save to a CSV File
# Explanation:
# - This function reads the list of URLs from a file.
# - It loops through each URL, scrapes the content, and saves the results to a CSV file.
def main():
    # Step 3.1: Load URLs from the file
    # Explanation:
    # - The URLs are stored in a text file, with one URL per line.
    # - We use pandas to read the file into a DataFrame for easier handling.
    try:
        urls = pd.read_csv(FILE_PATH, header=None, names=["URLs"])  # Load URLs into a DataFrame
    except FileNotFoundError:
        # If the file is not found, print an error message and stop the script.
        print("File not found. Please check the FILE_PATH.")
        return  # Stop the function execution.

    # Step 3.2: Initialize a list to store the scraped data
    # Explanation:
    # - This list will hold dictionaries with two keys: "URL" and "Content".
    # - Each dictionary represents one scraped webpage.
    scraped_data = []

    # Step 3.3: Loop through each URL and scrape content
    # Explanation:
    # - For each URL in the DataFrame, scrape the content using the `scrape_content` function.
    for idx, row in urls.iterrows():
        # Print progress (e.g., "Scraping URL 1/100...")
        print(f"Scraping URL {idx + 1}/{len(urls)}: {row['URLs']}")

        # Call the `scrape_content` function to get the webpage content
        content = scrape_content(row["URLs"])

        # If scraping is successful (i.e., content is not None), save the data
        if content:
            scraped_data.append({"URL": row["URLs"], "Content": content})  # Append to the list

        # Step 3.4: Preview the scraped content for the first 20 URLs
        # Explanation:
        # - To verify that the scraping is working, we display a short preview for the first 20 URLs.
        if idx < 20:  # Only preview the first 20 URLs
            print(f"\nPreview of URL {idx + 1}:\n")
            print(content[:500])  # Display the first 500 characters of the content
            print("\n" + "-" * 80 + "\n")  # Separator for better readability

    # Step 3.5: Save the scraped data to a CSV file
    # Explanation:
    # - The scraped data is saved in a CSV file for further analysis.
    # - Each row in the CSV contains the URL and the corresponding scraped content.
    scraped_df = pd.DataFrame(scraped_data)  # Convert the list of dictionaries into a DataFrame
    scraped_df.to_csv(OUTPUT_CSV, index=False)  # Save the DataFrame to a CSV file (no index column)
    print(f"Scraped data saved to {OUTPUT_CSV}")  # Confirm that the file has been saved.

# Step 4: Run the Main Function
# Explanation:
# - The `if __name__ == "__main__"` block ensures that the script runs only when executed directly.
# - This prevents accidental execution if the script is imported as a module.
if __name__ == "__main__":
    main()



# File Paths for Input and Output
SCRAPED_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/scraped_content.csv"
PROCESSED_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/preprocessed_content.csv"

# Import Required Libraries
import pandas as pd  # For handling tabular data
import re  # For text cleaning (e.g., removing special characters)
import nltk  # For natural language processing tasks
from nltk.corpus import stopwords  # To remove common, meaningless words (e.g., "the", "is")
from nltk.stem import WordNetLemmatizer  # To reduce words to their base form (e.g., "running" -> "run")
from google.colab import drive  # To access Google Drive in Colab

# Step 1: Mount Google Drive
# Why this step?
# - Google Colab runs in the cloud and doesn't directly access your files.
# - We mount Google Drive to read and save files stored in your Google Drive.
drive.mount('/content/drive')

# Step 2: Download NLTK Data Files
# Why this step?
# - NLTK is a library for processing text data, but it requires some files (like stopwords and lemmatization rules) to work.
# - We ensure these files are downloaded before running the script.
nltk.download('stopwords')  # Download stopwords for English
nltk.download('wordnet')  # Download WordNet for lemmatization

# Step 3: Initialize NLP Tools
# Explanation:
# - We set up a list of stopwords and initialize the lemmatizer for later use in text cleaning.
stop_words = set(stopwords.words('english'))  # Set of common words to exclude
lemmatizer = WordNetLemmatizer()  # To reduce words to their base form

# Step 4: Function to Clean and Preprocess Text
# Explanation:
# - This function cleans and preprocesses the text to make it suitable for analysis.
# - It removes unnecessary characters, stopwords, and performs lemmatization.

def preprocess_text(text):
    """
    Cleans and preprocesses the given text by:
    1. Removing special characters, punctuation, and numbers.
    2. Converting text to lowercase.
    3. Removing common stopwords.
    4. Performing lemmatization (reducing words to their base form).

    Why this step?
    - Text data from web scraping is often messy (e.g., contains HTML tags, special characters, etc.).
    - Preprocessing standardizes and simplifies the text for easier analysis.
    """
    try:
        # Remove special characters, punctuation, and numbers
        text = re.sub(r'[^a-zA-Z\s]', '', text)  # Keep only letters and spaces

        # Convert text to lowercase for consistency
        text = text.lower()

        # Tokenize the text into words
        words = text.split()

        # Remove stopwords (e.g., "the", "is", "in")
        words = [word for word in words if word not in stop_words]

        # Perform lemmatization to reduce words to their base form
        words = [lemmatizer.lemmatize(word) for word in words]

        # Rejoin the cleaned words into a single string
        return ' '.join(words)

    except Exception as e:
        # If an error occurs during preprocessing, print the error and return None
        print(f"Error during preprocessing: {e}")
        return None

# Step 5: Main Function to Preprocess the Dataset
# Explanation:
# - This function reads the scraped content, preprocesses it, and saves the results to a new file.
def main():
    try:
        # Step 5.1: Load the scraped content
        # Why this step?
        # - The input file contains web content scraped from various URLs.
        # - We need to load this data to process it.
        scraped_data = pd.read_csv(SCRAPED_CSV)  # Load the scraped data into a DataFrame

        # Step 5.2: Preprocess the 'Content' column
        # Why this step?
        # - The 'Content' column contains raw text scraped from webpages, which needs cleaning.
        print("Preprocessing text content...")
        scraped_data['Processed_Content'] = scraped_data['Content'].apply(preprocess_text)  # Apply preprocessing

        # Step 5.3: Save the preprocessed data to a new CSV file
        # Why this step?
        # - The cleaned text is saved for further analysis (e.g., topic modeling, similarity analysis).
        scraped_data.to_csv(PROCESSED_CSV, index=False)  # Save preprocessed data to a CSV file
        print(f"Preprocessed data saved to {PROCESSED_CSV}")  # Confirm that the data has been saved

        # Step 5.4: Display a preview of the preprocessed data
        # Why this step?
        # - To verify that the preprocessing worked as expected.
        print("\nPreview of Preprocessed Data:")
        print(scraped_data[['URL', 'Processed_Content']].head(40))  # Display the first 40 rows

    except FileNotFoundError:
        # If the input file is missing, print an error message
        print("Scraped content file not found. Please check the file path.")
    except Exception as e:
        # Catch any other errors and print them
        print(f"An error occurred: {e}")

# Step 6: Execute the Main Function
# Explanation:
# - This ensures that the script runs when executed directly.
if __name__ == "__main__":
    main()



# Import Required Libraries
import numpy as np  # For numerical computations, such as matrix operations
import pandas as pd  # For handling tabular data
from sklearn.feature_extraction.text import TfidfVectorizer  # For converting text to numerical representation
from sklearn.decomposition import TruncatedSVD  # For Latent Semantic Analysis (LSA)
import seaborn as sns  # For creating heatmaps
import matplotlib.pyplot as plt  # For plotting graphs
from google.colab import drive  # For accessing Google Drive in Colab

# Mount Google Drive
# Why this step?
# - This allows the script to access files stored in your Google Drive, which is essential for reading input files and saving output files.
drive.mount('/content/drive')

# File Paths
# Define paths for input and output files.
PROCESSED_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/preprocessed_content.csv"  # Input file
LSA_RESULTS_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/lsa_results.csv"  # Output for LSA results
ENHANCED_HEATMAP_PATH = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/enhanced_heatmap.png"  # Heatmap output
SIMILARITY_MATRIX_PATH = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/similarity_matrix.npy"  # Similarity matrix
WEBPAGE_MAPPING_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/webpage_mapping.csv"  # Document ID to webpage title mapping

def balanced_lsa_model():
    """
    This function performs Latent Semantic Analysis (LSA) on preprocessed text data.
    It uses a series of steps to extract meaningful topics from the text and visualize the similarity between documents.
    """
    try:
        # Step 1: Load Preprocessed Data
        # Why this step?
        # - This step reads the cleaned and preprocessed text data for further analysis.
        # - The data is assumed to be saved as a CSV file after preprocessing.
        print("Loading preprocessed content...")
        data = pd.read_csv(PROCESSED_CSV)
        print(f"Loaded {data.shape[0]} documents.\n")  # Show the number of documents loaded

        # Step 2: TF-IDF Vectorization
        # Why this step?
        # - Converts text data into a numerical format (TF-IDF matrix) suitable for machine learning.
        # - TF-IDF emphasizes important words in a document relative to a collection of documents.
        print("Vectorizing text data with TF-IDF...")
        vectorizer = TfidfVectorizer(max_features=3000, stop_words='english')  # Limit to 3000 features for efficiency
        tfidf_matrix = vectorizer.fit_transform(data['Processed_Content'])  # Convert text to numerical representation
        feature_names = vectorizer.get_feature_names_out()  # Get the names of the words in the TF-IDF matrix
        print(f"TF-IDF Vectorization Complete. Shape: {tfidf_matrix.shape}\n")  # Print the dimensions of the matrix

        # Step 3: Apply Latent Semantic Analysis (LSA)
        # Why this step?
        # - LSA reduces the high-dimensional TF-IDF matrix into a smaller set of topics (dimensionality reduction).
        # - Each topic captures key themes in the data.
        print("Applying Latent Semantic Analysis (LSA)...")
        num_topics = 20  # Specify the number of topics to extract
        lsa_model = TruncatedSVD(n_components=num_topics, random_state=42)  # Initialize the LSA model
        lsa_matrix = lsa_model.fit_transform(tfidf_matrix)  # Transform the TF-IDF matrix into a topic representation
        print(f"LSA Transformation Complete. Topics Extracted: {num_topics}\n")

        # Step 4: Extract Topics and Save Results
        # Why this step?
        # - Extract and display the top keywords for each topic to understand what each topic represents.
        print("Extracting topics and keywords...")
        topics = []
        for idx, component in enumerate(lsa_model.components_):
            # For each topic, select the top 10 words
            terms = [feature_names[i] for i in component.argsort()[-10:]]
            topics.append(" + ".join(terms))  # Combine the top words into a single string
            print(f"Topic {idx + 1}: {topics[-1]}")  # Print the topic keywords

        # Assign each document its most dominant topic
        data['Dominant_Topic'] = lsa_matrix.argmax(axis=1)  # Find the topic with the highest score for each document
        data['Topic_Keywords'] = data['Dominant_Topic'].apply(lambda x: topics[x])  # Add the keywords for each topic

        # Save the LSA results to a CSV file
        data.to_csv(LSA_RESULTS_CSV, index=False)
        print(f"\nLSA Results saved to: {LSA_RESULTS_CSV}")

        # Step 5: Generate Document-to-Webpage Mapping
        # Why this step?
        # - Maps each document ID to a webpage title for easier interpretation.
        print("\nHandling webpage mapping...")
        webpage_titles = [
            "Homepage", "Blog Page", "Contact Us", "Services Page", "About Us",
            # Add more titles corresponding to your documents...
        ]
        webpage_mapping = pd.DataFrame({
            'Document ID': range(len(webpage_titles)),
            'Webpage Title': webpage_titles[:len(data)]  # Ensure mapping matches the number of documents
        })
        webpage_mapping.to_csv(WEBPAGE_MAPPING_CSV, index=False)
        print(f"Webpage mapping saved to: {WEBPAGE_MAPPING_CSV}")

        # Step 6: Generate Topic Distribution Summary
        # Why this step?
        # - Summarizes the number of documents associated with each topic.
        print("\nTopic Distribution Summary:")
        topic_counts = data['Dominant_Topic'].value_counts()
        for topic, count in topic_counts.items():
            print(f"Topic {topic}: {count} documents")

        # Step 7: Compute and Save Similarity Matrix
        # Why this step?
        # - Computes a similarity matrix to find how similar documents are to one another.
        print("\nGenerating similarity matrix...")
        similarity_matrix = np.dot(lsa_matrix, lsa_matrix.T)  # Compute the similarity between documents
        np.save(SIMILARITY_MATRIX_PATH, similarity_matrix)  # Save the similarity matrix
        print(f"Similarity matrix saved to: {SIMILARITY_MATRIX_PATH}")

        # Step 8: Generate an Enhanced Heatmap
        # Why this step?
        # - Visualizes the similarity matrix as a heatmap for easier analysis of document relationships.
        print("\nGenerating enhanced heatmap...")
        plt.figure(figsize=(12, 10))  # Set the size of the heatmap
        heatmap = sns.heatmap(
            similarity_matrix,
            cmap="RdYlGn",  # Red-to-green color map
            annot=False,  # Disable cell annotations for better readability
            cbar_kws={'label': 'Similarity Score'}  # Add a color bar with a label
        )

        # Add titles and labels to the heatmap
        heatmap.set_title("Document Semantic Similarity Heatmap", fontsize=16)
        heatmap.set_xlabel("Documents", fontsize=12)
        heatmap.set_ylabel("Documents", fontsize=12)

        # Add a footnote explaining the heatmap
        plt.figtext(
            0.5, -0.1,
            "This heatmap shows the similarity between documents based on topics.\n"
            "Red = High Similarity (e.g., Documents on similar topics).\n"
            "Green = Low Similarity (e.g., Documents on unrelated topics).",
            wrap=True, horizontalalignment='center', fontsize=12
        )

        # Save and display the heatmap
        plt.tight_layout()
        plt.savefig(ENHANCED_HEATMAP_PATH)
        plt.show()
        print(f"Enhanced heatmap saved to: {ENHANCED_HEATMAP_PATH}")

    except Exception as e:
        # Print an error message if something goes wrong
        print(f"An error occurred: {e}")

# Execute the function
balanced_lsa_model()



# Import necessary libraries
import pandas as pd  # For working with tabular data
import numpy as np  # For numerical computations
from google.colab import drive  # To access files in Google Drive

# Step 1: Mount Google Drive
# Why this step?
# - To access input and output files stored in Google Drive.
# - This is necessary because the files used in this code are located in a specific folder in the user's Drive.
drive.mount('/content/drive')

# Step 2: File Paths
# Why this step?
# - Define file paths for input and output to ensure the script reads and writes files at the correct locations.
LSA_RESULTS_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/lsa_results.csv"  # Input LSA results
CONTENT_RECOMMENDATIONS_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/content_recommendations.csv"  # Output for recommendations
TOPIC_KEYWORDS_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/topic_keywords.csv"  # Output for topic keywords
CONTENT_GAPS_TXT = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/content_gaps.txt"  # Output for content gap analysis
SIMILARITY_MATRIX_PATH = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/similarity_matrix.npy"  # Input similarity matrix
WEBPAGE_MAPPING_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/webpage_mapping.csv"  # Input webpage mapping

def lsa_content_analysis():
    """
    Perform content analysis using Latent Semantic Analysis (LSA) results.
    The function:
    1. Generates content recommendations for webpages.
    2. Extracts high-value keywords for topics.
    3. Highlights content gaps for improvement.
    """
    try:
        # Step 3: Load LSA Results
        # Why this step?
        # - The LSA results contain processed documents and their dominant topics, essential for generating recommendations and analyzing gaps.
        print("Loading LSA results...")
        data = pd.read_csv(LSA_RESULTS_CSV)
        print(f"Loaded {data.shape[0]} documents.\n")  # Show the number of documents loaded

        # Step 4: Load Webpage Mapping
        # Why this step?
        # - The mapping file links each document (identified by its ID) to its corresponding webpage title for clarity.
        print("Loading webpage mapping...")
        webpage_mapping = pd.read_csv(WEBPAGE_MAPPING_CSV)

        # Step 5: Validate Webpage Mapping
        # Why this step?
        # - Ensure every document in the LSA results has a corresponding title.
        # - If the mapping is incomplete, add placeholders for missing entries.
        if len(webpage_mapping) < data.shape[0]:
            print("Warning: Webpage mapping is incomplete. Adding placeholders for missing entries.")
            missing_entries = [
                {'Document ID': missing_id, 'Webpage Title': f"Document {missing_id}"}
                for missing_id in range(len(webpage_mapping), data.shape[0])
            ]
            webpage_mapping = pd.concat([webpage_mapping, pd.DataFrame(missing_entries)], ignore_index=True)
        print("Webpage mapping loaded successfully.\n")

        # Step 6: Generate Content Recommendations
        # Why this step?
        # - Use the similarity matrix to find the most similar documents for each webpage.
        print("Generating content recommendations...")
        similarity_matrix = np.load(SIMILARITY_MATRIX_PATH)  # Load precomputed similarity matrix

        recommendations = []  # List to store recommendations
        for i, row in enumerate(similarity_matrix):
            # Find the top 5 most similar documents (excluding the document itself)
            similar_docs = row.argsort()[::-1][1:6]  # Sort in descending order and exclude the first element
            similar_titles = []  # List to store titles of similar documents
            for doc in similar_docs:
                # If the document ID is mapped, retrieve its title
                if doc in webpage_mapping['Document ID'].values:
                    title = webpage_mapping.loc[webpage_mapping['Document ID'] == doc, 'Webpage Title'].values[0]
                else:
                    title = f"Document {doc} (Title Missing)"
                similar_titles.append(title)  # Add the title to the list of similar titles
            # Store the recommendations for the current document
            recommendations.append({
                'Webpage_Title': webpage_mapping.loc[webpage_mapping['Document ID'] == i, 'Webpage Title'].values[0],
                'Similar_Webpages': similar_titles
            })

        # Save the recommendations to a CSV file
        recommendations_df = pd.DataFrame(recommendations)
        recommendations_df.to_csv(CONTENT_RECOMMENDATIONS_CSV, index=False)
        print(f"Content recommendations saved to {CONTENT_RECOMMENDATIONS_CSV}\n")

        # Preview the first 5 recommendations in the console
        print("Preview of Content Recommendations:")
        for _, rec in recommendations_df.head(5).iterrows():
            print(f"Webpage: {rec['Webpage_Title']}")
            print(f" - Similar Pages: {', '.join(rec['Similar_Webpages'])}")

        # Step 7: Extract High-Value Keywords
        # Why this step?
        # - Identify the most important keywords for each topic to provide actionable insights for content improvement.
        print("\nExtracting high-value keywords...")
        topic_keywords = data[['Dominant_Topic', 'Topic_Keywords']].drop_duplicates()
        topic_keywords = topic_keywords.sort_values('Dominant_Topic').reset_index(drop=True)

        # Save the keywords to a CSV file
        topic_keywords.to_csv(TOPIC_KEYWORDS_CSV, index=False)
        print(f"Topic keywords saved to {TOPIC_KEYWORDS_CSV}\n")

        # Preview the first 5 topic keywords in the console
        print("Preview of Topic Keywords:")
        print(topic_keywords.head(5).to_string(index=False))

        # Step 8: Highlight Content Gaps
        # Why this step?
        # - Analyze topic distribution to identify topics with insufficient content and suggest improvements.
        print("\nHighlighting content gaps...")
        topic_distribution = data['Dominant_Topic'].value_counts()  # Count documents per topic
        mean_documents_per_topic = topic_distribution.mean()  # Calculate the average number of documents per topic
        content_gaps = topic_distribution[topic_distribution < mean_documents_per_topic]  # Identify underrepresented topics

        # Save the content gaps to a text file
        with open(CONTENT_GAPS_TXT, 'w') as f:
            f.write("Content Gaps Report\n")
            f.write("=" * 50 + "\n")
            f.write(f"Average Documents per Topic: {mean_documents_per_topic:.2f}\n")
            f.write("Topics with fewer than average documents:\n\n")
            for topic, count in content_gaps.items():
                f.write(f"Topic {topic}: {count} documents\n")
        print(f"Content gaps report saved to {CONTENT_GAPS_TXT}\n")

        # Preview content gaps in the console
        print("Preview of Content Gaps:")
        print(f"Average Documents per Topic: {mean_documents_per_topic:.2f}")
        for topic, count in content_gaps.items():
            print(f"Topic {topic}: {count} documents")
            print(f"  Suggestion: Add more content for topic {topic}")

    except Exception as e:
        # Handle errors gracefully
        print(f"An error occurred: {e}")

# Execute the function
lsa_content_analysis()


# Ensure Required Libraries are Installed
# Why this step?
# - To ensure all necessary libraries (Dash, Plotly, WordCloud) are installed before running the code.
!pip install dash dash-table plotly wordcloud

# Import Required Libraries
# Why these libraries?
# - pandas: For reading and manipulating data in tables.
# - numpy: For numerical computations.
# - WordCloud: To generate word clouds for visualizing keywords.
# - dash: To create an interactive web-based dashboard.
# - plotly: For creating interactive charts and graphs.
import pandas as pd
import numpy as np
from wordcloud import WordCloud
import dash
from dash import dcc, html, dash_table
from dash.dependencies import Input, Output
import plotly.express as px

# File Paths for Input and Output Files
# Why these file paths?
# - These paths define where the data is stored and where the outputs (e.g., charts, recommendations) will be saved.
LSA_RESULTS_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/lsa_results.csv"
CONTENT_RECOMMENDATIONS_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/content_recommendations.csv"
TOPIC_KEYWORDS_CSV = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/topic_keywords.csv"
CONTENT_GAPS_TXT = "/content/drive/MyDrive/Dataset For Latent Semantic Analysis Model/content_gaps.txt"

# Step 1: Load Data from Files
# Why this step?
# - To load the results of the LSA model, recommendations, topic keywords, and content gap analysis from previously saved files.
print("Loading LSA results, recommendations, and keywords...")
lsa_results = pd.read_csv(LSA_RESULTS_CSV)  # Load document and topic data.
recommendations = pd.read_csv(CONTENT_RECOMMENDATIONS_CSV)  # Load content recommendations.
topic_keywords = pd.read_csv(TOPIC_KEYWORDS_CSV)  # Load keywords for each topic.

# Step 2: Parse Content Gaps
# Why this step?
# - The content gaps file (text file) needs to be processed into a structured format (like a table) for easy visualization.
print("Parsing content gaps...")
with open(CONTENT_GAPS_TXT, 'r') as f:
    content_gaps = f.readlines()

# Step 3: Process Content Gaps into a DataFrame
# Why this step?
# - To structure the raw content gap data into a table for visualization purposes.
content_gaps_dict = {}
for line in content_gaps[4:]:  # Skip the header lines in the file.
    if "Topic" in line:
        topic, count = line.strip().split(": ")  # Extract topic and count.
        count = int(count.split()[0])  # Extract the numeric count (e.g., '2' from '2 documents').
        content_gaps_dict[topic] = count

# Convert the dictionary to a DataFrame for creating bar charts.
content_gaps_df = pd.DataFrame({
    "Topic": list(content_gaps_dict.keys()),
    "Documents": list(content_gaps_dict.values())
})

# Step 4: Initialize Dash App
# Why this step?
# - Dash is used to create an interactive dashboard for visualizing LSA results, content recommendations, and gaps.
app = dash.Dash(__name__)

# Step 5: Define the Layout of the Dashboard
# Why this step?
# - The layout defines the structure and sections of the dashboard: recommendations, keywords, and gaps.
app.layout = html.Div([
    html.H1("Interactive Dashboard for LSA Results", style={'textAlign': 'center'}),

    # Section 1: Content Recommendations
    html.Div([
        html.H2("Content Recommendations"),
        dash_table.DataTable(
            id='recommendations-table',
            columns=[
                {"name": "Webpage Title", "id": "Webpage_Title"},
                {"name": "Similar Webpages", "id": "Similar_Webpages"}
            ],
            data=recommendations.to_dict('records'),  # Populate the table with recommendations data.
            style_table={'overflowX': 'auto'},
            style_cell={'textAlign': 'left', 'padding': '10px'},
            style_header={'backgroundColor': 'rgb(230, 230, 230)', 'fontWeight': 'bold'}
        )
    ], style={'margin': '20px'}),

    # Section 2: Topic Keywords with Word Cloud
    html.Div([
        html.H2("Topic Keywords"),
        dcc.Dropdown(
            id='topic-dropdown',
            options=[
                {'label': f'Topic {row["Dominant_Topic"]}', 'value': row["Dominant_Topic"]}
                for _, row in topic_keywords.iterrows()
            ],
            placeholder="Select a topic to view its word cloud"
        ),
        dcc.Graph(id='wordcloud')  # Placeholder for the word cloud.
    ], style={'margin': '20px'}),

    # Section 3: Content Gaps Visualization
    html.Div([
        html.H2("Content Gaps"),
        dcc.Graph(
            id='content-gaps-bar',
            figure=px.bar(
                content_gaps_df,
                x='Topic',
                y='Documents',
                title="Content Gaps by Topic",
                labels={'Documents': 'Number of Documents'},
                text='Documents'
            ).update_traces(textposition='outside')  # Show document counts above the bars.
        )
    ], style={'margin': '20px'}),
])

# Step 6: Callback for Word Cloud
# Why this step?
# - The callback updates the word cloud dynamically based on the selected topic in the dropdown.
@app.callback(
    Output('wordcloud', 'figure'),  # The output is the updated word cloud figure.
    [Input('topic-dropdown', 'value')]  # The input is the selected topic from the dropdown menu.
)
def update_wordcloud(selected_topic):
    """
    Updates the word cloud for the selected topic.
    """
    if selected_topic is None:
        return {}

    # Find the keywords for the selected topic.
    topic_keywords_row = topic_keywords[topic_keywords['Dominant_Topic'] == selected_topic].iloc[0]
    keywords = topic_keywords_row['Topic_Keywords']

    # Generate the word cloud.
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(keywords)

    # Convert the word cloud into a Plotly figure.
    fig = px.imshow(wordcloud, template="simple_white")
    fig.update_layout(
        title=f"Word Cloud for Topic {selected_topic}",
        xaxis=dict(showgrid=False, zeroline=False),
        yaxis=dict(showgrid=False, zeroline=False)
    )
    return fig

# Step 7: Run the Dashboard
# Why this step?
# - This starts the web server to display the dashboard in a web browser.
if __name__ == '__main__':
    app.run_server(debug=True, use_reloader=False)
````
