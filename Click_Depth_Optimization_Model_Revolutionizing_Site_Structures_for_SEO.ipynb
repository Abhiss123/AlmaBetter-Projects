### What is Click Depth Optimization?

When you visit a website, you start at the homepage and click through various links to reach other pages. The number of clicks it takes to reach a page from the homepage is called **click depth**.  

- **Optimal Click Depth**: Important pages should be reachable in 1-2 clicks.  
- **Suboptimal Click Depth**: If a page requires 4-5 or more clicks to reach, it becomes harder for users to access and search engines to index effectively.

The **Click Depth Optimization Model** analyzes the website’s current structure and provides **actionable suggestions** to make it simpler, more user-friendly, and better optimized for search engines.

# **2. Metadata Scraping**
**Purpose**: To gather essential information (metadata) for each webpage dynamically.

**What happens in this part:**
- **Loads URLs**: Reads the `urls_list.csv` file created earlier.
- **Scrapes metadata**: For each URL, it fetches:
  - **Title**: Describes what the page is about.
  - **Description**: Provides a summary of the page's content.
  - **Category**: Suggests a category for better organization.
- **Handles missing metadata**: Notes if any metadata is missing and logs errors for debugging.
- **Saves metadata**: Stores the collected information in a file (`metadata.csv`).

**Why this is important**: Metadata improves SEO (Search Engine Optimization) and user experience. For example, a missing title or description can reduce a page's visibility on search engines.

---


```python
import pandas as pd  # Importing pandas library to work with data in tabular format.
import requests  # Importing requests library to make HTTP requests to fetch webpage data.
from bs4 import BeautifulSoup  # BeautifulSoup is used to parse and scrape HTML content from webpages.
import warnings  # Used to handle and suppress unnecessary warnings during the program execution.

# File paths
# These are the file paths where the required datasets and output files are stored.
urls_list_path = '/content/drive/MyDrive/Click Depth Optimization Model/urls_list.csv'  # Input: List of URLs.
metadata_output_path = '/content/drive/MyDrive/Click Depth Optimization Model/metadata.csv'  # Output: Scraped metadata.
log_output_path = '/content/drive/MyDrive/Click Depth Optimization Model/skipped_urls.log'  # Output: Log file for errors.

# Suppress non-critical warnings
# Purpose: Suppress irrelevant warnings from the BeautifulSoup library to keep the program output clean.
warnings.filterwarnings("ignore", category=UserWarning, module='bs4')

try:
    # Step 1: Load the URLs list
    # Purpose: Read the list of URLs from a CSV file to analyze their metadata dynamically.
    # Use Case: Provides the source data (URLs) that we need to process in this script.
    urls_data = pd.read_csv(urls_list_path)  # Load the URLs list into a pandas DataFrame.
    urls_list = urls_data['URL'].tolist()  # Extract the 'URL' column as a list for processing.

    # Step 2: Define a function to scrape metadata dynamically
    # Purpose: Dynamically fetch metadata (title, description, and category) from each webpage.
    # Use Case: This is crucial for analyzing SEO performance and identifying missing metadata.
    def scrape_metadata(url):
        """
        This function fetches the title, description, and category from the given URL.
        If fetching fails, it logs the error and returns default metadata indicating the issue.
        """
        try:
            # Make an HTTP GET request to fetch the webpage content.
            # Timeout is set to 10 seconds to avoid hanging on unresponsive URLs.
            response = requests.get(url, timeout=10)
            if response.status_code == 200:  # Check if the request was successful.
                response.encoding = response.apparent_encoding  # Ensure proper encoding for webpage content.
                soup = BeautifulSoup(response.text, 'lxml')  # Parse the webpage HTML using BeautifulSoup.

                # Extract the title of the page.
                # Use Case: The title is a critical SEO element and provides the first impression of the webpage.
                title = soup.title.string.strip() if soup.title else "Title missing: Add a descriptive title."

                # Extract the meta description of the page.
                # Use Case: The description is used in search engine results to summarize the page content.
                description_tag = soup.find('meta', attrs={'name': 'description'})
                description = (
                    description_tag['content'].strip()
                    if description_tag
                    else "Description missing: Add a meta description to improve SEO."
                )

                # Extract the category (if available) of the page.
                # Use Case: Categorizing pages helps in organizing and optimizing site structure.
                category_tag = soup.find('meta', attrs={'name': 'category'})
                category = (
                    category_tag['content'].strip()
                    if category_tag
                    else "Category missing: Consider categorizing this page for better organization."
                )

                # Return a dictionary containing the scraped metadata for the current URL.
                return {'URL': url, 'Title': title, 'Description': description, 'Category': category}
            else:
                # Handle cases where the HTTP response is not successful.
                # Provide default metadata indicating the failure.
                return {'URL': url, 'Title': 'Failed to fetch', 'Description': 'Failed to fetch', 'Category': 'Failed to fetch'}
        except Exception as e:
            # Log error details for skipped URLs to a log file.
            # Purpose: Maintain a record of URLs that could not be processed due to errors.
            with open(log_output_path, 'a') as log_file:
                log_file.write(f"Failed URL: {url} | Error: {e}\n")
            # Return default metadata indicating the error.
            return {'URL': url, 'Title': 'Error', 'Description': 'Error', 'Category': 'Error'}

    # Step 3: Scrape metadata for all URLs dynamically
    # Purpose: Loop through the list of URLs and collect metadata for each one.
    # Use Case: Automates the process of collecting SEO-relevant information from webpages.
    metadata = [scrape_metadata(url) for url in urls_list]

    # Step 4: Convert metadata to a DataFrame and save it to a CSV file
    # Purpose: Store the collected metadata in a structured format for further analysis.
    # Use Case: Saving the data allows it to be used in later steps or shared with other tools.
    metadata_df = pd.DataFrame(metadata)  # Convert the list of dictionaries into a pandas DataFrame.
    metadata_df.to_csv(metadata_output_path, index=False)  # Save the DataFrame to a CSV file.
    print(f"Metadata saved to: {metadata_output_path}")  # Inform the user that the metadata has been saved.

    # Step 5: Display a preview of the scraped metadata
    # Purpose: Allow the user to verify the scraped data by displaying the first few rows.
    # Use Case: Ensures that the scraping process worked as expected.
    print("Preview of Scraped Metadata:")
    print(metadata_df.head(20))  # Display the first 20 rows of the DataFrame.

except Exception as e:
    # General error handling to catch any unforeseen issues during the execution.
    # Purpose: Make the program robust by handling errors gracefully and informing the user.
    print(f"Error occurred: {e}")

# Additional Step: Load and inspect additional datasets
# Purpose: Load supplementary datasets for inspection and further use in the project.
# Use Case: Helps in understanding the structure and content of related datasets.

# File paths for additional datasets.
pages_file_path = '/content/drive/MyDrive/Click Depth Optimization Model/Pages Of Thatware.csv'
internal_all_file_path = '/content/drive/MyDrive/Click Depth Optimization Model/internal_all.csv'

# Load and inspect the Pages dataset.
# Purpose: Understand the structure and columns in the Pages dataset.
pages_data = pd.read_csv(pages_file_path)
print("Columns in Pages dataset:")  # Display column names to help understand the dataset structure.
print(pages_data.columns)
print("First 20 rows of Pages dataset:")  # Show a sample of the dataset for quick inspection.
print(pages_data.head(20))

# Load and inspect the Internal All dataset.
# Purpose: Understand the structure and columns in the Internal All dataset.
internal_all_data = pd.read_csv(internal_all_file_path)
print("Columns in Internal All dataset:")  # Display column names to help understand the dataset structure.
print(internal_all_data.columns)
print("First 20 rows of Internal All dataset:")  # Show a sample of the dataset for quick inspection.
print(internal_all_data.head(20))
```

# **3. Merging Datasets and Calculating Click Depth**
**Purpose**: To unify all data into a single dataset and calculate the current and suggested click depths.

**What happens in this part:**
- **Loads datasets**: Combines data from `metadata.csv`, `Pages Of Thatware.csv`, and `internal_all.csv`.
- **Merges data**: Aligns information across all datasets based on URLs.
- **Handles missing data**: Adds placeholders for missing metadata (e.g., missing titles).
- **Calculates click depth**: Dynamically computes the current depth of each page based on the structure of its URL.
- **Suggests optimized click depth**: Reduces depth for pages where possible to improve accessibility.
- **Cleans the dataset**: Removes unnecessary columns and logs skipped rows.

**Why this is important**: A unified dataset with calculated click depths is essential for analyzing and improving the website's structure.

---


```python
import pandas as pd  # Importing pandas to handle data in tabular format efficiently.

# File paths
# These paths specify where the input datasets are located and where the output files will be saved.
metadata_path = '/content/drive/MyDrive/Click Depth Optimization Model/metadata.csv'  # File containing metadata for URLs.
pages_path = '/content/drive/MyDrive/Click Depth Optimization Model/Pages Of Thatware.csv'  # File containing page performance data.
internal_all_path = '/content/drive/MyDrive/Click Depth Optimization Model/internal_all.csv'  # File with detailed internal site data.
final_output_path = '/content/drive/MyDrive/Click Depth Optimization Model/unified_dataset.csv'  # Output file for the final dataset.
log_file_path = '/content/drive/MyDrive/Click Depth Optimization Model/merge_skipped_rows.log'  # Log file to track issues during merging.

# Step 1: Load datasets
# Purpose: Load the required datasets into pandas DataFrames so they can be analyzed and processed.
# Use Case: This step prepares all the raw data needed for the rest of the analysis.
metadata = pd.read_csv(metadata_path)  # Load metadata about the URLs.
pages = pd.read_csv(pages_path)  # Load data about page performance (e.g., clicks, impressions).
internal_all = pd.read_csv(internal_all_path)  # Load detailed internal site structure data.

# Step 2: Rename columns to ensure alignment for merging
# Purpose: Ensure that the column names in the datasets match so they can be merged successfully.
# Use Case: Column names like 'Top pages' need to be standardized to 'URL' for consistency.
if 'Top pages' in pages.columns:  # Check if the column name 'Top pages' exists in the dataset.
    pages.rename(columns={'Top pages': 'URL'}, inplace=True)  # Rename 'Top pages' to 'URL' for consistency.

# Step 3: Merge the datasets
# Purpose: Combine the three datasets into a single dataset for unified analysis.
# Use Case: Merging allows us to analyze all the information (metadata, performance, and structure) together.

# Merge metadata with the pages dataset based on the 'URL' column.
metadata_pages_merged = pd.merge(metadata, pages, on='URL', how='left')

# Merge the result with the internal_all dataset based on the 'URL' and 'Address' columns.
final_data = pd.merge(metadata_pages_merged, internal_all, left_on='URL', right_on='Address', how='left')

# Step 4: Flag and handle missing metadata
# Purpose: Fill in missing metadata (e.g., Title, Description, Category) with default values and actionable suggestions.
# Use Case: Missing metadata can harm SEO performance, so this step provides default suggestions to address such issues.
final_data['Title'] = final_data['Title'].fillna("Title missing: Add a descriptive title.")
final_data['Description'] = final_data['Description'].fillna("Description missing: Add a meta description to improve visibility.")
final_data['Category'] = final_data['Category'].fillna("Category missing: Consider categorizing this page.")

# Step 5: Calculate current click depth dynamically
# Purpose: Estimate the click depth of each URL based on its structure (e.g., counting the slashes in the URL).
# Use Case: Click depth is important for understanding how deep a page is in the site's structure.

def calculate_click_depth(url):
    """
    Calculate the click depth of a URL based on the number of slashes in its structure.
    - Assumes that depth is determined by the number of directory levels in the URL.
    - Example: 'https://example.com/section/page' has a depth of 2.
    """
    if isinstance(url, str):  # Check if the URL is a valid string.
        return url.count('/') - 2  # Subtract 2 to ignore the domain-level slashes.
    return 0  # Default depth for invalid URLs.

final_data['Crawl Depth'] = final_data['URL'].apply(calculate_click_depth)  # Apply the function to calculate depth.

# Step 6: Suggest optimized click depth
# Purpose: Suggest an optimized click depth by reducing the current depth by 1, ensuring a minimum depth of 1.
# Use Case: Helps in reorganizing the site structure to improve accessibility and SEO performance.

def suggest_click_depth(current_depth):
    """
    Suggest an optimized click depth for a page.
    - Reduces the depth by 1 while ensuring a minimum depth of 1.
    """
    return max(1, current_depth - 1)  # Ensure the minimum depth is 1.

final_data['Suggested Click Depth'] = final_data['Crawl Depth'].apply(suggest_click_depth)  # Apply the suggestion logic.

# Step 7: Log skipped rows for debugging
# Purpose: Identify and log rows that were skipped during merging due to missing data.
# Use Case: Helps in troubleshooting and fixing issues with the dataset.

skipped_rows = final_data[final_data['URL'].isna()]  # Filter rows with missing 'URL'.
if not skipped_rows.empty:  # Check if there are any skipped rows.
    skipped_rows.to_csv(log_file_path, index=False)  # Save skipped rows to a log file.
    print(f"Skipped rows logged to: {log_file_path}")  # Inform the user about the log.

# Step 8: Clean the dataset
# Purpose: Remove irrelevant columns that do not contribute to the analysis or SEO performance.
# Use Case: Makes the dataset cleaner and easier to work with by removing unnecessary information.

columns_to_drop = ['Grammar Errors', 'Spelling Errors', 'Redirect Type', 'Cookies']  # List of columns to drop.
# Drop columns only if they exist in the dataset.
final_data.drop(columns=[col for col in columns_to_drop if col in final_data.columns], inplace=True)

# Step 9: Save the final unified dataset
# Purpose: Save the processed and unified dataset to a CSV file for further analysis or sharing.
# Use Case: Provides a single dataset containing all the necessary information.
final_data.to_csv(final_output_path, index=False)  # Save the dataset to a CSV file.
print(f"Unified dataset saved to: {final_output_path}")  # Inform the user about the saved file.

# Step 10: Preview the dataset
# Purpose: Display a small sample of the processed dataset to verify that the steps were successful.
# Use Case: Helps in quick validation of the dataset.
print("Preview of Unified Dataset:")
print(final_data.head(20))  # Display the first 20 rows of the dataset.
```

# **4. Generating the Click Depth Report**
**Purpose**: To create a report summarizing the optimization insights.

**What happens in this part:**
- **Generates actionable suggestions**: For each page, it provides:
  - What to improve (e.g., move a page closer to the homepage).
  - The priority of the improvement (High, Medium, Low).
  - The SEO impact of the suggestion.
- **Saves the report**: Stores the detailed suggestions in a file (`click_depth_report.csv`).
- **Prepares data for visualization**: Adds a simplified dataset for visual analysis.

**Why this is important**: The report provides clear, actionable insights for improving the website's structure, making it easier to plan optimizations.

---


```python
import pandas as pd  # Importing pandas to work with tabular data efficiently.

# File paths
# These paths specify the location of input data and where the output will be saved.
unified_dataset_path = '/content/drive/MyDrive/Click Depth Optimization Model/unified_dataset.csv'  # Input: Unified dataset containing URL, metadata, and click depth information.
click_depth_report_path = '/content/drive/MyDrive/Click Depth Optimization Model/click_depth_report.csv'  # Output: File to save the Click Depth Report.

# Step 1: Load the unified dataset
# Purpose: Read the unified dataset that contains all the information about the website's pages, such as URLs, metadata, and calculated click depth.
# Use Case: This step is critical as it provides the base data to generate insights and actionable suggestions.
unified_data = pd.read_csv(unified_dataset_path)  # Load the dataset into a pandas DataFrame.

# Step 2: Define a function to generate actionable suggestions
# Purpose: Create specific suggestions for each webpage to optimize its click depth.
# Use Case: These suggestions help improve the site's structure, making pages more accessible to users and search engines.

def generate_actionable_suggestions(row):
    """
    Generate actionable suggestions to optimize click depth for a webpage.

    Parameters:
    - row: A single row of the DataFrame representing one webpage's data.

    Returns:
    - A tuple containing:
      1. An actionable suggestion for improving the page's click depth.
      2. A priority level (High, Medium, Low) based on how urgent the change is.
      3. SEO impact insights explaining the benefit of the suggestion.
    """
    current_depth = row['Crawl Depth']  # The current depth of the page in the site structure.
    suggested_depth = row['Suggested Click Depth']  # The ideal depth for the page.
    priority = None  # Will store the urgency of the suggested change.
    seo_impact = None  # Will store insights about how the suggestion impacts SEO.

    # Check if the page is too deep in the site's structure
    if current_depth > 2:  # If the page is more than 2 clicks away from the homepage.
        priority = "High"  # High priority because deeper pages are harder to access.
        seo_impact = "Significantly improves accessibility and reduces crawl depth issues."
        return (f"Move '{row['Title']}' closer to the homepage or add internal links from pages at depth {current_depth - 1}.",
                priority, seo_impact)
    # Check if the page is at depth 2 (slightly suboptimal)
    elif current_depth == 2:
        priority = "Medium"  # Medium priority as the page is reasonably accessible but can be improved.
        seo_impact = "Enhances user navigation by providing more direct access."
        return (f"Add internal links to '{row['Title']}' from top-level pages.",
                priority, seo_impact)
    else:  # If the page is at depth 1 (optimal depth)
        priority = "Low"  # Low priority because the page is already in an ideal position.
        seo_impact = "Page is optimally placed for accessibility and user experience."
        return ("Page is optimally placed.", priority, seo_impact)

# Step 3: Apply the function to generate suggestions for all pages
# Purpose: Process each row of the dataset to generate suggestions, priority levels, and SEO insights.
# Use Case: This step creates actionable information that can be used to improve the site's structure.
unified_data[['Actionable Suggestions', 'Priority', 'SEO Impact']] = unified_data.apply(
    lambda row: pd.Series(generate_actionable_suggestions(row)), axis=1
)

# Step 4: Create a Click Depth Report
# Purpose: Extract the relevant columns (e.g., URL, Title, Click Depth, Suggestions) and save them as a separate report.
# Use Case: The report provides a clear and concise overview of improvements needed for the site's structure.
click_depth_report = unified_data[[  # Select only the columns relevant for the report.
    'URL', 'Title', 'Crawl Depth', 'Suggested Click Depth',
    'Actionable Suggestions', 'Priority', 'SEO Impact'
]]

# Save the Click Depth Report to a CSV file
# Purpose: Save the report for sharing with stakeholders or for further analysis.
# Use Case: The saved file acts as documentation of the site's optimization plan.
click_depth_report.to_csv(click_depth_report_path, index=False)  # Save the report without row indices.

# Step 5: Prepare for visualization (Optional)
# Purpose: Prepare a simplified dataset that can be used for creating visualizations, such as bar charts or graphs.
# Use Case: Visualizations help in understanding the distribution of click depths and proposed improvements.
visualization_data = unified_data[['URL', 'Crawl Depth', 'Suggested Click Depth', 'Priority']]  # Select relevant columns for visualization.

# Placeholder for visualization logic
# In the next part of the project, this data can be used to generate visual representations of the site's structure.

# Step 6: Preview the Click Depth Report
# Purpose: Display a sample of the report to verify that it has been created correctly.
# Use Case: Ensures that the report contains accurate and meaningful information.
print("Click Depth Report saved to:", click_depth_report_path)  # Inform the user where the report has been saved.
print("Preview of Click Depth Report:")
print(click_depth_report.head(40))  # Display the first 40 rows of the report for verification.
```

```python
import pandas as pd  # Pandas is used to load and manipulate tabular data.
import matplotlib.pyplot as plt  # Matplotlib is used for creating visualizations like bar charts.
import networkx as nx  # NetworkX is used to represent the site structure as a graph.

# File paths
# These file paths specify the location of the Click Depth Report and where the output visualizations will be generated.
click_depth_report_path = '/content/drive/MyDrive/Click Depth Optimization Model/click_depth_report.csv'

# Step 1: Load the Click Depth Report
# Purpose: Read the report that contains details about the site's pages, such as their current and suggested click depths.
# Use Case: This step prepares the data for further analysis and visualization.
click_depth_data = pd.read_csv(click_depth_report_path)  # Load the CSV file into a pandas DataFrame for easy processing.

# Step 2: Generate Bar Charts for Click Depth
# Purpose: Compare the current structure of the website with the optimized (suggested) structure using visual bar charts.
# Use Case: Visualizing the click depths helps in understanding the distribution of pages and identifying areas for improvement.

# Setting up the plotting area
plt.figure(figsize=(14, 7))  # Create a figure with a specific size for better visibility.

# Subplot 1: Current Click Depth
plt.subplot(1, 2, 1)  # Create the first subplot to show the current site structure.
click_depth_data['Crawl Depth'].value_counts().sort_index().plot(
    kind='bar', alpha=0.8, edgecolor='black'  # Create a bar chart showing the number of pages at each current depth.
)
plt.title('Current Site Structure', fontsize=14)  # Title of the chart.
plt.xlabel('Crawl Depth', fontsize=12)  # Label for the x-axis.
plt.ylabel('Number of Pages', fontsize=12)  # Label for the y-axis.
plt.xticks(fontsize=10)  # Set the font size of the x-axis labels.
plt.yticks(fontsize=10)  # Set the font size of the y-axis labels.
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add a grid for better readability of the chart.

# Subplot 2: Suggested Click Depth
plt.subplot(1, 2, 2)  # Create the second subplot to show the optimized site structure.
click_depth_data['Suggested Click Depth'].value_counts().sort_index().plot(
    kind='bar', color='green', alpha=0.8, edgecolor='black'  # Create a bar chart showing the number of pages at each suggested depth.
)
plt.title('Proposed Optimized Structure', fontsize=14)  # Title of the chart.
plt.xlabel('Suggested Click Depth', fontsize=12)  # Label for the x-axis.
plt.ylabel('Number of Pages', fontsize=12)  # Label for the y-axis.
plt.xticks(fontsize=10)  # Set the font size of the x-axis labels.
plt.yticks(fontsize=10)  # Set the font size of the y-axis labels.
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add a grid for better readability of the chart.

# Adjust layout to avoid overlapping of charts
plt.tight_layout()  # Ensures the charts are properly aligned and there is no overlap.
plt.show()  # Display the bar charts.

# Step 3: Visualize the Site Structure as a Graph
# Purpose: Represent the current structure of the site as a graph where nodes represent pages, and edges represent relationships between pages.
# Use Case: This visualization helps in understanding how pages are connected and identifying any isolated or poorly connected pages.

# Create a directed graph
G = nx.DiGraph()  # A directed graph means the connections (edges) have a direction, representing parent-child relationships between pages.

# Add nodes and edges to the graph
for _, row in click_depth_data.iterrows():  # Loop through each page in the data.
    # Add a node for the current URL with its click depth as an attribute.
    G.add_node(row['URL'], depth=row['Crawl Depth'])
    # If the page is not at the top level (depth > 1), add an edge to its parent page.
    if row['Crawl Depth'] > 1:
        parent_depth = row['Crawl Depth'] - 1  # Calculate the depth of the parent page.
        parent_rows = click_depth_data[click_depth_data['Crawl Depth'] == parent_depth]  # Find the parent page(s) at the calculated depth.
        if not parent_rows.empty:  # If a parent page exists, add a connection (edge) between the parent and current page.
            parent_url = parent_rows['URL'].iloc[0]
            G.add_edge(parent_url, row['URL'])

# Visualize the graph
plt.figure(figsize=(16, 16))  # Set the size of the plot for better visibility of the graph.
pos = nx.spring_layout(G, seed=42)  # Use a spring layout to arrange the nodes in a circular pattern.
nx.draw(
    G,
    pos,
    with_labels=False,  # Do not display the labels (URLs) to avoid clutter.
    node_size=30,  # Set the size of each node.
    node_color='skyblue',  # Set the color of the nodes.
    edge_color='gray',  # Set the color of the edges.
    alpha=0.7  # Make the edges slightly transparent for better readability.
)
plt.title('Site Structure Graph (Current)', fontsize=16)  # Add a title to the graph.
plt.show()  # Display the graph.

# Step 4: Generate Summary Statistics
# Purpose: Provide key numerical insights into the site's structure to help in decision-making.
# Use Case: These metrics give a high-level overview of the current state and potential improvements in the site structure.

# Calculate average current and suggested click depths
average_current_depth = click_depth_data['Crawl Depth'].mean()  # Average depth of pages in the current structure.
average_suggested_depth = click_depth_data['Suggested Click Depth'].mean()  # Average depth of pages in the optimized structure.

# Display the statistics with explanations
print(f"Average Current Click Depth: {average_current_depth:.2f}")  # Show the average current depth.
print(f"Average Suggested Click Depth: {average_suggested_depth:.2f}")  # Show the average optimized depth.

# Additional metrics for better understanding
total_pages = click_depth_data.shape[0]  # Total number of pages in the site.
optimal_depth_pages = click_depth_data[click_depth_data['Suggested Click Depth'] == 1].shape[0]  # Count pages at the optimal depth (1).
print(f"Total Number of Pages: {total_pages}")  # Display the total number of pages.
print(f"Number of Pages at Optimal Depth (Suggested Depth = 1): {optimal_depth_pages}")  # Show how many pages are at the best depth.
print(
    f"Percentage of Pages at Optimal Depth: {(optimal_depth_pages / total_pages) * 100:.2f}%"  # Calculate and display the percentage of pages at optimal depth.
)
```

# **What is the Click Depth Optimization Model?**
The **Click Depth Optimization Model** is a tool designed to improve the **SEO (Search Engine Optimization)** performance of a website by analyzing the structure and arrangement of its web pages. It focuses on **click depth**, which refers to the number of clicks a user needs to reach a specific page from the homepage.

---

### **Understanding the Output**
Here’s a detailed breakdown of the output:

#### **1. URL**
- **What it is:** This column lists the unique URLs (web addresses) of the pages on the website.
- **Use case:** It helps identify all the pages analyzed by the model. Each URL represents a webpage that needs optimization.

#### **2. Title**
- **What it is:** The title of each page, often shown in search results and browser tabs.
- **Use case:** Titles are crucial for SEO and user engagement. Pages with missing or generic titles (e.g., "Title missing: Add a descriptive title") need immediate attention to create compelling, keyword-rich titles.

### **3. Crawl Depth**
#### **What it is:**
Crawl Depth measures how far a webpage is from the homepage, based on the number of clicks needed to access it. For example:
- A crawl depth of 1 means the page is directly accessible from the homepage.
- A crawl depth of 5 means the user or search engine needs to click through five levels of links to find the page.

#### **Use Case:**
- **For Users:** Pages with higher crawl depth are harder to find, leading to a poor user experience. Users may give up if they can't quickly locate important pages.
- **For Search Engines:** Search engines prioritize pages that are easier to reach. If a page is buried too deeply, it may not get indexed properly, lowering its visibility in search results.

#### **Benefit for Website Owners:**
- It identifies pages that are "hidden" deep within the site structure and helps improve their accessibility, ensuring these pages are not overlooked by users or search engines.

#### **Steps to Take:**
1. **Identify Critical Pages:** Look at pages with high crawl depth (e.g., 5 or more). Determine if they are important for your business or SEO.
2. **Reduce Crawl Depth:** Move critical pages closer to the homepage by:
   - Linking them from high-traffic pages.
   - Including them in the main navigation or footer.
3. **Regular Audits:** Keep track of crawl depth to prevent new pages from being buried.

---

### **4. Suggested Click Depth**
#### **What it is:**
The Suggested Click Depth provides a recommendation for the optimal number of clicks needed to access a page. This value is calculated to enhance accessibility while maintaining a logical site structure.

#### **Use Case:**
- It acts as a benchmark for restructuring the site.
- Ensures critical pages are no more than 2-3 clicks away from the homepage for optimal visibility.

#### **Benefit for Website Owners:**
- Improves the likelihood of pages being crawled by search engines.
- Makes it easier for users to navigate the site and find important content.

#### **Steps to Take:**
1. **Follow the Recommendations:** Adjust the site structure to match the suggested click depth.
2. **Optimize Internal Links:** Use internal linking to connect pages at higher depths with pages closer to the homepage.
3. **Test Navigation:** Ensure the restructured navigation is intuitive for users and improves their experience.

---

- "Move 'Title missing: Add a descriptive title' closer to the homepage."

#### **Use Case:**
- **For Website Owners or SEO Teams:** It provides clear, actionable steps to improve site navigation and SEO.
- **For Content Teams:** Highlights areas where metadata (like titles or descriptions) is missing and needs improvement.

#### **Benefit for Website Owners:**
- Offers a step-by-step guide tailored to each page, making it easier to implement changes without guesswork.
- Increases the efficiency of SEO and development teams by focusing efforts where they are most needed.

#### **Steps to Take:**
1. **Implement Suggestions:** Follow the specific steps provided for each page to reduce click depth and improve metadata.
2. **Track Changes:** Monitor the impact of these changes on site traffic and user behavior.
3. **Iterate:** Use the actionable suggestions to continuously refine the site structure.

### **6. Priority**
#### **What it is:**
