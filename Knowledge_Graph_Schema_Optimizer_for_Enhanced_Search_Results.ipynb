````python
import requests
from bs4 import BeautifulSoup
import json

# JSON-LD Schema Markup Code for ThatWare LLP
schema_markup = {
    "@context": "https://schema.org",
    "@type": "Organization",
    "name": "ThatWare LLP",
    "url": "https://thatware.co",
    "logo": "https://thatware.co/wp-content/uploads/2023/03/logo-03.png",
    "contactPoint": {
        "@type": "ContactPoint",
        "telephone": "+91-7044080698",
        "contactType": "Customer Service",
        "email": "info@thatware.co"
    },
    "sameAs": [
        "https://www.facebook.com/thatware.co",
        "https://www.instagram.com/thatware.co/",
        "https://www.linkedin.com/company/thatware-llp/",
        "https://www.youtube.com/TuhinBanika2z"
    ],
    "address": [
        {
            "@type": "PostalAddress",
            "streetAddress": "Arunava Sarani, Sukriti Apartment – G Floor, North Ghosh Para, Bally",
            "addressLocality": "Howrah",
            "postalCode": "711227",
            "addressCountry": "IN"
        },
        {
            "@type": "PostalAddress",
            "streetAddress": "DC Neogi Road, Srijoni Apartment – G Floor, North Ghosh Para, Bally",
            "addressLocality": "Howrah",
            "postalCode": "711227",
            "addressCountry": "IN"
        },
        {
            "@type": "PostalAddress",
            "streetAddress": "GMZ-F01-104, Danube Gemz, Al Asayel St – Al Furjan",
            "addressLocality": "Dubai",
            "addressCountry": "AE"
        },
        {
            "@type": "PostalAddress",
            "streetAddress": "71-75 Shelton Street",
            "addressLocality": "Covent Garden",
            "addressRegion": "London",
            "postalCode": "WC2H 9JQ",
            "addressCountry": "GB"
        }
    ],
    "description": "At ThatWare, we work hard to support businesses of all sizes by providing customized digital marketing and SEO solutions to enhance online visibility and drive growth.",
    "foundingDate": "2018",
    "keywords": [
        "Advanced SEO Services",
        "Local SEO",
        "E-Commerce SEO",
        "Voice SEO",
        "Knowledge Graph Optimization",
        "SEO Consulting",
        "On-Page SEO",
        "Technical SEO",
        "Enterprise SEO",
        "Link Building",
        "SEO Audits",
        "Content Writing",
        "AI SEO Solutions",
        "Brand SEO"
    ]
}

# Convert schema markup to JSON format
json_ld_schema = f'<script type="application/ld+json">\n{json.dumps(schema_markup, indent=2)}\n</script>'

# List of URLs where schema should be added/updated
urls = [
    'https://thatware.co/',
    'https://thatware.co/services/',
    'https://thatware.co/advanced-seo-services/',
    'https://thatware.co/digital-marketing-services/',
    'https://thatware.co/business-intelligence-services/',
    'https://thatware.co/link-building-services/',
    'https://thatware.co/branding-press-release-services/',
    'https://thatware.co/conversion-rate-optimization/',
    'https://thatware.co/social-media-marketing/',
    'https://thatware.co/content-proofreading-services/',
    'https://thatware.co/website-design-services/',
    'https://thatware.co/web-development-services/',
    'https://thatware.co/app-development-services/',
    'https://thatware.co/website-maintenance-services/',
    'https://thatware.co/bug-testing-services/',
    'https://thatware.co/software-development-services/',
    'https://thatware.co/competitor-keyword-analysis/'# Add all other URLs as provided
]

def inject_schema(url):
    """
    Function to inject or update JSON-LD schema markup on a webpage.
    """
    try:
        # Step 1: Request the webpage content
        response = requests.get(url)
        response.raise_for_status()  # Check if request was successful

        # Step 2: Parse the webpage HTML content with BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Step 3: Check if JSON-LD script already exists
        existing_schema = soup.find('script', type="application/ld+json")

        if existing_schema:
            # Step 4: Update existing schema if found
            print(f"Schema exists on {url}, updating it.")
            existing_schema.string = json.dumps(schema_markup, indent=2)
        else:
            # Step 5: Inject new schema if none exists
            print(f"No schema found on {url}, injecting new schema.")
            new_schema_tag = soup.new_tag("script", type="application/ld+json")
            new_schema_tag.string = json.dumps(schema_markup, indent=2)
            soup.head.append(new_schema_tag)  # Append the schema to the <head> section

        # (Simulated Step) Normally, you would save the updated HTML back to the server.
        # Here we simply print the updated HTML to show the changes.
        print(f"Updated HTML for {url}:\n", soup.prettify())

    except requests.RequestException as e:
        print(f"Error accessing {url}: {e}")
    except Exception as e:
        print(f"An error occurred with {url}: {e}")

# Loop through each URL and apply schema injection
for url in urls:
    inject_schema(url)
````
