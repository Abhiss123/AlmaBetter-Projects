#### **What are Attention Mechanisms?**

**Attention Mechanisms** are a concept used in **machine learning and artificial intelligence (AI)** that allows models (like transformers) to **"pay attention"** to the most important parts of the input data. Imagine reading a long article—your brain naturally focuses more on specific sentences or words to understand the main point. Similarly, Attention Mechanisms help a model focus on the most relevant parts of the text, which improves its understanding and output.

#### **Why are Attention Mechanisms Important?**

These mechanisms are crucial for dealing with complex data because they allow models to weigh the significance of each part of the data. This "attention" leads to better **content generation**, like generating human-like text, and better understanding of **keyword relevance**, which means identifying the most important words or phrases in a given context.

````python
# Import necessary libraries for working with NLP models, text processing, and web scraping

# `transformers` library provides pre-trained BERT model and tokenizer
# - BERT (Bidirectional Encoder Representations from Transformers) is used to analyze text.
# - It is pre-trained to understand relationships between words and context.
from transformers import BertModel, BertTokenizer

# `torch` library provides tools for performing tensor operations.
# - A tensor is like a multi-dimensional array used for calculations.
# - It's required for handling inputs and outputs with BERT.
import torch

# `requests` library is used for sending HTTP requests to websites.
# - We use it to fetch the raw HTML content of a webpage given its URL.
import requests

# `BeautifulSoup` is a library used to parse and extract readable text from raw HTML content.
# - HTML content is the code that defines a webpage's structure.
# - BeautifulSoup allows us to only extract the text, ignoring tags like `<div>`, `<p>`, etc.
from bs4 import BeautifulSoup

# `matplotlib.pyplot` is a plotting library used for visualizing data.
# - We use it to create bar charts to visualize the attention scores of words.
# - It makes it easy to understand which words are more significant.
import matplotlib.pyplot as plt

# `pandas` is a data manipulation library used for managing tabular data.
# - We use it to create DataFrames (like tables) and export data to CSV files.
# - A DataFrame is similar to an Excel sheet where we can store data in rows and columns.
import pandas as pd

# `os` module provides functions for interacting with the operating system.
# - It allows us to create directories, check file paths, and manage folders.
import os

# `re` is the regular expressions library for text cleaning.
# - It helps find and replace unwanted characters like special symbols, numbers, etc.
import re

# `nltk` (Natural Language Toolkit) is a library for various language processing tasks.
# - It includes tools like stopwords removal, tokenization, and more.
# - Here, we use it to remove common words that don't add much meaning (e.g., 'the', 'is').
import nltk

# Step 1: Download the stopwords from the NLTK library.
# Stopwords are common words like 'a', 'an', 'the', which don't contribute much to the context.
# - Example: In the sentence "The cat sat on the mat", 'the' and 'on' are stopwords.
# Downloading these stopwords ensures we have the list available for our text cleaning.
nltk.download('stopwords')


# Define lists of common stopwords and unwanted symbols for text cleaning
# Stopwords are common words like 'a', 'the', 'and', which don't add value to the content.
stopwords = set([
    'a', 'an', 'the', 'and', 'or', 'if', 'in', 'on', 'at', 'to', 'from', 'with',
    'that', 'this', 'these', 'those', 'it', 'is', 'are', 'was', 'were', 'of',
    'for', 'by', 'be', 'has', 'have', 'not', 'but', 'as', 'so', 'now', 'also'
])

# Unwanted symbols: Characters like '@', '#', digits, etc., which are not meaningful in most contexts.
unwanted_symbols = set(['&', '<', '>', '(', ')', '*', '@', '#', '!', '?', '%', '-', '_', '=', '+', '[', ']', '{', '}', '/', '\\', ':', ';', ',', '.', '1', '2', '3', '4', '5', '6', '7', '8', '9', '0'])

# Define a function to fetch and clean text content from a webpage URL
def fetch_webpage_text(url):
    """
    This function is used to retrieve text content from a webpage and clean it.
    The purpose of cleaning is to remove unwanted symbols, characters, and common words that don’t add much meaning (stopwords).
    This will help in extracting only the relevant information from the webpage.

    Parameters:
    url (str): The URL of the webpage to fetch text from.

    Returns:
    str: Cleaned and filtered text content of the webpage.
    """

    try:
        # Step 1: Send a GET request to the webpage to fetch its HTML content.
        # Explanation:
        # This line of code sends an HTTP request to the provided URL.
        # It asks the server to send back the HTML content of the webpage, which will be stored in the 'response' variable.
        # If the webpage is accessible, it returns the content; otherwise, it throws an error.
        response = requests.get(url)

        # Step 2: Use BeautifulSoup to parse the HTML and extract readable text from the webpage.
        # Explanation:
        # BeautifulSoup is a library that helps in parsing (analyzing) the HTML content to extract only the readable text.
        # This removes all the unnecessary HTML tags (like <div>, <p>, etc.), so only the human-readable content is left.
        # The 'soup' variable will hold the cleaned HTML content.
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract text from the HTML content and remove excessive whitespaces.
        # Explanation:
        # `soup.get_text()` extracts only the text part from the HTML content.
        # `separator=' '` ensures that paragraphs are separated by spaces and not clumped together.
        # `strip=True` removes any leading or trailing spaces around the text.
        raw_text = soup.get_text(separator=' ', strip=True)

        # Step 3: Clean the extracted text using regular expressions to remove unwanted symbols.
        # Explanation:
        # Regular expressions (`re.sub`) are used to define a pattern for unwanted symbols and characters.
        # In this case, the pattern `[^A-Za-z\s]` means:
        #   - Keep only letters (A-Z, a-z) and spaces.
        #   - Remove all other symbols like '@', '#', digits (0-9), etc.
        # The result is stored in 'cleaned_text'.
        cleaned_text = re.sub(r'[^A-Za-z\s]', '', raw_text)

        # Convert the text to lowercase for uniformity.
        # Explanation:
        # To ensure uniformity, the entire text is converted to lowercase letters.
        # Example: If we have both 'SEO' and 'seo' in the text, converting to lowercase treats them as the same word.
        cleaned_text = cleaned_text.lower()

        # Step 4: Split the text into individual words for filtering.
        # Explanation:
        # The text is split into a list of individual words.
        # Example: "This is a test" will become ['this', 'is', 'a', 'test'].
        # This makes it easier to filter out unwanted words and symbols in the next step.
        words = cleaned_text.split()

        # Step 5: Remove stopwords and unwanted symbols from the text using the provided method.
        # Explanation:
        # 'stopwords' is a list of common words like 'a', 'the', 'and' that are very frequent but don’t add much meaning.
        # 'unwanted_symbols' is a list of symbols like '@', '#', digits, etc., that are not useful.
        # The filtering step checks each word, and if it’s not in 'stopwords' and not an unwanted symbol, it keeps it.
        filtered_words = [word for word in words if word not in stopwords and word not in unwanted_symbols]

        # Step 6: Join the cleaned words back into a single string and return it.
        # Explanation:
        # The cleaned words are joined back into a single sentence using spaces.
        # Example: ['this', 'is', 'test'] will become "this is test".
        # This cleaned text is then returned as the final output of the function.
        return ' '.join(filtered_words)

    except Exception as e:
        # Print an error message if something goes wrong during the fetching or cleaning process.
        # Explanation:
        # If the URL is invalid, or there’s a problem with the internet connection, this block will handle the error.
        # Instead of breaking the code, it will print an error message and return an empty string.
        print(f"Error fetching or cleaning content from {url}: {e}")
        return ""  # Return an empty string if there is an error

# List of URLs to analyze (all the provided URLs)
# These URLs represent different pages of the website to extract text for analysis.
urls = [
    'https://thatware.co/',
    'https://thatware.co/services/',
    'https://thatware.co/advanced-seo-services/',
    'https://thatware.co/digital-marketing-services/',
    'https://thatware.co/business-intelligence-services/',
    'https://thatware.co/link-building-services/',
    'https://thatware.co/branding-press-release-services/',
    'https://thatware.co/conversion-rate-optimization/',
    'https://thatware.co/social-media-marketing/',
    'https://thatware.co/content-proofreading-services/',
    'https://thatware.co/website-design-services/',
    'https://thatware.co/web-development-services/',
    'https://thatware.co/app-development-services/',
    'https://thatware.co/website-maintenance-services/',
    'https://thatware.co/bug-testing-services/',
    'https://thatware.co/software-development-services/',
    'https://thatware.co/competitor-keyword-analysis/'
]

# Fetch and clean text content for each URL in the list.
# Store the cleaned text in a dictionary where the keys are URLs and values are the cleaned text.
webpage_texts = {url: fetch_webpage_text(url) for url in urls}

# Load a pre-trained BERT model and its corresponding tokenizer
# BERT uses attention mechanisms to analyze text and identify the importance of each word.
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased', output_attentions=True)

# Define a function to compute attention scores for each word in a given text using the BERT model
def get_attention_scores(text, tokenizer, model):
    """
    This function calculates the attention scores for each word in the input text using BERT.
    Attention scores represent how much focus (or attention) the model gives to each word in the text.
    This helps us understand which words are considered more important by the model.

    Parameters:
    text (str): The cleaned input text for which we want to compute attention scores.
    tokenizer (BertTokenizer): The BERT tokenizer to convert text into smaller units called tokens.
    model (BertModel): The pre-trained BERT model to compute attention scores for each word.

    Returns:
    dict: A dictionary where each word (or token) is mapped to its computed attention score.
    """

    # Step 1: Tokenize the input text and prepare it for the BERT model
    # Explanation:
    # The tokenizer breaks down the text into smaller pieces called tokens, which are easier for the model to process.
    # 'return_tensors='pt'' converts the tokens into a PyTorch tensor format, which is the input format required by BERT.
    # 'truncation=True' ensures that if the text is too long, it will be cut down to a maximum length specified by 'max_length'.
    inputs = tokenizer(text, return_tensors='pt', truncation=True, max_length=128)  # Tokenize the input text into BERT-readable format.

    # Step 2: Run the BERT model to compute attention scores without calculating gradients
    # Explanation:
    # 'torch.no_grad()' disables gradient calculations since we are only using the model for analysis, not training.
    # We pass the tokenized inputs into the BERT model, and the model returns various outputs, including attention scores.
    with torch.no_grad():  # Disable gradient calculation to save computation power and memory.
        outputs = model(**inputs)  # Get the outputs from the model using the tokenized inputs.
        attentions = outputs.attentions  # Extract attention scores from the model's outputs.

    # Step 3: Convert token IDs back to readable words
    # Explanation:
    # After processing the text, the model outputs token IDs (unique numbers representing each word).
    # We convert these IDs back to their corresponding readable words or tokens using the tokenizer's method.
    tokens = tokenizer.convert_ids_to_tokens(inputs['input_ids'][0])  # Convert token IDs back to human-readable words.

    # Step 4: Calculate average attention scores for each token across all heads in the last layer of BERT
    # Explanation:
    # BERT uses multiple "attention heads" in each layer to focus on different parts of the text.
    # We take the average attention scores across all these heads to get a single value for each token.
    avg_attention = attentions[-1].mean(dim=1)  # Compute average attention scores across all attention heads.

    # Step 5: Calculate the mean attention score for each token in the input text
    # Explanation:
    # After calculating the average across heads, we also take the mean value for each token.
    # This gives us a final attention score for each token, indicating its importance in the context of the entire text.
    mean_attention_scores = avg_attention[0].mean(dim=0)  # Get the mean attention score for each token.

    # Step 6: Create a dictionary to map each token (word) to its corresponding attention score
    # Explanation:
    # This step pairs each token (word) with its calculated attention score and stores it in a dictionary.
    # Example: {'word1': 0.002, 'word2': 0.005, ...}
    return {tokens[i]: mean_attention_scores[i].item() for i in range(len(tokens))}

# Calculate and store attention scores for each webpage's cleaned text content
# Explanation:
# This loop goes through each webpage's cleaned text, calculates attention scores using the above function,
# and stores the results in the 'attention_scores' dictionary, with the URL as the key.
attention_scores = {url: get_attention_scores(text, tokenizer, model) for url, text in webpage_texts.items() if text}

# Save the computed attention scores into separate CSV files
# Explanation:
# This part of the code ensures that all the calculated attention scores are saved into CSV files.
# Each file is named based on the URL of the webpage to make it easy to identify which scores belong to which page.
output_folder = "/mnt/data/attention_scores"  # Specify the folder path to save CSV files.

# Step 1: Create the directory if it doesn’t already exist
# Explanation:
# This line of code checks if the folder specified by 'output_folder' exists. If not, it creates the folder.
# The purpose is to ensure that we have a place to store our output CSV files.
os.makedirs(output_folder, exist_ok=True)  # Create the directory if it doesn’t exist (won’t throw an error if it already exists).



# Loop through each URL and its corresponding attention scores in the attention_scores dictionary.
# This dictionary contains URLs as keys and attention scores as values.
for url, scores in attention_scores.items():

    # Step 1: Convert the keys of the dictionary (which are words or tokens) into a list.
    # This 'words' list will contain all the individual words that were analyzed for the current URL.
    words = list(scores.keys())

    # Step 2: Convert the values of the dictionary (which are the attention scores for each word) into a list.
    # This 'score_values' list will contain the corresponding attention scores for each word in the 'words' list.
    score_values = list(scores.values())

    # Step 3: Create a Pandas DataFrame to organize the data.
    # A DataFrame is like a table (similar to an Excel sheet) where each column represents a specific type of data.
    # Here, we create two columns: 'Word' and 'Attention Score'.
    # 'Word' column contains the list of words, and 'Attention Score' column contains the corresponding scores.
    df = pd.DataFrame({'Word': words, 'Attention Score': score_values})

    # Step 4: Construct a valid filename for each URL to save its data as a CSV file.
    # The filename is created by taking the URL, removing the 'https://' or 'http://' part, and replacing slashes with underscores.
    # Example: For 'https://thatware.co/services/', the filename would be 'thatware.co_services_.csv'.
    # This ensures that each file has a unique name based on its URL.
    filename = os.path.join(output_folder, url.split('//')[1].replace('/', '_') + ".csv")

    # Step 5: Save the DataFrame as a CSV file in the specified directory ('output_folder').
    # The 'index=False' parameter means that the row numbers won't be saved as a separate column in the CSV.
    # Example: The DataFrame will look like:
    # | Word    | Attention Score |
    # | ------- | --------------- |
    # | keyword | 0.003           |
    # | seo     | 0.005           |
    df.to_csv(filename, index=False)

    # Step 6: Print a message to confirm that the attention scores for the given URL have been saved.
    # This message helps the user know that the file has been successfully created.
    print(f"Attention scores for {url} saved to {filename}")


# Step 1: Select a sample URL from the list of URLs to analyze
# Here, we choose the first URL in the 'urls' list (i.e., urls[0]) to create a sample visualization.
# This sample URL will be used to generate a bar chart showing the attention scores for its words.
sample_url = urls[0]  # Select the first URL in the list

# Step 2: Check if attention scores are available for the selected sample URL
# We verify if the selected URL (sample_url) has corresponding attention scores in the 'attention_scores' dictionary.
# If the URL exists in the dictionary, proceed to extract and process its attention scores.
if sample_url in attention_scores:

    # Step 3: Extract the attention scores for the selected sample URL
    # Retrieve the attention scores for the sample URL, stored in the form of a dictionary where:
    # - Keys represent words (tokens), and
    # - Values represent their corresponding attention scores.
    sample_scores = attention_scores[sample_url]  # Get the attention scores for the sample URL

    # Step 4: Convert the words (keys) and their scores (values) into separate lists
    # The 'sample_words' list will contain all the words analyzed for the selected URL.
    # The 'sample_score_values' list will contain the corresponding attention scores for each word.
    sample_words = list(sample_scores.keys())  # Convert words (keys) to a list
    sample_score_values = list(sample_scores.values())  # Convert scores (values) to a list

    # Step 5: Create a Pandas DataFrame to structure the extracted data
    # A DataFrame is like a table (similar to an Excel sheet) that organizes data into columns.
    # Here, we create two columns:
    # - 'Word' (contains the words analyzed for the sample URL)
    # - 'Attention Score' (contains the attention scores corresponding to each word)
    sample_df = pd.DataFrame({'Word': sample_words, 'Attention Score': sample_score_values})

    # Step 6: Save the DataFrame as a CSV file in the specified directory
    # The CSV file will be stored at '/mnt/data/sample_attention_scores.csv'.
    # This file will contain the words and their respective attention scores, which can be viewed later.
    sample_df.to_csv("/mnt/data/sample_attention_scores.csv", index=False)  # Save as CSV file

    # Step 7: Print a confirmation message indicating that the sample CSV file has been saved successfully
    # This message tells the user that the CSV file for the sample URL has been created and is ready for download or viewing.
    print(f"Sample attention scores saved to /mnt/data/sample_attention_scores.csv")

    # Step 8: Print the DataFrame to display the saved data
    # This step shows a preview of the DataFrame created, so the user can see the structure of the saved data.
    print(sample_df)

    # Step 9: Plot the attention scores for the top 20 words in the sample URL
    # This block of code creates a bar chart to visually display the top 20 words with the highest attention scores.
    # It helps the user quickly identify the most important words on the sample webpage based on the model's analysis.
    plt.figure(figsize=(15, 5))  # Create a new figure for the plot with a size of 15 units wide and 5 units high.

    # Step 10: Generate a bar chart using the top 20 words and their attention scores
    # 'sample_words[:20]' selects the first 20 words from the 'sample_words' list.
    # 'sample_score_values[:20]' selects the first 20 corresponding attention scores.
    plt.bar(sample_words[:20], sample_score_values[:20])  # Create a bar chart using words as x-axis and their scores as y-axis.

    # Step 11: Label the x-axis to show that it represents the words in the sample URL
    plt.xlabel('Words')  # Set the label for the x-axis as 'Words'.

    # Step 12: Label the y-axis to show that it represents the attention scores for each word
    plt.ylabel('Attention Score')  # Set the label for the y-axis as 'Attention Score'.

    # Step 13: Add a title to the bar chart for better understanding
    # The title dynamically includes the sample URL being analyzed, so the user knows which webpage the words belong to.
    plt.title(f'Attention Scores for Sample URL: {sample_url}')  # Set the title of the plot with the sample URL.

    # Step 14: Rotate the x-axis labels (words) by 45 degrees for better readability
    # This helps prevent overlapping text labels, making it easier to read the words on the x-axis.
    plt.xticks(rotation=45)  # Rotate the words on the x-axis by 45 degrees.

    # Step 15: Display the plot to the user
    # This command renders the bar chart with the attention scores for the top 20 words.
    plt.show()  # Display the bar chart with the attention scores for the top 20 words.

else:
    # Step 16: If the sample URL does not have any attention scores, print a message
    # This indicates that the attention scores for the selected URL were not found.
    print(f"No attention scores available for {sample_url}.")
````

### Explanation of the Output and Guidance

#### 1. **Understanding the Bar Chart:**
   - The **X-axis** represents individual **words** (or tokens) in the input text. In the chart, some words have special symbols like `##` before them. These indicate sub-words or segments of words because BERT sometimes breaks down complex words into smaller tokens.
   - The **Y-axis** shows the **Attention Score** assigned to each word, ranging from 0 to a maximum value (in this case, around 0.0075).
   - Each bar corresponds to a specific word, and the height of the bar reflects the **attention score**. A higher bar means that the word is given more importance by the model in understanding the content.

#### 2. **Attention Scores in Detail:**
   - **`[CLS]` and `[SEP]` Tokens**: These special tokens are part of the BERT model's input format.
     - **`[CLS]`** is used to indicate the beginning of the text.
     - **`[SEP]`** indicates the end of a sentence or text. These tokens have higher attention scores because they are used as markers by the BERT model to structure the input.
   - **Top Attention Scores**: Words like **"powered", "seo", and "google"** have relatively **high attention scores.** This means the BERT model found these words to be more relevant in the given context.
   - **Low Attention Scores**: Words like **"that" and "managed"** have **low attention scores**, indicating they are less important in the context of the analyzed text.

#### 3. **Interpreting the DataFrame Output:**
